````python
# Demo: Docstring quality checker
import re
import inspect

class DocstringChecker:
    """Check docstring quality for vLLM code."""
    
    def __init__(self):
        self.issues = []
    
    def check_function(self, func):
        """Check a function's docstring."""
        self.issues = []
        name = func.__name__
        doc = func.__doc__
        sig = inspect.signature(func)
        
        # Check 1: Has docstring
        if not doc:
            self.issues.append(f"Function '{name}' has no docstring")
            return self.issues
        
        # Check 2: Has summary line
        lines = doc.strip().split('\n')
        if not lines[0].strip():
            self.issues.append("Missing summary line")
        elif len(lines[0].strip()) > 80:
            self.issues.append("Summary line too long (> 80 chars)")
        
        # Check 3: All parameters documented
        params = [p for p in sig.parameters if p != 'self' and p != 'cls']
        doc_lower = doc.lower()
        for param in params:
            if param not in doc:
                self.issues.append(f"Parameter '{param}' not documented")
        
        # Check 4: Has return documentation (if function returns something)
        source = inspect.getsource(func)
        if 'return ' in source and source.count('return') > source.count('return None'):
            if 'returns' not in doc_lower and 'return' not in doc_lower:
                self.issues.append("Missing return value documentation")
        
        # Check 5: Has example
        if 'example' not in doc_lower and '>>>' not in doc:
            self.issues.append("Missing usage example (nice to have)")
        
        return self.issues
    
    def print_report(self, func):
        issues = self.check_function(func)
        name = func.__name__
        if issues:
            print(f"[NEEDS WORK] {name}: {len(issues)} issue(s)")
            for issue in issues:
                print(f"  - {issue}")
        else:
            print(f"[GOOD] {name}: docstring is complete")

# Test functions with varying docstring quality
def bad_function(x, y, z):
    return x + y + z

def ok_function(x, y):
    """Add two numbers."""
    return x + y

def good_function(x, y):
    """Add two numbers together.
    
    Args:
        x: The first number.
        y: The second number.
    
    Returns:
        The sum of x and y.
    
    Example:
        >>> good_function(1, 2)
        3
    """
    return x + y

checker = DocstringChecker()
print("Docstring Quality Check")
print("=" * 50)
for func in [bad_function, ok_function, good_function, google_style_example]:
    checker.print_report(func)
    print()
````

## 3. Creating Example Scripts

````python
# Demo: Structure of a good example script

example_script = '''#!/usr/bin/env python3
"""Example: Offline inference with vLLM.

This example demonstrates how to use vLLM for offline batch inference.
It loads a model, generates completions for multiple prompts, and
displays the results.

Requirements:
    - vLLM installed: pip install vllm
    - GPU with at least 4GB memory
    - Internet connection (to download the model)

Usage:
    python offline_inference.py
    python offline_inference.py --model meta-llama/Llama-2-7b-hf
    python offline_inference.py --model facebook/opt-125m --max-tokens 50
"""

import argparse
from vllm import LLM, SamplingParams


def parse_args():
    """Parse command-line arguments."""
    parser = argparse.ArgumentParser(
        description="Offline inference example with vLLM"
    )
    parser.add_argument(
        "--model",
        type=str,
        default="facebook/opt-125m",
        help="HuggingFace model name or path (default: facebook/opt-125m)",
    )
    parser.add_argument(
        "--max-tokens",
        type=int,
        default=100,
        help="Maximum number of tokens to generate (default: 100)",
    )
    parser.add_argument(
        "--temperature",
        type=float,
        default=0.8,
        help="Sampling temperature (default: 0.8)",
    )
    return parser.parse_args()


def main():
    """Main function demonstrating offline inference."""
    args = parse_args()

    # Step 1: Define sample prompts
    prompts = [
        "Hello, my name is",
        "The president of the United States is",
        "The capital of France is",
        "The future of AI is",
    ]

    # Step 2: Configure sampling parameters
    sampling_params = SamplingParams(
        temperature=args.temperature,
        top_p=0.95,
        max_tokens=args.max_tokens,
    )

    # Step 3: Initialize the LLM
    print(f"Loading model: {args.model}")
    llm = LLM(model=args.model)

    # Step 4: Generate completions
    print("Generating completions...")
    outputs = llm.generate(prompts, sampling_params)

    # Step 5: Display results
    print("\\n" + "=" * 60)
    for output in outputs:
        prompt = output.prompt
        generated_text = output.outputs[0].text
        print(f"\\nPrompt:    {prompt}")
        print(f"Generated: {generated_text}")
        print("-" * 40)


if __name__ == "__main__":
    main()
'''

print("Example Script Template:")
print("=" * 60)
print(example_script)
````

## 4. API Documentation with Type Hints

````python
# Demo: Type hints for comprehensive API documentation

from typing import (
    List, Dict, Optional, Union, Tuple, 
    Sequence, Iterator, Any, Callable, TypeVar
)
from dataclasses import dataclass, field
from enum import Enum

class FinishReason(Enum):
    """Reason why generation stopped.
    
    Attributes:
        STOP: Generation stopped due to stop token or stop string.
        LENGTH: Generation stopped due to reaching max_tokens.
        ABORT: Generation was aborted by the user.
    """
    STOP = "stop"
    LENGTH = "length"
    ABORT = "abort"

@dataclass
class SamplingParamsExample:
    """Parameters for controlling text generation sampling.
    
    This class defines how the model selects the next token at each
    generation step. It supports greedy decoding (temperature=0),
    random sampling, and various filtering strategies.
    
    Args:
        temperature: Controls randomness. 0 = greedy, higher = more random.
            Must be >= 0. Default: 1.0.
        top_p: Nucleus sampling threshold. Only tokens with cumulative
            probability >= top_p are considered. Must be in (0, 1].
            Default: 1.0.
        top_k: Only sample from the top-k most likely tokens.
            -1 means no restriction. Default: -1.
        max_tokens: Maximum number of tokens to generate.
            None means use model's default. Default: 16.
        stop: List of strings that stop generation when encountered.
            Default: empty list.
        presence_penalty: Penalize tokens that have appeared in the output.
            Range: [-2.0, 2.0]. Default: 0.0.
        frequency_penalty: Penalize tokens based on their frequency.
            Range: [-2.0, 2.0]. Default: 0.0.
        logprobs: Number of log-probabilities to return per token.
            None means don't return logprobs. Default: None.
        seed: Random seed for reproducible sampling.
            None means non-deterministic. Default: None.
    
    Example:
        >>> # Greedy decoding
        >>> params = SamplingParams(temperature=0)
        
        >>> # Creative sampling
        >>> params = SamplingParams(temperature=1.2, top_p=0.9)
        
        >>> # Deterministic sampling
        >>> params = SamplingParams(temperature=0.7, seed=42)
    """
    temperature: float = 1.0
    top_p: float = 1.0
    top_k: int = -1
    max_tokens: Optional[int] = 16
    stop: List[str] = field(default_factory=list)
    presence_penalty: float = 0.0
    frequency_penalty: float = 0.0
    logprobs: Optional[int] = None
    seed: Optional[int] = None
    
    def __post_init__(self):
        """Validate parameters after initialization."""
        if self.temperature < 0:
            raise ValueError(f"temperature must be >= 0, got {self.temperature}")
        if not 0 < self.top_p <= 1:
            raise ValueError(f"top_p must be in (0, 1], got {self.top_p}")

@dataclass
class CompletionOutput:
    """Output of a single completion.
    
    Attributes:
        index: The index of this output in the request.
        text: The generated text.
        token_ids: The generated token IDs.
        finish_reason: Why generation stopped.
        logprobs: Log-probabilities for each token (if requested).
    """
    index: int
    text: str
    token_ids: List[int]
    finish_reason: Optional[FinishReason] = None
    logprobs: Optional[List[Dict[int, float]]] = None

@dataclass
class RequestOutput:
    """Output for a single request, containing one or more completions.
    
    Attributes:
        request_id: Unique identifier for this request.
        prompt: The original prompt text.
        prompt_token_ids: Token IDs of the prompt.
        outputs: List of generated completions.
        finished: Whether all outputs are complete.
    """
    request_id: str
    prompt: str
    prompt_token_ids: List[int]
    outputs: List[CompletionOutput]
    finished: bool = False


# Display the API documentation
print("API Documentation Generated from Type Hints and Docstrings")
print("=" * 60)

for cls in [SamplingParamsExample, CompletionOutput, RequestOutput]:
    print(f"\nclass {cls.__name__}:")
    print(cls.__doc__)
    print(f"  Fields:")
    for field_name, field_type in cls.__dataclass_fields__.items():
        default = field_type.default if field_type.default is not field_type.default_factory else "(factory)"
        if default is not dataclass:
            print(f"    {field_name}: {field_type.type} = {default}")
````

````python
# Demo: Tutorial quality evaluation framework

class TutorialEvaluator:
    """Evaluate tutorial quality against best practices."""
    
    def __init__(self):
        self.criteria = {
            "Structure": [
                ("Has clear title", lambda t: t.startswith('#')),
                ("Has prerequisites section", lambda t: 'prerequisit' in t.lower()),
                ("Has learning objectives", lambda t: 'learn' in t.lower() and ('will' in t.lower() or 'objective' in t.lower())),
                ("Has numbered steps", lambda t: '## Step' in t or '1.' in t),
                ("Has troubleshooting", lambda t: 'troubleshoot' in t.lower()),
                ("Has next steps", lambda t: 'next step' in t.lower()),
            ],
            "Content": [
                ("Has code examples", lambda t: '```' in t),
                ("Has expected output", lambda t: 'expected' in t.lower() or 'output' in t.lower()),
                ("Has command-line examples", lambda t: '```bash' in t or '$ ' in t),
                ("Code is copy-pasteable", lambda t: '```python' in t),
            ],
            "Accessibility": [
                ("Uses simple language", lambda t: True),  # Would need NLP for real check
                ("Explains jargon", lambda t: True),
                ("Progressive difficulty", lambda t: 'step 1' in t.lower()),
            ]
        }
    
    def evaluate(self, tutorial_text):
        """Evaluate a tutorial."""
        print("Tutorial Quality Evaluation")
        print("=" * 50)
        
        total = 0
        passed = 0
        
        for category, checks in self.criteria.items():
            print(f"\n{category}:")
            for description, check_func in checks:
                result = check_func(tutorial_text)
                total += 1
                if result:
                    passed += 1
                status = "PASS" if result else "MISS"
                print(f"  [{status}] {description}")
        
        score = 100 * passed / total if total > 0 else 0
        print(f"\nScore: {passed}/{total} ({score:.0f}%)")
        return score

evaluator = TutorialEvaluator()
evaluator.evaluate(tutorial_template)
````

````python
# Demo: Auto-generating documentation from code
import inspect

class DocGenerator:
    """Generate documentation from Python source code."""
    
    def generate_class_doc(self, cls):
        """Generate markdown documentation for a class."""
        lines = []
        lines.append(f"## `{cls.__name__}`")
        lines.append("")
        
        if cls.__doc__:
            lines.append(cls.__doc__.strip())
            lines.append("")
        
        # List methods
        methods = inspect.getmembers(cls, predicate=inspect.isfunction)
        public_methods = [(name, m) for name, m in methods if not name.startswith('_')]
        
        if public_methods:
            lines.append("### Methods")
            lines.append("")
            for name, method in public_methods:
                sig = inspect.signature(method)
                params = []
                for pname, param in sig.parameters.items():
                    if pname == 'self':
                        continue
                    if param.annotation != inspect.Parameter.empty:
                        params.append(f"{pname}: {param.annotation.__name__ if hasattr(param.annotation, '__name__') else str(param.annotation)}")
                    else:
                        params.append(pname)
                
                params_str = ", ".join(params)
                return_type = ""
                if sig.return_annotation != inspect.Signature.empty:
                    return_type = f" -> {sig.return_annotation}"
                
                lines.append(f"#### `{name}({params_str}){return_type}`")
                if method.__doc__:
                    lines.append(method.__doc__.strip())
                lines.append("")
        
        return "\n".join(lines)

# Test with a sample class
class SampleScheduler:
    """A sample scheduler for demonstration.
    
    This scheduler manages request queues and allocates resources.
    """
    
    def add_request(self, request_id: str, prompt: str) -> bool:
        """Add a new request to the scheduler.
        
        Args:
            request_id: Unique identifier for the request.
            prompt: The input text prompt.
        
        Returns:
            True if the request was added successfully.
        """
        return True
    
    def schedule(self) -> dict:
        """Run one scheduling step.
        
        Returns:
            Dictionary with scheduling results.
        """
        return {}
    
    def abort(self, request_id: str) -> None:
        """Abort a running request."""
        pass

gen = DocGenerator()
doc = gen.generate_class_doc(SampleScheduler)
print("Auto-Generated Documentation:")
print("=" * 60)
print(doc)
````

## 7. Creating Getting-Started Examples

````python
# Demo: Create a getting-started example for SGLang

sglang_getting_started = '''
"""Getting Started with SGLang.

This example demonstrates the core features of SGLang:
1. Starting the SGLang runtime
2. Using the SGLang language for structured generation
3. Parallel generation and branching
4. Constrained decoding

Requirements:
    pip install "sglang[all]"
    GPU with at least 16GB memory

Usage:
    python getting_started_sglang.py
"""

import sglang as sgl

# === Example 1: Basic text generation ===
@sgl.function
def basic_generation(s):
    s += "Tell me a joke about programming.\\n"
    s += sgl.gen("joke", max_tokens=100)

# === Example 2: Multi-turn conversation ===
@sgl.function
def multi_turn(s):
    s += sgl.system("You are a helpful assistant.")
    s += sgl.user("What is Python?")
    s += sgl.assistant(sgl.gen("response_1", max_tokens=100))
    s += sgl.user("Give me a code example.")
    s += sgl.assistant(sgl.gen("response_2", max_tokens=200))

# === Example 3: Parallel generation (fork-join) ===
@sgl.function
def parallel_generation(s):
    s += "Generate three different greetings:\\n"
    
    # Fork into 3 parallel branches
    forks = s.fork(3)
    for i, f in enumerate(forks):
        f += f"Greeting {i+1}: "
        f += sgl.gen(f"greeting_{i}", max_tokens=30)
    
    # Join results
    s += sgl.join(forks)

# === Example 4: Constrained decoding ===
@sgl.function
def constrained_output(s):
    s += "Rate this movie from 1-5: The Matrix\\n"
    s += "Rating: "
    s += sgl.gen("rating", choices=["1", "2", "3", "4", "5"])


def main():
    # Set the backend
    sgl.set_default_backend(
        sgl.RuntimeEndpoint("http://localhost:30000")
    )
    
    # Run examples
    print("=== Example 1: Basic Generation ===")
    result = basic_generation.run()
    print(result["joke"])
    
    print("\\n=== Example 2: Multi-Turn ===")
    result = multi_turn.run()
    print(f"Response 1: {result['response_1']}")
    print(f"Response 2: {result['response_2']}")
    
    print("\\n=== Example 3: Parallel Generation ===")
    result = parallel_generation.run()
    for i in range(3):
        print(f"Greeting {i+1}: {result[f'greeting_{i}']}")
    
    print("\\n=== Example 4: Constrained Output ===")
    result = constrained_output.run()
    print(f"Rating: {result['rating']}")


if __name__ == "__main__":
    main()
'''

print("SGLang Getting Started Example:")
print("=" * 60)
print(sglang_getting_started)
````
