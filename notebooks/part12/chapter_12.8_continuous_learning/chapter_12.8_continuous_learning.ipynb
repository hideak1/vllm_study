## Part 1: Hot Model Swapping — Dual-Buffer Strategy

### The Problem
Loading a large model (e.g., 70B parameters) takes minutes. During this time, the server can't serve requests if we simply replace the model.

### The Solution: Dual-Buffer
1. Keep the **active model** serving requests
2. Load the **new model** in background into a second buffer
3. Once loaded, atomically swap the active pointer
4. Drain remaining requests on the old model
5. Unload the old model to free memory

```
Time ──────────────────────────────────►

Buffer A: [Model v1 ACTIVE] ─────────── [drain] [unload]
Buffer B: [empty] [loading v2...] [v2 ACTIVE] ──────────
                                   ▲
                              atomic swap
```

````python
class ModelState(Enum):
    EMPTY = "empty"
    LOADING = "loading"
    READY = "ready"
    ACTIVE = "active"
    DRAINING = "draining"
    UNLOADING = "unloading"


@dataclass
class ModelVersion:
    """Represents a model version."""
    version_id: str
    model_name: str
    checkpoint_path: str
    created_at: float = field(default_factory=time.time)
    metadata: Dict = field(default_factory=dict)
    
    # Simulated properties
    load_time_seconds: float = 30.0  # Simulated load time
    memory_gb: float = 16.0
    
    def __repr__(self):
        return f"Model({self.version_id}, {self.model_name})"


@dataclass
class ModelBuffer:
    """A buffer slot for holding a model."""
    buffer_id: str
    state: ModelState = ModelState.EMPTY
    model: Optional[ModelVersion] = None
    active_requests: int = 0
    loaded_at: Optional[float] = None


class DualBufferModelManager:
    """Manages model lifecycle with dual-buffer hot swapping.
    
    This ensures zero downtime during model updates.
    """
    
    def __init__(self):
        self.buffers = {
            'A': ModelBuffer('A'),
            'B': ModelBuffer('B'),
        }
        self.active_buffer_id: Optional[str] = None
        self.swap_history: List[dict] = []
        self.event_log: List[dict] = []
    
    def _log_event(self, event: str, details: dict = None):
        self.event_log.append({
            'time': time.time(),
            'event': event,
            'details': details or {}
        })
    
    def _get_inactive_buffer(self) -> Optional[str]:
        for bid, buf in self.buffers.items():
            if bid != self.active_buffer_id:
                return bid
        return None
    
    def load_initial_model(self, model: ModelVersion) -> str:
        """Load the first model (blocking)."""
        buf_id = 'A'
        buf = self.buffers[buf_id]
        
        buf.state = ModelState.LOADING
        self._log_event('load_start', {'buffer': buf_id, 'model': model.version_id})
        
        # Simulate loading (instant in simulation)
        time.sleep(0.01)  # Tiny delay for simulation
        
        buf.model = model
        buf.state = ModelState.ACTIVE
        buf.loaded_at = time.time()
        self.active_buffer_id = buf_id
        
        self._log_event('load_complete', {'buffer': buf_id, 'model': model.version_id})
        return buf_id
    
    def hot_swap(self, new_model: ModelVersion) -> dict:
        """Perform a hot swap to a new model version.
        
        Returns swap details including timing.
        """
        swap_start = time.time()
        old_buffer_id = self.active_buffer_id
        new_buffer_id = self._get_inactive_buffer()
        
        if new_buffer_id is None:
            raise RuntimeError("No inactive buffer available")
        
        old_model = self.buffers[old_buffer_id].model
        new_buffer = self.buffers[new_buffer_id]
        
        # Step 1: Load new model into inactive buffer
        new_buffer.state = ModelState.LOADING
        self._log_event('swap_load_start', {
            'new_buffer': new_buffer_id, 'model': new_model.version_id
        })
        
        time.sleep(0.01)  # Simulate load time
        
        new_buffer.model = new_model
        new_buffer.state = ModelState.READY
        new_buffer.loaded_at = time.time()
        
        # Step 2: Atomic swap
        self.active_buffer_id = new_buffer_id
        new_buffer.state = ModelState.ACTIVE
        self.buffers[old_buffer_id].state = ModelState.DRAINING
        
        self._log_event('swap_active', {
            'new_buffer': new_buffer_id, 'old_buffer': old_buffer_id,
            'new_model': new_model.version_id,
            'old_model': old_model.version_id if old_model else None
        })
        
        # Step 3: Wait for old buffer to drain
        # In production, wait for active_requests to reach 0
        self.buffers[old_buffer_id].state = ModelState.UNLOADING
        
        # Step 4: Unload old model
        self.buffers[old_buffer_id].model = None
        self.buffers[old_buffer_id].state = ModelState.EMPTY
        self.buffers[old_buffer_id].active_requests = 0
        
        swap_end = time.time()
        
        result = {
            'old_model': old_model.version_id if old_model else None,
            'new_model': new_model.version_id,
            'swap_time_ms': (swap_end - swap_start) * 1000,
            'downtime_ms': 0,  # Zero downtime!
            'active_buffer': new_buffer_id
        }
        self.swap_history.append(result)
        return result
    
    def get_active_model(self) -> Optional[ModelVersion]:
        if self.active_buffer_id:
            return self.buffers[self.active_buffer_id].model
        return None
    
    def status(self) -> dict:
        return {
            'active_buffer': self.active_buffer_id,
            'buffers': {
                bid: {
                    'state': buf.state.value,
                    'model': buf.model.version_id if buf.model else None,
                    'active_requests': buf.active_requests
                }
                for bid, buf in self.buffers.items()
            },
            'total_swaps': len(self.swap_history)
        }


# Demo: hot model swapping
manager = DualBufferModelManager()

# Define model versions
v1 = ModelVersion("v1.0", "llama-3-8b", "/models/llama3-v1", memory_gb=16.0)
v2 = ModelVersion("v2.0", "llama-3-8b", "/models/llama3-v2", memory_gb=16.0)
v3 = ModelVersion("v3.0", "llama-3-8b-instruct", "/models/llama3-instruct", memory_gb=16.5)

# Initial load
manager.load_initial_model(v1)
print(f"Initial model loaded: {manager.get_active_model()}")
print(f"Status: {json.dumps(manager.status(), indent=2)}")

# Hot swap to v2
print(f"\n=== Hot Swap v1 -> v2 ===")
result = manager.hot_swap(v2)
print(f"Swap result: {json.dumps(result, indent=2)}")
print(f"Active model: {manager.get_active_model()}")

# Hot swap to v3
print(f"\n=== Hot Swap v2 -> v3 ===")
result = manager.hot_swap(v3)
print(f"Swap result: {json.dumps(result, indent=2)}")
print(f"Downtime: {result['downtime_ms']} ms (zero downtime!)")
````

## Part 2: A/B Testing Model Versions

A/B testing allows comparing model versions in production by splitting traffic.

````python
@dataclass
class ABTestConfig:
    """Configuration for an A/B test."""
    test_id: str
    model_a: str  # Control (current model)
    model_b: str  # Treatment (new model)
    traffic_split: float = 0.5  # Fraction of traffic to model B
    start_time: float = field(default_factory=time.time)
    min_samples: int = 1000
    metrics: List[str] = field(default_factory=lambda: ['latency', 'quality', 'errors'])
    sticky_sessions: bool = True  # Same user always gets same model


class ABTestFramework:
    """A/B testing framework for model versions."""
    
    def __init__(self):
        self.active_tests: Dict[str, ABTestConfig] = {}
        self.results: Dict[str, Dict[str, List]] = {}  # test_id -> model -> metrics
        self.user_assignments: Dict[str, Dict[str, str]] = {}  # test_id -> user -> model
    
    def create_test(self, config: ABTestConfig):
        self.active_tests[config.test_id] = config
        self.results[config.test_id] = {
            config.model_a: {'latency': [], 'quality': [], 'errors': 0, 'requests': 0},
            config.model_b: {'latency': [], 'quality': [], 'errors': 0, 'requests': 0},
        }
        self.user_assignments[config.test_id] = {}
    
    def assign_model(self, test_id: str, user_id: str) -> str:
        """Assign a user to a model variant."""
        config = self.active_tests[test_id]
        
        # Sticky sessions: same user always gets same model
        if config.sticky_sessions:
            if user_id in self.user_assignments.get(test_id, {}):
                return self.user_assignments[test_id][user_id]
        
        # Hash-based assignment for consistency
        hash_val = int(hashlib.md5(f"{test_id}:{user_id}".encode()).hexdigest(), 16)
        assigned = config.model_b if (hash_val % 100) < (config.traffic_split * 100) else config.model_a
        
        if config.sticky_sessions:
            self.user_assignments.setdefault(test_id, {})[user_id] = assigned
        
        return assigned
    
    def record_result(self, test_id: str, model: str, latency: float, 
                      quality: float, error: bool = False):
        """Record a request result."""
        results = self.results[test_id][model]
        results['latency'].append(latency)
        results['quality'].append(quality)
        results['requests'] += 1
        if error:
            results['errors'] += 1
    
    def analyze(self, test_id: str) -> dict:
        """Analyze A/B test results."""
        config = self.active_tests[test_id]
        results = self.results[test_id]
        
        analysis = {}
        for model_name, data in results.items():
            latencies = data['latency']
            qualities = data['quality']
            n = data['requests']
            
            if n == 0:
                continue
            
            analysis[model_name] = {
                'requests': n,
                'error_rate': data['errors'] / n,
                'avg_latency_ms': np.mean(latencies) if latencies else 0,
                'p50_latency_ms': np.percentile(latencies, 50) if latencies else 0,
                'p95_latency_ms': np.percentile(latencies, 95) if latencies else 0,
                'p99_latency_ms': np.percentile(latencies, 99) if latencies else 0,
                'avg_quality': np.mean(qualities) if qualities else 0,
                'quality_std': np.std(qualities) if qualities else 0,
            }
        
        # Statistical significance (simple z-test)
        if len(analysis) == 2:
            models = list(analysis.keys())
            q_a = results[models[0]]['quality']
            q_b = results[models[1]]['quality']
            
            if len(q_a) > 30 and len(q_b) > 30:
                mean_diff = np.mean(q_b) - np.mean(q_a)
                pooled_std = np.sqrt(np.var(q_a)/len(q_a) + np.var(q_b)/len(q_b))
                z_score = mean_diff / pooled_std if pooled_std > 0 else 0
                analysis['statistical'] = {
                    'quality_difference': mean_diff,
                    'z_score': z_score,
                    'significant_at_95': abs(z_score) > 1.96,
                    'winner': models[1] if z_score > 1.96 else (models[0] if z_score < -1.96 else 'no_winner')
                }
        
        return analysis


# Demo: run an A/B test
ab = ABTestFramework()
test_config = ABTestConfig(
    test_id="test-001",
    model_a="llama-3-8b-v1",
    model_b="llama-3-8b-v2",
    traffic_split=0.3,  # 30% to new model
    min_samples=100
)
ab.create_test(test_config)

# Simulate traffic
random.seed(42)
np.random.seed(42)

for i in range(2000):
    user_id = f"user-{i % 200}"  # 200 unique users
    model = ab.assign_model("test-001", user_id)
    
    # Simulate model performance (v2 is slightly better quality, similar latency)
    if model == "llama-3-8b-v1":
        latency = np.random.lognormal(3.0, 0.3)  # ~20ms avg
        quality = np.random.normal(0.75, 0.1)
        error = random.random() < 0.02  # 2% error rate
    else:
        latency = np.random.lognormal(3.0, 0.3)  # similar latency
        quality = np.random.normal(0.80, 0.1)   # 5% better quality
        error = random.random() < 0.015  # 1.5% error rate
    
    ab.record_result("test-001", model, latency, quality, error)

# Analyze
analysis = ab.analyze("test-001")
print("=== A/B Test Results ===")
for model, stats in analysis.items():
    if model == 'statistical':
        print(f"\n--- Statistical Analysis ---")
        print(f"  Quality difference: {stats['quality_difference']:.4f}")
        print(f"  Z-score: {stats['z_score']:.2f}")
        print(f"  Significant at 95%: {stats['significant_at_95']}")
        print(f"  Winner: {stats['winner']}")
    else:
        print(f"\n--- {model} ---")
        print(f"  Requests: {stats['requests']}")
        print(f"  Avg latency: {stats['avg_latency_ms']:.1f}ms")
        print(f"  P99 latency: {stats['p99_latency_ms']:.1f}ms")
        print(f"  Avg quality: {stats['avg_quality']:.4f}")
        print(f"  Error rate: {stats['error_rate']:.2%}")
````

## Part 3: Canary Deployments

A canary deployment gradually increases traffic to a new model while monitoring for problems. If issues are detected, traffic is automatically rolled back.

````python
class HealthCheck:
    """Health monitoring for canary deployments."""
    
    def __init__(
        self,
        max_error_rate: float = 0.05,
        max_latency_p99_ms: float = 200.0,
        min_quality: float = 0.6,
        window_size: int = 100
    ):
        self.max_error_rate = max_error_rate
        self.max_latency_p99_ms = max_latency_p99_ms
        self.min_quality = min_quality
        self.window_size = window_size
    
    def check(self, metrics: dict) -> Tuple[bool, List[str]]:
        """Check if metrics are healthy. Returns (healthy, violations)."""
        violations = []
        
        if metrics.get('error_rate', 0) > self.max_error_rate:
            violations.append(f"Error rate {metrics['error_rate']:.2%} > {self.max_error_rate:.2%}")
        
        if metrics.get('p99_latency', 0) > self.max_latency_p99_ms:
            violations.append(f"P99 latency {metrics['p99_latency']:.0f}ms > {self.max_latency_p99_ms:.0f}ms")
        
        if metrics.get('avg_quality', 1.0) < self.min_quality:
            violations.append(f"Quality {metrics['avg_quality']:.3f} < {self.min_quality:.3f}")
        
        return len(violations) == 0, violations


class CanaryDeployment:
    """Canary deployment with automatic rollback.
    
    Traffic ramp-up stages:
    1% -> 5% -> 10% -> 25% -> 50% -> 100%
    
    At each stage, monitor health for a period.
    If unhealthy, rollback to 0%.
    """
    
    STAGES = [0.01, 0.05, 0.10, 0.25, 0.50, 1.00]
    
    def __init__(
        self,
        canary_model: str,
        stable_model: str,
        health_check: HealthCheck,
        samples_per_stage: int = 200
    ):
        self.canary_model = canary_model
        self.stable_model = stable_model
        self.health_check = health_check
        self.samples_per_stage = samples_per_stage
        
        self.current_stage = 0
        self.traffic_fraction = 0.0
        self.state = 'not_started'  # not_started, rolling_out, completed, rolled_back
        self.stage_history: List[dict] = []
        self.metrics_buffer: Dict[str, List] = defaultdict(list)
    
    def route_request(self, user_id: str) -> str:
        """Route a request to canary or stable."""
        if self.state in ('not_started', 'rolled_back', 'completed'):
            return self.stable_model if self.state != 'completed' else self.canary_model
        
        # Hash-based routing
        hash_val = int(hashlib.md5(user_id.encode()).hexdigest(), 16) % 100
        return self.canary_model if hash_val < (self.traffic_fraction * 100) else self.stable_model
    
    def record_and_check(self, model: str, latency: float, quality: float, error: bool):
        """Record a result and check health if we have enough data."""
        if model != self.canary_model:
            return
        
        self.metrics_buffer['latency'].append(latency)
        self.metrics_buffer['quality'].append(quality)
        self.metrics_buffer['errors'].append(1 if error else 0)
    
    def evaluate_stage(self) -> dict:
        """Evaluate current stage health and decide next action."""
        latencies = self.metrics_buffer['latency']
        qualities = self.metrics_buffer['quality']
        errors = self.metrics_buffer['errors']
        
        if len(latencies) < 10:
            return {'action': 'wait', 'reason': 'insufficient data'}
        
        metrics = {
            'error_rate': sum(errors) / len(errors),
            'p99_latency': np.percentile(latencies, 99),
            'avg_quality': np.mean(qualities),
            'samples': len(latencies)
        }
        
        healthy, violations = self.health_check.check(metrics)
        
        stage_result = {
            'stage': self.current_stage,
            'traffic_fraction': self.traffic_fraction,
            'metrics': metrics,
            'healthy': healthy,
            'violations': violations
        }
        self.stage_history.append(stage_result)
        
        if not healthy:
            self.state = 'rolled_back'
            self.traffic_fraction = 0.0
            return {'action': 'rollback', 'reason': violations, 'stage': stage_result}
        
        # Advance to next stage
        self.current_stage += 1
        self.metrics_buffer = defaultdict(list)  # Reset
        
        if self.current_stage >= len(self.STAGES):
            self.state = 'completed'
            self.traffic_fraction = 1.0
            return {'action': 'complete', 'stage': stage_result}
        
        self.traffic_fraction = self.STAGES[self.current_stage]
        return {'action': 'advance', 'new_fraction': self.traffic_fraction, 'stage': stage_result}
    
    def start(self):
        self.state = 'rolling_out'
        self.current_stage = 0
        self.traffic_fraction = self.STAGES[0]


# Demo: successful canary deployment
print("=== Successful Canary Deployment ===")
health = HealthCheck(max_error_rate=0.05, max_latency_p99_ms=200, min_quality=0.6)
canary = CanaryDeployment("llama-v2", "llama-v1", health, samples_per_stage=50)
canary.start()

random.seed(42)
np.random.seed(42)

for stage_idx in range(len(CanaryDeployment.STAGES)):
    # Simulate requests for this stage
    for i in range(100):
        user = f"user-{random.randint(0, 999)}"
        model = canary.route_request(user)
        
        # Good model: low errors, good quality
        latency = np.random.lognormal(2.5, 0.3)
        quality = np.random.normal(0.8, 0.08)
        error = random.random() < 0.02
        canary.record_and_check(model, latency, quality, error)
    
    result = canary.evaluate_stage()
    print(f"Stage {stage_idx}: {canary.STAGES[stage_idx]*100:.0f}% traffic -> {result['action'].upper()}")
    
    if result['action'] in ('complete', 'rollback'):
        break

print(f"\nFinal state: {canary.state}")
````

````python
# Demo: canary that triggers rollback
print("=== Canary with Rollback ===")
health = HealthCheck(max_error_rate=0.05, max_latency_p99_ms=100, min_quality=0.7)
bad_canary = CanaryDeployment("llama-v3-bad", "llama-v1", health, samples_per_stage=50)
bad_canary.start()

random.seed(123)
np.random.seed(123)

for stage_idx in range(len(CanaryDeployment.STAGES)):
    for i in range(100):
        user = f"user-{random.randint(0, 999)}"
        model = bad_canary.route_request(user)
        
        if model == "llama-v3-bad":
            # Bad model: high errors at higher traffic
            error_rate = 0.02 + stage_idx * 0.03  # Increases with load
            latency = np.random.lognormal(3.0, 0.5)  # Higher latency
            quality = np.random.normal(0.65, 0.15)    # Lower quality
            error = random.random() < error_rate
        else:
            latency = np.random.lognormal(2.5, 0.3)
            quality = np.random.normal(0.8, 0.08)
            error = random.random() < 0.02
        
        bad_canary.record_and_check(model, latency, quality, error)
    
    result = bad_canary.evaluate_stage()
    action = result['action'].upper()
    
    if result['action'] == 'rollback':
        print(f"Stage {stage_idx}: {bad_canary.STAGES[min(stage_idx, len(bad_canary.STAGES)-1)]*100:.0f}% traffic -> ROLLBACK!")
        print(f"  Violations: {result['reason']}")
        break
    else:
        print(f"Stage {stage_idx}: {bad_canary.STAGES[stage_idx]*100:.0f}% traffic -> {action}")

print(f"\nFinal state: {bad_canary.state}")
print("The bad model was automatically rolled back before affecting all users!")
````

## Part 4: Version Management and Instant Rollback

````python
class ModelVersionManager:
    """Manages model versions with instant rollback capability.
    
    Features:
    - Version registry with metadata
    - Deployment history tracking
    - One-click rollback to any previous version
    - Version comparison
    """
    
    def __init__(self):
        self.versions: Dict[str, ModelVersion] = {}
        self.deployment_history: List[dict] = []
        self.current_version: Optional[str] = None
        self.model_manager = DualBufferModelManager()
    
    def register_version(
        self, version_id: str, model_name: str, checkpoint_path: str,
        metadata: Dict = None
    ) -> ModelVersion:
        version = ModelVersion(
            version_id=version_id,
            model_name=model_name,
            checkpoint_path=checkpoint_path,
            metadata=metadata or {}
        )
        self.versions[version_id] = version
        return version
    
    def deploy(self, version_id: str, reason: str = "") -> dict:
        """Deploy a specific version."""
        if version_id not in self.versions:
            raise ValueError(f"Version {version_id} not registered")
        
        version = self.versions[version_id]
        old_version = self.current_version
        
        if self.current_version is None:
            self.model_manager.load_initial_model(version)
        else:
            self.model_manager.hot_swap(version)
        
        self.current_version = version_id
        
        record = {
            'timestamp': time.time(),
            'action': 'deploy',
            'from_version': old_version,
            'to_version': version_id,
            'reason': reason
        }
        self.deployment_history.append(record)
        return record
    
    def rollback(self, to_version: Optional[str] = None) -> dict:
        """Rollback to a previous version.
        
        If to_version is None, rolls back to the immediately previous version.
        """
        if to_version is None:
            # Find the previous version from history
            for record in reversed(self.deployment_history):
                if record['from_version'] and record['from_version'] != self.current_version:
                    to_version = record['from_version']
                    break
        
        if to_version is None:
            raise ValueError("No previous version found for rollback")
        
        return self.deploy(to_version, reason=f"rollback from {self.current_version}")
    
    def history(self) -> List[dict]:
        return self.deployment_history
    
    def list_versions(self) -> List[dict]:
        return [
            {
                'version_id': v.version_id,
                'model_name': v.model_name,
                'is_active': v.version_id == self.current_version,
                'metadata': v.metadata
            }
            for v in self.versions.values()
        ]


# Demo: version management
vm = ModelVersionManager()

# Register versions
vm.register_version("v1.0", "llama-3-8b", "/models/v1", {"fine_tuned": False, "eval_score": 0.75})
vm.register_version("v1.1", "llama-3-8b", "/models/v1.1", {"fine_tuned": True, "eval_score": 0.78})
vm.register_version("v2.0", "llama-3-8b-instruct", "/models/v2", {"fine_tuned": True, "eval_score": 0.82})
vm.register_version("v2.1", "llama-3-8b-instruct", "/models/v2.1", {"fine_tuned": True, "eval_score": 0.80})

# Deploy sequence
print("=== Deployment Sequence ===")
vm.deploy("v1.0", "initial deployment")
print(f"1. Deployed v1.0")

vm.deploy("v1.1", "fine-tuning improvement")
print(f"2. Deployed v1.1 (fine-tuned)")

vm.deploy("v2.0", "major version upgrade")
print(f"3. Deployed v2.0 (instruct model)")

vm.deploy("v2.1", "hotfix")
print(f"4. Deployed v2.1 (hotfix)")

# Oops, v2.1 has issues!
print(f"\n=== Emergency Rollback ===")
result = vm.rollback()
print(f"Rolled back to: {vm.current_version}")
print(f"Reason: {result['reason']}")

# Show history
print(f"\n=== Deployment History ===")
for i, record in enumerate(vm.history()):
    print(f"  {i+1}. {record['from_version'] or 'none':>5s} -> {record['to_version']:<5s} ({record['reason']})")

print(f"\n=== Available Versions ===")
for v in vm.list_versions():
    active = ' [ACTIVE]' if v['is_active'] else ''
    print(f"  {v['version_id']}: {v['model_name']} (eval={v['metadata'].get('eval_score', 'N/A')}){active}")
````

## Part 5: Online Fine-Tuning Integration

Continuous learning involves feeding production data back into training.

````python
class ContinuousLearningPipeline:
    """Pipeline for continuous model improvement.
    
    Flow:
    1. Collect production data (user feedback, quality metrics)
    2. Filter and prepare training data
    3. Fine-tune model (offline)
    4. Evaluate new version
    5. Deploy via canary
    6. Monitor and decide: keep or rollback
    """
    
    def __init__(self, version_manager: ModelVersionManager):
        self.vm = version_manager
        self.data_buffer: List[dict] = []
        self.training_runs: List[dict] = []
        self.min_data_for_training = 1000
    
    def collect_feedback(self, prompt: str, response: str, 
                        rating: float, model_version: str):
        """Collect user feedback from production."""
        self.data_buffer.append({
            'prompt': prompt,
            'response': response,
            'rating': rating,
            'model_version': model_version,
            'timestamp': time.time()
        })
    
    def prepare_training_data(self) -> dict:
        """Filter and prepare data for training."""
        if len(self.data_buffer) < self.min_data_for_training:
            return {'ready': False, 'samples': len(self.data_buffer), 
                    'needed': self.min_data_for_training}
        
        # Filter: keep high-rated examples for SFT
        good_examples = [d for d in self.data_buffer if d['rating'] >= 4.0]
        # Keep pairs for DPO/RLHF
        # (simplified: just count)
        
        return {
            'ready': True,
            'total_samples': len(self.data_buffer),
            'high_quality_samples': len(good_examples),
            'avg_rating': np.mean([d['rating'] for d in self.data_buffer]),
            'models_seen': list(set(d['model_version'] for d in self.data_buffer))
        }
    
    def simulate_training(self, base_version: str) -> str:
        """Simulate a fine-tuning run."""
        base = self.vm.versions[base_version]
        
        # Generate new version ID
        parts = base_version.split('.')
        new_minor = int(parts[-1]) + 1
        new_version_id = f"{'.'.join(parts[:-1])}.{new_minor}"
        
        # Simulate improvement
        old_score = base.metadata.get('eval_score', 0.75)
        improvement = np.random.normal(0.02, 0.01)  # Small improvement
        new_score = min(old_score + improvement, 1.0)
        
        self.vm.register_version(
            new_version_id,
            base.model_name,
            f"/models/{new_version_id}",
            {
                'fine_tuned': True,
                'eval_score': round(new_score, 4),
                'base_version': base_version,
                'training_samples': len(self.data_buffer)
            }
        )
        
        training_record = {
            'base_version': base_version,
            'new_version': new_version_id,
            'eval_score_before': old_score,
            'eval_score_after': new_score,
            'training_samples': len(self.data_buffer),
            'timestamp': time.time()
        }
        self.training_runs.append(training_record)
        
        # Clear data buffer
        self.data_buffer = []
        
        return new_version_id


# Demo: continuous learning pipeline
pipeline = ContinuousLearningPipeline(vm)

# Simulate production feedback collection
print("=== Continuous Learning Pipeline ===")
random.seed(42)

for i in range(1500):
    pipeline.collect_feedback(
        prompt=f"Question {i}",
        response=f"Answer {i}",
        rating=random.uniform(1.0, 5.0),
        model_version=vm.current_version
    )

# Check data readiness
data_status = pipeline.prepare_training_data()
print(f"\nData collection status:")
print(json.dumps(data_status, indent=2))

# Train new version
if data_status['ready']:
    new_version = pipeline.simulate_training(vm.current_version)
    print(f"\nTrained new version: {new_version}")
    print(f"Training record: {json.dumps(pipeline.training_runs[-1], indent=2)}")
    
    # Deploy via hot swap
    vm.deploy(new_version, "continuous learning update")
    print(f"\nDeployed {new_version}. Active model: {vm.current_version}")
````

````python
class ProductionMLSystem:
    """Complete production ML system with all components."""
    
    def __init__(self):
        self.version_manager = ModelVersionManager()
        self.ab_framework = ABTestFramework()
        self.pipeline = ContinuousLearningPipeline(self.version_manager)
        self.request_count = 0
        self.active_canary: Optional[CanaryDeployment] = None
    
    def handle_request(self, user_id: str, prompt: str) -> dict:
        """Handle a production request through the full pipeline."""
        self.request_count += 1
        
        # Route through canary if active
        if self.active_canary and self.active_canary.state == 'rolling_out':
            model = self.active_canary.route_request(user_id)
        # Route through A/B test if active
        elif self.ab_framework.active_tests:
            test_id = list(self.ab_framework.active_tests.keys())[0]
            model = self.ab_framework.assign_model(test_id, user_id)
        else:
            model = self.version_manager.current_version
        
        # Simulate inference
        latency = np.random.lognormal(2.5, 0.3)
        quality = np.random.normal(0.8, 0.1)
        response = f"[Response from {model}]"
        
        return {
            'model': model,
            'response': response,
            'latency_ms': latency,
            'quality': quality
        }
    
    def status(self) -> dict:
        return {
            'current_model': self.version_manager.current_version,
            'total_requests': self.request_count,
            'active_ab_tests': len(self.ab_framework.active_tests),
            'canary_active': self.active_canary is not None and self.active_canary.state == 'rolling_out',
            'versions_registered': len(self.version_manager.versions),
            'deployment_count': len(self.version_manager.deployment_history)
        }


# Quick demo
system = ProductionMLSystem()
system.version_manager.register_version("v1.0", "llama-3", "/v1", {"eval_score": 0.8})
system.version_manager.deploy("v1.0", "initial")

# Process some requests
for i in range(100):
    system.handle_request(f"user-{i % 20}", f"query {i}")

print("=== Production System Status ===")
print(json.dumps(system.status(), indent=2))
````

## Summary

| Strategy | When to Use | Downtime | Risk |
|----------|------------|----------|------|
| Hot Swap (Dual Buffer) | Routine updates | Zero | Low (instant rollback) |
| A/B Testing | Comparing versions | Zero | Low (controlled traffic) |
| Canary Deployment | New version validation | Zero | Very Low (auto-rollback) |
| Blue/Green | Major changes | Near-zero | Low |
| Instant Rollback | Emergency | Zero | None |

### Key Takeaways
1. **Dual-buffer swapping** eliminates downtime by loading the new model before switching
2. **A/B testing** requires statistical rigor — use z-tests and sufficient sample sizes
3. **Canary deployments** gradually increase risk exposure and auto-rollback on failure
4. **Version management** with instant rollback is essential for production safety
5. **Continuous learning** closes the feedback loop: production data improves the model
