# Chapter 12.2: KV-Cache Compression

## Overview

KV-cache is the dominant memory consumer during LLM inference. For a 70B model serving a 4K-token sequence, the KV-cache alone consumes ~1.3 GB per request. Compression techniques can dramatically reduce this, enabling higher batch sizes and longer context lengths.

### Compression Approaches
1. **Quantization**: Reduce precision of KV-cache entries (FP16 -> FP8/INT8/INT4)
2. **Token Eviction**: Remove less important tokens (H2O, StreamingLLM)
3. **Sparse Attention**: Compute attention over a subset of tokens

### Key Papers
- **H2O** (Heavy Hitter Oracle): Token eviction based on attention score accumulation
- **StreamingLLM**: Attention sink + sliding window for infinite context
- **KIVI**: KV-cache quantization with per-channel/per-token schemes

### Learning Objectives
1. Implement KV-cache quantization (FP8, INT8, INT4)
2. Build H2O-style token eviction
3. Understand attention sink phenomenon
4. Measure quality-compression trade-offs

## 1. KV-Cache Memory Analysis

Let's start by understanding how much memory KV-cache consumes and why compression matters.

```python
def kv_cache_memory(num_layers: int, num_heads: int, head_dim: int,
                     seq_len: int, batch_size: int, dtype_bytes: int = 2) -> float:
    """Calculate KV-cache memory in GB."""
    # K and V each: [batch, num_heads, seq_len, head_dim]
    return 2 * num_layers * num_heads * seq_len * head_dim * batch_size * dtype_bytes / (1024**3)

# Common model configs
models = {
    'LLaMA-7B':  {'layers': 32, 'heads': 32, 'head_dim': 128},
    'LLaMA-13B': {'layers': 40, 'heads': 40, 'head_dim': 128},
    'LLaMA-70B': {'layers': 80, 'heads': 64, 'head_dim': 128},
    'GPT-4-est': {'layers': 120, 'heads': 96, 'head_dim': 128},
}

seq_lengths = [1024, 2048, 4096, 8192, 16384, 32768, 131072]
dtype_names = {'FP16': 2, 'FP8': 1, 'INT8': 1, 'INT4': 0.5}

print(f"KV-Cache Memory (GB) per request - FP16")
print(f"{'Model':>15}", end='')
for sl in seq_lengths:
    print(f"{sl:>10}", end='')
print()
print("-" * (15 + 10 * len(seq_lengths)))

for name, cfg in models.items():
    print(f"{name:>15}", end='')
    for sl in seq_lengths:
        mem = kv_cache_memory(cfg['layers'], cfg['heads'], cfg['head_dim'], sl, 1)
        print(f"{mem:>9.2f}G", end='')
    print()

print(f"\n\nCompression savings for LLaMA-70B, seq_len=4096, batch_size=32:")
for dtype_name, dtype_bytes in dtype_names.items():
    mem = kv_cache_memory(80, 64, 128, 4096, 32, dtype_bytes)
    print(f"  {dtype_name:>5}: {mem:>8.2f} GB")
```

```python
# Visualize memory savings
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Memory vs sequence length for different dtypes
seq_range = np.array([512, 1024, 2048, 4096, 8192, 16384, 32768, 65536, 131072])
colors = {'FP16': 'red', 'FP8': 'orange', 'INT8': 'blue', 'INT4': 'green'}

for dtype_name, dtype_bytes in dtype_names.items():
    mems = [kv_cache_memory(80, 64, 128, sl, 1, dtype_bytes) for sl in seq_range]
    ax1.semilogy(seq_range, mems, 'o-', color=colors[dtype_name],
                 linewidth=2, label=dtype_name)

ax1.axhline(y=80, color='gray', linestyle='--', alpha=0.5, label='A100/H100 80GB')
ax1.set_xlabel('Sequence Length')
ax1.set_ylabel('KV-Cache Memory (GB)')
ax1.set_title('LLaMA-70B KV-Cache Memory per Request')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Max batch size given 40GB KV budget
kv_budget_gb = 40  # Reserve 40GB for KV-cache
for dtype_name, dtype_bytes in dtype_names.items():
    max_batch = [kv_budget_gb / kv_cache_memory(80, 64, 128, sl, 1, dtype_bytes)
                 for sl in seq_range]
    ax2.semilogx(seq_range, max_batch, 'o-', color=colors[dtype_name],
                  linewidth=2, label=dtype_name)

ax2.set_xlabel('Sequence Length')
ax2.set_ylabel('Max Batch Size')
ax2.set_title(f'Max Batch Size ({kv_budget_gb}GB KV Budget)')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('kv_cache_memory.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 2. Demo: KV-Cache Quantization

Implement quantization schemes for KV-cache: symmetric, asymmetric, per-channel, and per-token.

```python
class KVCacheQuantizer:
    """Quantize KV-cache values to lower precision."""

    @staticmethod
    def symmetric_quantize(tensor: np.ndarray, bits: int) -> Tuple[np.ndarray, float]:
        """Symmetric quantization: scale based on max absolute value."""
        max_val = np.max(np.abs(tensor))
        qmax = (1 << (bits - 1)) - 1
        scale = max_val / qmax if max_val > 0 else 1.0
        quantized = np.clip(np.round(tensor / scale), -qmax, qmax).astype(np.int8 if bits <= 8 else np.int16)
        return quantized, scale

    @staticmethod
    def symmetric_dequantize(quantized: np.ndarray, scale: float) -> np.ndarray:
        return quantized.astype(np.float32) * scale

    @staticmethod
    def asymmetric_quantize(tensor: np.ndarray, bits: int) -> Tuple[np.ndarray, float, float]:
        """Asymmetric quantization: uses both zero-point and scale."""
        min_val = np.min(tensor)
        max_val = np.max(tensor)
        qmax = (1 << bits) - 1
        scale = (max_val - min_val) / qmax if max_val != min_val else 1.0
        zero_point = -np.round(min_val / scale)
        quantized = np.clip(np.round(tensor / scale + zero_point), 0, qmax).astype(np.uint8)
        return quantized, scale, zero_point

    @staticmethod
    def asymmetric_dequantize(quantized: np.ndarray, scale: float,
                               zero_point: float) -> np.ndarray:
        return (quantized.astype(np.float32) - zero_point) * scale

    @staticmethod
    def per_channel_quantize(tensor: np.ndarray, bits: int,
                              axis: int = -1) -> Tuple[np.ndarray, np.ndarray]:
        """Per-channel quantization: separate scale per channel."""
        qmax = (1 << (bits - 1)) - 1
        # Move quantization axis to last
        moved = np.moveaxis(tensor, axis, -1)
        orig_shape = moved.shape
        flat = moved.reshape(-1, orig_shape[-1])

        max_vals = np.max(np.abs(flat), axis=0)
        scales = max_vals / qmax
        scales = np.where(scales == 0, 1.0, scales)

        quantized = np.clip(np.round(flat / scales), -qmax, qmax).astype(np.int8)
        quantized = quantized.reshape(orig_shape)
        quantized = np.moveaxis(quantized, -1, axis)
        return quantized, scales

    @staticmethod
    def per_channel_dequantize(quantized: np.ndarray, scales: np.ndarray,
                                axis: int = -1) -> np.ndarray:
        moved = np.moveaxis(quantized, axis, -1)
        orig_shape = moved.shape
        flat = moved.reshape(-1, orig_shape[-1]).astype(np.float32)
        dequant = flat * scales
        dequant = dequant.reshape(orig_shape)
        return np.moveaxis(dequant, -1, axis)


# Test quantization quality
quantizer = KVCacheQuantizer()

# Simulate KV-cache values (typically normally distributed)
batch, heads, seq_len, head_dim = 1, 8, 256, 128
kv_cache = np.random.randn(batch, heads, seq_len, head_dim).astype(np.float32)
# Add some outliers (common in real KV-caches)
outlier_mask = np.random.random(kv_cache.shape) < 0.01
kv_cache[outlier_mask] *= 10

print(f"Original KV-cache: shape={kv_cache.shape}, "
      f"range=[{kv_cache.min():.3f}, {kv_cache.max():.3f}]")

# Test each quantization method
methods = {}
for bits in [8, 4]:
    # Symmetric
    q, s = quantizer.symmetric_quantize(kv_cache, bits)
    dq = quantizer.symmetric_dequantize(q, s)
    err = np.mean((kv_cache - dq) ** 2)
    methods[f'Symmetric INT{bits}'] = {'mse': err, 'bits': bits}

    # Per-channel
    q, s = quantizer.per_channel_quantize(kv_cache, bits, axis=-1)
    dq = quantizer.per_channel_dequantize(q, s, axis=-1)
    err = np.mean((kv_cache - dq) ** 2)
    methods[f'Per-Channel INT{bits}'] = {'mse': err, 'bits': bits}

print(f"\n{'Method':>25} {'MSE':>12} {'PSNR (dB)':>12} {'Compression':>12}")
print("-" * 65)
for name, info in methods.items():
    psnr = 10 * np.log10(np.max(kv_cache**2) / info['mse']) if info['mse'] > 0 else float('inf')
    compression = 16 / info['bits']
    print(f"{name:>25} {info['mse']:>12.6f} {psnr:>10.1f}dB {compression:>10.1f}x")
```

```python
# Visualize quantization error distributions
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Original distribution
axes[0, 0].hist(kv_cache.flatten(), bins=100, density=True, alpha=0.7, color='blue')
axes[0, 0].set_title('Original KV-Cache Distribution')
axes[0, 0].set_xlabel('Value')

# INT8 symmetric error
q8, s8 = quantizer.symmetric_quantize(kv_cache, 8)
dq8 = quantizer.symmetric_dequantize(q8, s8)
err8 = (kv_cache - dq8).flatten()
axes[0, 1].hist(err8, bins=100, density=True, alpha=0.7, color='orange')
axes[0, 1].set_title(f'INT8 Symmetric Error (MSE={np.mean(err8**2):.6f})')
axes[0, 1].set_xlabel('Error')

# INT4 symmetric error
q4, s4 = quantizer.symmetric_quantize(kv_cache, 4)
dq4 = quantizer.symmetric_dequantize(q4, s4)
err4 = (kv_cache - dq4).flatten()
axes[1, 0].hist(err4, bins=100, density=True, alpha=0.7, color='red')
axes[1, 0].set_title(f'INT4 Symmetric Error (MSE={np.mean(err4**2):.6f})')
axes[1, 0].set_xlabel('Error')

# INT4 per-channel error
q4c, s4c = quantizer.per_channel_quantize(kv_cache, 4, axis=-1)
dq4c = quantizer.per_channel_dequantize(q4c, s4c, axis=-1)
err4c = (kv_cache - dq4c).flatten()
axes[1, 1].hist(err4c, bins=100, density=True, alpha=0.7, color='green')
axes[1, 1].set_title(f'INT4 Per-Channel Error (MSE={np.mean(err4c**2):.6f})')
axes[1, 1].set_xlabel('Error')

plt.tight_layout()
plt.savefig('quantization_errors.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 3. Demo: H2O (Heavy Hitter Oracle) Token Eviction

H2O identifies "heavy hitter" tokens that consistently receive high attention scores. The eviction policy keeps:
1. **Attention sinks**: Initial tokens that serve as attention anchors
2. **Heavy hitters**: Tokens with accumulated high attention scores
3. **Recent tokens**: Tokens within a sliding window

```python
class H2OCache:
    """Heavy Hitter Oracle (H2O) KV-cache eviction."""

    def __init__(self, max_cache_size: int, num_heads: int, head_dim: int,
                 num_sink_tokens: int = 4, recent_window: int = 64):
        self.max_cache_size = max_cache_size
        self.num_heads = num_heads
        self.head_dim = head_dim
        self.num_sink_tokens = num_sink_tokens
        self.recent_window = recent_window
        self.heavy_hitter_budget = max_cache_size - num_sink_tokens - recent_window

        # Storage
        self.key_cache = None  # [num_heads, cache_size, head_dim]
        self.value_cache = None
        self.attention_scores = None  # Accumulated attention scores
        self.token_positions = []  # Original position of each cached token
        self.current_size = 0
        self.eviction_history = []

    def _softmax(self, x, axis=-1):
        e = np.exp(x - np.max(x, axis=axis, keepdims=True))
        return e / np.sum(e, axis=axis, keepdims=True)

    def append(self, key: np.ndarray, value: np.ndarray, position: int):
        """Add a new token's KV to cache."""
        if self.key_cache is None:
            self.key_cache = key.reshape(self.num_heads, 1, self.head_dim)
            self.value_cache = value.reshape(self.num_heads, 1, self.head_dim)
            self.attention_scores = np.zeros((self.num_heads, 1))
            self.token_positions = [position]
            self.current_size = 1
            return

        # Add new KV
        self.key_cache = np.concatenate([
            self.key_cache,
            key.reshape(self.num_heads, 1, self.head_dim)
        ], axis=1)
        self.value_cache = np.concatenate([
            self.value_cache,
            value.reshape(self.num_heads, 1, self.head_dim)
        ], axis=1)
        self.attention_scores = np.concatenate([
            self.attention_scores,
            np.zeros((self.num_heads, 1))
        ], axis=1)
        self.token_positions.append(position)
        self.current_size += 1

        # Evict if over capacity
        if self.current_size > self.max_cache_size:
            self._evict()

    def compute_attention(self, query: np.ndarray) -> np.ndarray:
        """Compute attention and update accumulated scores."""
        # query: [num_heads, head_dim]
        # scores: [num_heads, cache_size]
        scale = 1.0 / np.sqrt(self.head_dim)
        scores = np.einsum('hd,hsd->hs', query, self.key_cache) * scale
        attn_weights = self._softmax(scores, axis=-1)

        # Update accumulated attention scores
        self.attention_scores += attn_weights

        # Compute output
        output = np.einsum('hs,hsd->hd', attn_weights, self.value_cache)
        return output, attn_weights

    def _evict(self):
        """Evict tokens using H2O policy."""
        # Never evict sink tokens or recent tokens
        evictable_start = self.num_sink_tokens
        evictable_end = self.current_size - self.recent_window

        if evictable_end <= evictable_start:
            return

        # Find token with lowest accumulated attention in evictable range
        evictable_scores = self.attention_scores[:, evictable_start:evictable_end]
        avg_scores = np.mean(evictable_scores, axis=0)  # Average across heads
        evict_idx = evictable_start + np.argmin(avg_scores)

        self.eviction_history.append({
            'position': self.token_positions[evict_idx],
            'score': float(np.mean(self.attention_scores[:, evict_idx]))
        })

        # Remove evicted token
        self.key_cache = np.delete(self.key_cache, evict_idx, axis=1)
        self.value_cache = np.delete(self.value_cache, evict_idx, axis=1)
        self.attention_scores = np.delete(self.attention_scores, evict_idx, axis=1)
        self.token_positions.pop(evict_idx)
        self.current_size -= 1


# Simulate H2O cache behavior
num_heads, head_dim = 8, 64
full_seq_len = 1024
cache_budget = 256  # Only keep 256 tokens in cache

h2o_cache = H2OCache(max_cache_size=cache_budget, num_heads=num_heads,
                      head_dim=head_dim, num_sink_tokens=4, recent_window=64)

# Generate synthetic sequence
keys = np.random.randn(full_seq_len, num_heads, head_dim).astype(np.float32) * 0.1
values = np.random.randn(full_seq_len, num_heads, head_dim).astype(np.float32) * 0.1
queries = np.random.randn(full_seq_len, num_heads, head_dim).astype(np.float32) * 0.1

# Make some tokens "heavy hitters" with stronger keys
heavy_hitter_positions = [10, 50, 100, 200, 300, 500]
for pos in heavy_hitter_positions:
    keys[pos] *= 5.0  # These tokens will attract more attention

attention_history = []
for t in range(full_seq_len):
    h2o_cache.append(keys[t], values[t], t)
    if t > 0:
        _, attn_weights = h2o_cache.compute_attention(queries[t])
        attention_history.append(attn_weights.copy())

print(f"H2O Cache Summary:")
print(f"  Full sequence length: {full_seq_len}")
print(f"  Cache budget: {cache_budget}")
print(f"  Compression ratio: {full_seq_len/cache_budget:.1f}x")
print(f"  Tokens evicted: {len(h2o_cache.eviction_history)}")
print(f"  Retained positions: {h2o_cache.token_positions[:10]}... (first 10)")
print(f"  Heavy hitters retained: ", end='')
retained = set(h2o_cache.token_positions)
print([p for p in heavy_hitter_positions if p in retained])
```

```python
# Visualize H2O eviction and attention patterns
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Retained token positions
retained_positions = sorted(h2o_cache.token_positions)
ax = axes[0, 0]
ax.scatter(retained_positions, [1]*len(retained_positions), s=2, alpha=0.5)
for hp in heavy_hitter_positions:
    if hp in retained:
        ax.axvline(hp, color='red', alpha=0.3, linewidth=1)
ax.set_xlabel('Token Position')
ax.set_title(f'Retained Tokens (cache={cache_budget}/{full_seq_len})')
ax.set_yticks([])

# Accumulated attention scores
ax = axes[0, 1]
avg_scores = np.mean(h2o_cache.attention_scores, axis=0)
ax.bar(range(len(avg_scores)), avg_scores, width=1.0, alpha=0.7)
ax.set_xlabel('Cache Slot')
ax.set_ylabel('Accumulated Attention')
ax.set_title('Accumulated Attention Scores in Cache')

# Eviction timeline
ax = axes[1, 0]
evict_positions = [e['position'] for e in h2o_cache.eviction_history]
evict_scores = [e['score'] for e in h2o_cache.eviction_history]
ax.scatter(range(len(evict_positions)), evict_positions, s=1, alpha=0.3, c=evict_scores, cmap='coolwarm')
ax.set_xlabel('Eviction Step')
ax.set_ylabel('Evicted Token Position')
ax.set_title('Eviction Timeline')

# Attention pattern for last query
ax = axes[1, 1]
if attention_history:
    last_attn = attention_history[-1]
    ax.bar(range(last_attn.shape[1]), np.mean(last_attn, axis=0), width=1.0, alpha=0.7)
    ax.set_xlabel('Cache Position')
    ax.set_ylabel('Attention Weight')
    ax.set_title('Final Attention Pattern (avg across heads)')

plt.tight_layout()
plt.savefig('h2o_cache.png', dpi=150, bbox_inches='tight')
plt.show()
```

## 4. Demo: Attention Sink Phenomenon & StreamingLLM

The "attention sink" phenomenon: initial tokens receive disproportionately high attention regardless of content. StreamingLLM exploits this by always keeping the first few tokens.

```python
class StreamingLLMCache:
    """StreamingLLM: attention sinks + sliding window."""

    def __init__(self, num_sink_tokens: int, window_size: int,
                 num_heads: int, head_dim: int):
        self.num_sink = num_sink_tokens
        self.window_size = window_size
        self.num_heads = num_heads
        self.head_dim = head_dim
        self.total_budget = num_sink_tokens + window_size

        self.sink_keys = None   # [num_heads, num_sink, head_dim]
        self.sink_values = None
        self.window_keys = None  # [num_heads, window_size, head_dim]
        self.window_values = None
        self.tokens_seen = 0

    def append(self, key: np.ndarray, value: np.ndarray):
        k = key.reshape(self.num_heads, 1, self.head_dim)
        v = value.reshape(self.num_heads, 1, self.head_dim)

        if self.tokens_seen < self.num_sink:
            if self.sink_keys is None:
                self.sink_keys = k
                self.sink_values = v
            else:
                self.sink_keys = np.concatenate([self.sink_keys, k], axis=1)
                self.sink_values = np.concatenate([self.sink_values, v], axis=1)
        else:
            if self.window_keys is None:
                self.window_keys = k
                self.window_values = v
            else:
                self.window_keys = np.concatenate([self.window_keys, k], axis=1)
                self.window_values = np.concatenate([self.window_values, v], axis=1)
                # Evict oldest from window if over budget
                if self.window_keys.shape[1] > self.window_size:
                    self.window_keys = self.window_keys[:, -self.window_size:, :]
                    self.window_values = self.window_values[:, -self.window_size:, :]

        self.tokens_seen += 1

    def get_kv(self):
        """Get combined sink + window KV."""
        if self.window_keys is None:
            return self.sink_keys, self.sink_values
        keys = np.concatenate([self.sink_keys, self.window_keys], axis=1)
        values = np.concatenate([self.sink_values, self.window_values], axis=1)
        return keys, values


# Demonstrate attention sink phenomenon
def simulate_attention_pattern(seq_len: int, num_heads: int, head_dim: int,
                                sink_strength: float = 3.0) -> np.ndarray:
    """Simulate realistic attention patterns with attention sinks."""
    keys = np.random.randn(seq_len, num_heads, head_dim).astype(np.float32) * 0.5
    queries = np.random.randn(seq_len, num_heads, head_dim).astype(np.float32) * 0.5

    # Attention sink: first few tokens have strong keys
    keys[:4] *= sink_strength

    # Compute attention for last token
    q = queries[-1]  # [num_heads, head_dim]
    scale = 1.0 / np.sqrt(head_dim)
    scores = np.einsum('hd,shd->hs', q, keys) * scale
    # Apply causal mask (all positions visible to last token)
    attn = np.exp(scores - np.max(scores, axis=-1, keepdims=True))
    attn = attn / np.sum(attn, axis=-1, keepdims=True)
    return attn

# Visualize attention sink
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for ax, sink_str, title in zip(axes,
    [1.0, 3.0, 5.0],
    ['No Sink (strength=1)', 'Moderate Sink (3x)', 'Strong Sink (5x)']):

    attn = simulate_attention_pattern(512, 8, 64, sink_strength=sink_str)
    avg_attn = np.mean(attn, axis=0)
    ax.bar(range(len(avg_attn)), avg_attn, width=1.0, alpha=0.7)
    ax.set_xlabel('Token Position')
    ax.set_ylabel('Attention Weight')
    ax.set_title(title)
    ax.set_xlim(-5, 520)
    # Highlight first 4 tokens
    for i in range(4):
        ax.axvline(i, color='red', alpha=0.3, linewidth=1)

plt.suptitle('Attention Sink Phenomenon', fontsize=14, y=1.02)
plt.tight_layout()
plt.savefig('attention_sink.png', dpi=150, bbox_inches='tight')
plt.show()
print("Red lines mark the 'attention sink' tokens (positions 0-3).")
```

```python
# Compare full attention vs StreamingLLM vs sliding window
def compare_cache_strategies(seq_len: int = 1024, num_heads: int = 8,
                              head_dim: int = 64) -> dict:
    """Compare output quality of different cache strategies."""
    # Generate sequence with attention sinks
    keys = np.random.randn(seq_len, num_heads, head_dim).astype(np.float32) * 0.5
    values = np.random.randn(seq_len, num_heads, head_dim).astype(np.float32) * 0.5
    queries = np.random.randn(seq_len, num_heads, head_dim).astype(np.float32) * 0.5
    keys[:4] *= 3.0  # Attention sinks

    results = {}

    # Full attention (ground truth)
    scale = 1.0 / np.sqrt(head_dim)
    q = queries[-1]
    scores = np.einsum('hd,shd->hs', q, keys) * scale
    attn = np.exp(scores - np.max(scores, axis=-1, keepdims=True))
    attn = attn / np.sum(attn, axis=-1, keepdims=True)
    full_output = np.einsum('hs,shd->hd', attn, values)
    results['Full Attention'] = full_output

    # Sliding window only (no sinks)
    for window in [64, 128, 256, 512]:
        window_keys = keys[-window:]
        window_values = values[-window:]
        scores = np.einsum('hd,shd->hs', q, window_keys) * scale
        attn = np.exp(scores - np.max(scores, axis=-1, keepdims=True))
        attn = attn / np.sum(attn, axis=-1, keepdims=True)
        output = np.einsum('hs,shd->hd', attn, window_values)
        results[f'Window-{window}'] = output

    # StreamingLLM (sinks + window)
    for window in [60, 124, 252, 508]:
        sink_keys = keys[:4]
        sink_values = values[:4]
        win_keys = keys[-window:]
        win_values = values[-window:]
        combined_keys = np.concatenate([sink_keys, win_keys], axis=0)
        combined_values = np.concatenate([sink_values, win_values], axis=0)
        scores = np.einsum('hd,shd->hs', q, combined_keys) * scale
        attn = np.exp(scores - np.max(scores, axis=-1, keepdims=True))
        attn = attn / np.sum(attn, axis=-1, keepdims=True)
        output = np.einsum('hs,shd->hd', attn, combined_values)
        total = 4 + window
        results[f'StreamLLM-{total}'] = output

    # Compute MSE relative to full attention
    metrics = {}
    for name, output in results.items():
        if name == 'Full Attention':
            continue
        mse = np.mean((full_output - output) ** 2)
        cos_sim = np.mean([
            np.dot(full_output[h], output[h]) /
            (np.linalg.norm(full_output[h]) * np.linalg.norm(output[h]) + 1e-8)
            for h in range(num_heads)
        ])
        metrics[name] = {'mse': mse, 'cosine_sim': cos_sim}

    return metrics

metrics = compare_cache_strategies()
print(f"{'Strategy':>20} {'MSE':>12} {'Cosine Sim':>12}")
print("-" * 48)
for name, m in sorted(metrics.items()):
    print(f"{name:>20} {m['mse']:>12.6f} {m['cosine_sim']:>12.4f}")
```

## Assignment 1: Build a KV-Cache Compression Pipeline

Combine quantization and token eviction into a unified compression pipeline. The pipeline should:
1. Apply token eviction first (reduce token count)
2. Then quantize remaining tokens
3. Support configurable compression ratios

```python
# ASSIGNMENT 1: Build a KV-Cache Compression Pipeline

class KVCacheCompressionPipeline:
    """Unified KV-cache compression with eviction + quantization."""

    def __init__(self, target_compression_ratio: float = 4.0,
                 eviction_strategy: str = 'h2o',
                 quant_bits: int = 8,
                 num_sink_tokens: int = 4,
                 recent_window_fraction: float = 0.25):
        """
        target_compression_ratio: Overall target (e.g., 4x means 1/4 of original size)
        eviction_strategy: 'h2o', 'random', 'fifo'
        quant_bits: Quantization bit width for remaining tokens
        """
        self.target_ratio = target_compression_ratio
        self.eviction_strategy = eviction_strategy
        self.quant_bits = quant_bits
        self.num_sink_tokens = num_sink_tokens
        self.recent_window_fraction = recent_window_fraction

    def compress(self, keys: np.ndarray, values: np.ndarray,
                 attention_scores: Optional[np.ndarray] = None) -> dict:
        """
        TODO: Implement the compression pipeline.

        Args:
            keys: [num_heads, seq_len, head_dim]
            values: [num_heads, seq_len, head_dim]
            attention_scores: [num_heads, seq_len] accumulated attention

        Returns:
            dict with compressed keys, values, metadata for decompression
        """
        num_heads, seq_len, head_dim = keys.shape
        quant_compression = 16 / self.quant_bits
        eviction_ratio = self.target_ratio / quant_compression
        keep_tokens = max(self.num_sink_tokens + 1, int(seq_len / eviction_ratio))

        # YOUR CODE HERE:
        # Step 1: Token eviction
        # - Keep sink tokens (first num_sink_tokens)
        # - Keep recent window
        # - For middle tokens, use eviction_strategy to select which to keep

        # Step 2: Quantization
        # - Quantize the remaining tokens to quant_bits precision

        # Step 3: Return compressed result with decompression metadata

        # Placeholder implementation
        recent_window = int(keep_tokens * self.recent_window_fraction)
        sink_end = self.num_sink_tokens
        middle_budget = keep_tokens - sink_end - recent_window

        if self.eviction_strategy == 'h2o' and attention_scores is not None:
            middle_scores = np.mean(attention_scores[:, sink_end:-recent_window], axis=0)
            top_indices = np.argsort(middle_scores)[-middle_budget:]
            top_indices = np.sort(top_indices) + sink_end
        elif self.eviction_strategy == 'random':
            middle_range = np.arange(sink_end, seq_len - recent_window)
            top_indices = np.sort(np.random.choice(middle_range, middle_budget, replace=False))
        else:  # fifo
            top_indices = np.arange(sink_end, sink_end + middle_budget)

        keep_indices = np.concatenate([
            np.arange(sink_end),
            top_indices,
            np.arange(seq_len - recent_window, seq_len)
        ]).astype(int)

        evicted_keys = keys[:, keep_indices, :]
        evicted_values = values[:, keep_indices, :]

        # Quantize
        quantizer = KVCacheQuantizer()
        q_keys, s_keys = quantizer.per_channel_quantize(evicted_keys, self.quant_bits)
        q_values, s_values = quantizer.per_channel_quantize(evicted_values, self.quant_bits)

        original_size = keys.nbytes + values.nbytes
        compressed_size = q_keys.nbytes + q_values.nbytes + s_keys.nbytes + s_values.nbytes

        return {
            'keys': q_keys, 'values': q_values,
            'key_scales': s_keys, 'value_scales': s_values,
            'keep_indices': keep_indices,
            'original_seq_len': seq_len,
            'compression_ratio': original_size / compressed_size,
            'tokens_kept': len(keep_indices),
        }

    def decompress(self, compressed: dict) -> Tuple[np.ndarray, np.ndarray]:
        """TODO: Decompress KV-cache for inference."""
        quantizer = KVCacheQuantizer()
        keys = quantizer.per_channel_dequantize(compressed['keys'],
                                                 compressed['key_scales'])
        values = quantizer.per_channel_dequantize(compressed['values'],
                                                   compressed['value_scales'])
        return keys, values


# Test the pipeline
pipeline = KVCacheCompressionPipeline(
    target_compression_ratio=4.0,
    eviction_strategy='h2o',
    quant_bits=4,
    num_sink_tokens=4,
)

# Create test KV-cache
test_keys = np.random.randn(8, 1024, 64).astype(np.float32)
test_values = np.random.randn(8, 1024, 64).astype(np.float32)
test_attn = np.random.exponential(0.01, (8, 1024)).astype(np.float32)
test_attn[:, :4] = 0.5  # Sinks have high accumulated attention

compressed = pipeline.compress(test_keys, test_values, test_attn)
decompressed_keys, decompressed_values = pipeline.decompress(compressed)

print(f"Compression Pipeline Results:")
print(f"  Original: {test_keys.shape[1]} tokens, {test_keys.nbytes + test_values.nbytes} bytes")
print(f"  Compressed: {compressed['tokens_kept']} tokens")
print(f"  Compression ratio: {compressed['compression_ratio']:.2f}x")

# Quality check on retained tokens
retained = compressed['keep_indices']
orig_retained_keys = test_keys[:, retained, :]
mse = np.mean((orig_retained_keys - decompressed_keys) ** 2)
print(f"  Quantization MSE (retained tokens): {mse:.6f}")
```

## Assignment 2: Measure Quality Impact of Compression

Systematically evaluate how different compression levels affect output quality.

```python
# ASSIGNMENT 2: Quality Impact Measurement

def measure_attention_output_quality(keys: np.ndarray, values: np.ndarray,
                                      query: np.ndarray,
                                      compressed_keys: np.ndarray,
                                      compressed_values: np.ndarray) -> dict:
    """
    TODO: Compare attention output with full vs compressed KV-cache.

    Metrics to compute:
    - MSE of attention output
    - Cosine similarity of attention output
    - KL divergence of attention distributions
    - Top-k attention position overlap
    """
    head_dim = keys.shape[-1]
    scale = 1.0 / np.sqrt(head_dim)

    # Full attention
    full_scores = np.einsum('hd,hsd->hs', query, keys) * scale
    full_attn = np.exp(full_scores - np.max(full_scores, axis=-1, keepdims=True))
    full_attn = full_attn / np.sum(full_attn, axis=-1, keepdims=True)
    full_output = np.einsum('hs,hsd->hd', full_attn, values)

    # Compressed attention
    comp_scores = np.einsum('hd,hsd->hs', query, compressed_keys) * scale
    comp_attn = np.exp(comp_scores - np.max(comp_scores, axis=-1, keepdims=True))
    comp_attn = comp_attn / np.sum(comp_attn, axis=-1, keepdims=True)
    comp_output = np.einsum('hs,hsd->hd', comp_attn, compressed_values)

    # YOUR CODE HERE: compute metrics
    mse = float(np.mean((full_output - comp_output) ** 2))

    cos_sims = []
    for h in range(keys.shape[0]):
        norm_f = np.linalg.norm(full_output[h])
        norm_c = np.linalg.norm(comp_output[h])
        cos_sims.append(np.dot(full_output[h], comp_output[h]) / (norm_f * norm_c + 1e-10))
    cos_sim = float(np.mean(cos_sims))

    return {'mse': mse, 'cosine_similarity': cos_sim}


# Sweep compression ratios
compression_ratios = [1.5, 2, 3, 4, 6, 8, 12, 16]
strategies = ['h2o', 'random', 'fifo']
quant_configs = [(8, 'INT8'), (4, 'INT4')]

results = {}
query = np.random.randn(8, 64).astype(np.float32)

for strategy in strategies:
    for bits, bits_name in quant_configs:
        key_name = f"{strategy}+{bits_name}"
        results[key_name] = []
        for ratio in compression_ratios:
            pipeline = KVCacheCompressionPipeline(
                target_compression_ratio=ratio,
                eviction_strategy=strategy,
                quant_bits=bits,
            )
            compressed = pipeline.compress(test_keys, test_values, test_attn)
            dk, dv = pipeline.decompress(compressed)
            quality = measure_attention_output_quality(
                test_keys, test_values, query, dk, dv
            )
            quality['compression_ratio'] = compressed['compression_ratio']
            results[key_name].append(quality)

# Plot quality vs compression
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

for key_name, data in results.items():
    ratios = [d['compression_ratio'] for d in data]
    mses = [d['mse'] for d in data]
    cos_sims = [d['cosine_similarity'] for d in data]
    ax1.semilogy(ratios, mses, 'o-', label=key_name)
    ax2.plot(ratios, cos_sims, 'o-', label=key_name)

ax1.set_xlabel('Compression Ratio')
ax1.set_ylabel('MSE')
ax1.set_title('Output Quality vs Compression')
ax1.legend(fontsize=8)
ax1.grid(True, alpha=0.3)

ax2.set_xlabel('Compression Ratio')
ax2.set_ylabel('Cosine Similarity')
ax2.set_title('Output Similarity vs Compression')
ax2.legend(fontsize=8)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('compression_quality.png', dpi=150, bbox_inches='tight')
plt.show()
```

## Assignment 3: Design a Hybrid Compression Strategy

Design a strategy that adapts compression per-layer and per-head based on importance.

```python
# ASSIGNMENT 3: Hybrid Compression Strategy

class HybridCompressionStrategy:
    """
    TODO: Implement an adaptive compression strategy that:
    1. Uses different quantization bits per layer (higher bits for important layers)
    2. Uses different eviction ratios per head (keep more for important heads)
    3. Adapts based on attention patterns
    """

    def __init__(self, num_layers: int, num_heads: int,
                 target_overall_compression: float = 4.0):
        self.num_layers = num_layers
        self.num_heads = num_heads
        self.target_compression = target_overall_compression

        # Layer importance scores (to be computed)
        self.layer_importance = np.ones(num_layers) / num_layers
        # Head importance scores
        self.head_importance = np.ones((num_layers, num_heads)) / num_heads

    def analyze_importance(self, attention_patterns: np.ndarray):
        """
        TODO: Analyze attention patterns to determine layer/head importance.

        Args:
            attention_patterns: [num_layers, num_heads, seq_len] attention scores
        """
        # YOUR CODE HERE
        # Hint: Important layers/heads have high entropy or high variance
        for layer in range(self.num_layers):
            layer_attn = attention_patterns[layer]  # [num_heads, seq_len]
            # Entropy-based importance
            entropy = -np.sum(layer_attn * np.log(layer_attn + 1e-10), axis=-1)
            self.layer_importance[layer] = np.mean(entropy)
            self.head_importance[layer] = entropy

        # Normalize
        self.layer_importance /= np.sum(self.layer_importance)
        for l in range(self.num_layers):
            self.head_importance[l] /= np.sum(self.head_importance[l])

    def get_layer_config(self, layer_idx: int) -> dict:
        """
        TODO: Return compression config for a specific layer.

        Returns dict with:
        - 'quant_bits': int (4 or 8)
        - 'eviction_ratio': float (fraction of tokens to keep)
        """
        # YOUR CODE HERE
        importance = self.layer_importance[layer_idx]
        # More important layers get more bits and less eviction
        if importance > np.median(self.layer_importance):
            return {'quant_bits': 8, 'eviction_ratio': 0.7}
        else:
            return {'quant_bits': 4, 'eviction_ratio': 0.4}

    def compress_all_layers(self, kv_caches: dict,
                            attention_patterns: np.ndarray) -> dict:
        """
        TODO: Compress KV-caches across all layers with adaptive strategy.
        """
        self.analyze_importance(attention_patterns)

        compressed_layers = {}
        total_original = 0
        total_compressed = 0

        for layer in range(self.num_layers):
            config = self.get_layer_config(layer)
            keys = kv_caches[layer]['keys']
            values = kv_caches[layer]['values']
            attn = attention_patterns[layer]

            pipeline = KVCacheCompressionPipeline(
                target_compression_ratio=1.0 / config['eviction_ratio'] * (16 / config['quant_bits']),
                eviction_strategy='h2o',
                quant_bits=config['quant_bits'],
            )
            compressed = pipeline.compress(keys, values, attn)
            compressed_layers[layer] = compressed
            compressed_layers[layer]['config'] = config

            total_original += keys.nbytes + values.nbytes
            total_compressed += compressed['keys'].nbytes + compressed['values'].nbytes

        return {
            'layers': compressed_layers,
            'overall_compression': total_original / total_compressed,
            'total_original_mb': total_original / 1024**2,
            'total_compressed_mb': total_compressed / 1024**2,
        }


# Test hybrid strategy
num_layers, num_heads, seq_len, head_dim = 8, 8, 1024, 64

# Generate test data
kv_caches = {}
for l in range(num_layers):
    kv_caches[l] = {
        'keys': np.random.randn(num_heads, seq_len, head_dim).astype(np.float32),
        'values': np.random.randn(num_heads, seq_len, head_dim).astype(np.float32),
    }

attention_patterns = generate_multilayer_attention(seq_len, num_layers, num_heads, head_dim)

hybrid = HybridCompressionStrategy(num_layers, num_heads, target_overall_compression=4.0)
result = hybrid.compress_all_layers(kv_caches, attention_patterns)

print(f"Hybrid Compression Results:")
print(f"  Overall compression: {result['overall_compression']:.2f}x")
print(f"  Original size: {result['total_original_mb']:.1f} MB")
print(f"  Compressed size: {result['total_compressed_mb']:.1f} MB")
print(f"\n  Per-layer config:")
for l in range(num_layers):
    cfg = result['layers'][l]['config']
    tokens = result['layers'][l]['tokens_kept']
    print(f"    Layer {l}: {cfg['quant_bits']}-bit, {tokens}/{seq_len} tokens kept, "
          f"importance={hybrid.layer_importance[l]:.3f}")
```

1. **KV-cache quantization** from FP16 to INT4 gives 4x compression with modest quality loss; per-channel quantization significantly improves accuracy

2. **H2O token eviction** identifies "heavy hitters" that consistently receive high attention, enabling significant token reduction while preserving output quality

3. **Attention sinks** are real -- the first few tokens disproportionately attract attention regardless of content; StreamingLLM exploits this for infinite-length generation

4. **Hybrid strategies** that adapt compression per-layer and per-head based on importance analysis achieve better quality-compression trade-offs than uniform compression
