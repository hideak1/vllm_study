```python
# Demo 2: Programmatic Decision Engine

class DecisionEngine:
    """Automated decision engine for choosing between vLLM and SGLang."""
    
    def __init__(self):
        # Decision factors and their weights for each system
        self.factors = {
            'need_frontend_dsl': {
                'question': 'Do you need a frontend DSL for complex LLM programs?',
                'vLLM_weight': -5, 'SGLang_weight': 10,
            },
            'prefix_caching_critical': {
                'question': 'Is prefix caching for multi-turn conversations critical?',
                'vLLM_weight': 3, 'SGLang_weight': 8,
            },
            'constrained_decoding': {
                'question': 'Do you need fast constrained/JSON decoding?',
                'vLLM_weight': 5, 'SGLang_weight': 9,
            },
            'broad_quant_support': {
                'question': 'Do you need GGUF, SqueezeLLM, or niche quantization formats?',
                'vLLM_weight': 9, 'SGLang_weight': 3,
            },
            'beam_search': {
                'question': 'Is beam search important for your use case?',
                'vLLM_weight': 9, 'SGLang_weight': 4,
            },
            'production_stability': {
                'question': 'Is production stability and maturity a top priority?',
                'vLLM_weight': 8, 'SGLang_weight': 6,
            },
            'ray_integration': {
                'question': 'Do you need Ray-based distributed serving?',
                'vLLM_weight': 9, 'SGLang_weight': 3,
            },
            'openai_api_compat': {
                'question': 'Is OpenAI API compatibility essential?',
                'vLLM_weight': 8, 'SGLang_weight': 7,
            },
            'batch_processing': {
                'question': 'Is offline batch processing the primary use case?',
                'vLLM_weight': 7, 'SGLang_weight': 7,
            },
            'multi_lora': {
                'question': 'Do you need to serve many LoRA adapters?',
                'vLLM_weight': 8, 'SGLang_weight': 7,
            },
            'disaggregated_serving': {
                'question': 'Do you want disaggregated prefill/decode?',
                'vLLM_weight': 4, 'SGLang_weight': 8,
            },
            'community_support': {
                'question': 'Is a large community and ecosystem important?',
                'vLLM_weight': 9, 'SGLang_weight': 6,
            },
        }
    
    def evaluate(self, answers: Dict[str, bool]) -> Dict:
        """Evaluate based on yes/no answers to each factor."""
        vllm_score = 0
        sglang_score = 0
        breakdown = []
        
        for factor, config in self.factors.items():
            if factor in answers and answers[factor]:
                vllm_score += config['vLLM_weight']
                sglang_score += config['SGLang_weight']
                breakdown.append({
                    'factor': factor,
                    'question': config['question'],
                    'vLLM_contribution': config['vLLM_weight'],
                    'SGLang_contribution': config['SGLang_weight'],
                    'winner': 'vLLM' if config['vLLM_weight'] > config['SGLang_weight'] else 'SGLang',
                })
        
        recommendation = 'vLLM' if vllm_score > sglang_score else 'SGLang'
        confidence = abs(vllm_score - sglang_score) / max(vllm_score + sglang_score, 1) * 100
        
        return {
            'recommendation': recommendation,
            'vLLM_score': vllm_score,
            'SGLang_score': sglang_score,
            'confidence': confidence,
            'breakdown': breakdown,
        }
    
    def print_questions(self):
        for i, (factor, config) in enumerate(self.factors.items(), 1):
            print(f"  {i}. [{factor}] {config['question']}")

engine = DecisionEngine()
print("Decision factors:")
engine.print_questions()
```

```python
# Demo 11: Migration effort estimator

def estimate_migration_effort(source: str, features_used: List[str]) -> Dict:
    """Estimate the effort to migrate from one system to the other."""
    
    effort_map = {
        'vLLM_to_SGLang': {
            'OpenAI API calls': ('low', 'Change port/URL, minor param adjustments'),
            'Batch inference': ('low', 'Similar API, rename some parameters'),
            'LoRA serving': ('medium', 'Different config format, test thoroughly'),
            'Ray distributed': ('high', 'Need to replace with SGLang TP/DP'),
            'Custom sampling': ('medium', 'API differences in logits processors'),
            'Beam search': ('high', 'SGLang has limited beam search support'),
            'GGUF models': ('blocker', 'SGLang may not support GGUF format'),
        },
        'SGLang_to_vLLM': {
            'OpenAI API calls': ('low', 'Change port/URL'),
            'Frontend DSL': ('high', 'No equivalent; must decompose into API calls'),
            'Fork/Join patterns': ('high', 'Must implement manual orchestration'),
            'RadixCache tuning': ('medium', 'Switch to APC configuration'),
            'Constrained choices': ('medium', 'Use guided_choice parameter'),
            'Data parallelism': ('medium', 'Use Ray for equivalent functionality'),
            'Jump-forward decoding': ('medium', 'Use Outlines integration'),
        },
    }
    
    direction = f"{source}_to_{'SGLang' if source == 'vLLM' else 'vLLM'}"
    effort_data = effort_map.get(direction, {})
    
    total_effort = 0
    effort_values = {'low': 1, 'medium': 3, 'high': 5, 'blocker': 10}
    
    print(f"\nMigration Effort Estimate: {direction.replace('_', ' ')}")
    print(f"{'Feature':<25} {'Effort':<10} {'Details'}")
    print('=' * 80)
    
    for feature in features_used:
        if feature in effort_data:
            level, details = effort_data[feature]
            total_effort += effort_values[level]
            emoji = {'low': '[EASY]', 'medium': '[MODERATE]', 'high': '[HARD]', 'blocker': '[BLOCKER]'}[level]
            print(f"{feature:<25} {emoji:<10} {details}")
        else:
            print(f"{feature:<25} {'[N/A]':<10} Not in migration map")
    
    max_possible = len(features_used) * 10
    difficulty = total_effort / max_possible * 100 if max_possible > 0 else 0
    
    print(f"\nOverall difficulty: {difficulty:.0f}% ({total_effort}/{max_possible} points)")
    if difficulty < 30:
        print("Assessment: Straightforward migration (1-2 days)")
    elif difficulty < 60:
        print("Assessment: Moderate effort (1-2 weeks)")
    else:
        print("Assessment: Significant effort (2-4 weeks, may require architecture changes)")

# Example: Migrating a vLLM chatbot to SGLang
estimate_migration_effort('vLLM', ['OpenAI API calls', 'LoRA serving', 'Custom sampling'])

# Example: Migrating an SGLang DSL app to vLLM
estimate_migration_effort('SGLang', ['Frontend DSL', 'Fork/Join patterns', 'Constrained choices'])
```

## Hands-on Assignment 3: Build a System Selection Questionnaire

Create an interactive questionnaire that non-technical stakeholders can use.

```python
# Assignment 3: System selection questionnaire

class SystemSelectionQuestionnaire:
    """An interactive questionnaire for selecting between vLLM and SGLang.
    
    Designed for non-technical decision makers.
    """
    
    def __init__(self):
        self.questions = [
            {
                'id': 'workload',
                'question': 'What is your primary workload type?',
                'options': {
                    'a': ('Interactive chatbot', {'vLLM': 7, 'SGLang': 9}),
                    'b': ('Batch data processing', {'vLLM': 8, 'SGLang': 7}),
                    'c': ('Code generation', {'vLLM': 8, 'SGLang': 8}),
                    'd': ('RAG application', {'vLLM': 7, 'SGLang': 9}),
                    'e': ('Multi-agent system', {'vLLM': 5, 'SGLang': 9}),
                },
            },
            {
                'id': 'team_size',
                'question': 'What is your team size?',
                'options': {
                    'a': ('1-3 developers', {'vLLM': 8, 'SGLang': 7}),
                    'b': ('4-10 developers', {'vLLM': 8, 'SGLang': 8}),
                    'c': ('10+ developers (enterprise)', {'vLLM': 9, 'SGLang': 6}),
                },
            },
            {
                'id': 'priority',
                'question': 'What is your top priority?',
                'options': {
                    'a': ('Maximum performance', {'vLLM': 8, 'SGLang': 9}),
                    'b': ('Ease of use', {'vLLM': 9, 'SGLang': 7}),
                    'c': ('Feature flexibility', {'vLLM': 7, 'SGLang': 9}),
                    'd': ('Production stability', {'vLLM': 9, 'SGLang': 7}),
                },
            },
            {
                'id': 'hardware',
                'question': 'What hardware do you have?',
                'options': {
                    'a': ('Single GPU (consumer)', {'vLLM': 8, 'SGLang': 7}),
                    'b': ('Single GPU (datacenter)', {'vLLM': 9, 'SGLang': 9}),
                    'c': ('Multi-GPU (2-8)', {'vLLM': 9, 'SGLang': 8}),
                    'd': ('Multi-node cluster', {'vLLM': 9, 'SGLang': 7}),
                },
            },
            {
                'id': 'structured_output',
                'question': 'How important is structured (JSON) output?',
                'options': {
                    'a': ('Not needed', {'vLLM': 5, 'SGLang': 5}),
                    'b': ('Nice to have', {'vLLM': 7, 'SGLang': 7}),
                    'c': ('Critical requirement', {'vLLM': 7, 'SGLang': 10}),
                },
            },
        ]
    
    def run_questionnaire(self, answers: Dict[str, str]) -> Dict:
        """Run the questionnaire with provided answers.
        
        answers: dict of {question_id: option_letter}
        """
        vllm_total = 0
        sglang_total = 0
        details = []
        
        for q in self.questions:
            answer = answers.get(q['id'], 'a')
            if answer in q['options']:
                desc, scores = q['options'][answer]
                vllm_total += scores['vLLM']
                sglang_total += scores['SGLang']
                details.append({
                    'question': q['question'],
                    'answer': desc,
                    'vLLM': scores['vLLM'],
                    'SGLang': scores['SGLang'],
                })
        
        winner = 'vLLM' if vllm_total > sglang_total else 'SGLang'
        margin = abs(vllm_total - sglang_total)
        
        return {
            'recommendation': winner,
            'vLLM_score': vllm_total,
            'SGLang_score': sglang_total,
            'margin': margin,
            'strength': 'Strong' if margin > 5 else 'Moderate' if margin > 2 else 'Close call',
            'details': details,
        }
    
    def display_results(self, result: Dict):
        print(f"\n{'='*60}")
        print(f"RECOMMENDATION: {result['recommendation']}")
        print(f"{'='*60}")
        print(f"\nScores: vLLM={result['vLLM_score']}, SGLang={result['SGLang_score']}")
        print(f"Strength: {result['strength']} (margin: {result['margin']})")
        print(f"\nBreakdown:")
        print(f"{'Question':<40} {'Answer':<25} {'vLLM':<8} {'SGLang':<8}")
        print('-' * 80)
        for d in result['details']:
            print(f"{d['question']:<40} {d['answer']:<25} {d['vLLM']:<8} {d['SGLang']:<8}")

# Run the questionnaire
questionnaire = SystemSelectionQuestionnaire()

# Example: Chatbot developer answers
chatbot_answers = {
    'workload': 'a',        # Interactive chatbot
    'team_size': 'a',       # 1-3 developers
    'priority': 'a',        # Maximum performance
    'hardware': 'b',        # Single datacenter GPU
    'structured_output': 'c',  # Critical
}

result = questionnaire.run_questionnaire(chatbot_answers)
questionnaire.display_results(result)

# TODO: Run with your own answers
print("\nTODO: Create your own answers and run the questionnaire.")
print("Share the result with your team for discussion.")
```
