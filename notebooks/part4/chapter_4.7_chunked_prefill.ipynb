## 1. The Problem: Long Prefills Block Everything

### Standard Scheduling (No Chunking)

```
Without chunked prefill:

Time →
│ Prefill (4096 tok) │ Decode │ Decode │ Decode │ Decode │
├────────────────────┤────────┤────────┤────────┤────────┤
0ms               200ms    210ms   220ms   230ms   240ms

Problem: During the 200ms prefill, ALL decode requests are blocked!
  - Existing conversations stall (no new tokens)
  - ITL spikes for decode requests
  - P99 latency is terrible

With chunked prefill (chunk_size=512):

Time →
│Chunk0+Dec│Chunk1+Dec│Chunk2+Dec│...│Chunk7+Dec│Decode│
├──────────┤──────────┤──────────┤   ├──────────┤──────┤
0ms      30ms      60ms      90ms    240ms     260ms  270ms

Now: Decode requests get serviced every 30ms instead of waiting 200ms!
  - Consistent ITL
  - Slightly higher TTFT (prefill spread across iterations)
  - Much better P99 latency
```

````python
import torch
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from typing import List, Dict, Optional, Tuple
from dataclasses import dataclass, field
import time
import random

random.seed(42)
np.random.seed(42)

print("Imports loaded.")
````

````python
@dataclass
class Request:
    """Represents a serving request."""
    id: int
    prompt_len: int
    max_output_len: int
    arrival_time: float
    
    # State tracking
    prefill_tokens_remaining: int = 0
    tokens_generated: int = 0
    first_token_time: Optional[float] = None
    completion_time: Optional[float] = None
    token_times: List[float] = field(default_factory=list)
    
    def __post_init__(self):
        self.prefill_tokens_remaining = self.prompt_len
    
    @property
    def is_prefilling(self) -> bool:
        return self.prefill_tokens_remaining > 0
    
    @property
    def is_complete(self) -> bool:
        return self.tokens_generated >= self.max_output_len
    
    @property
    def ttft(self) -> Optional[float]:
        if self.first_token_time is not None:
            return self.first_token_time - self.arrival_time
        return None
    
    @property
    def itl_stats(self) -> dict:
        if len(self.token_times) < 2:
            return {'mean': 0, 'p50': 0, 'p99': 0}
        itls = [self.token_times[i+1] - self.token_times[i] 
                for i in range(len(self.token_times)-1)]
        return {
            'mean': np.mean(itls),
            'p50': np.percentile(itls, 50),
            'p99': np.percentile(itls, 99),
        }

print("Request dataclass defined.")
````

````python
class SchedulerSimulator:
    """
    Simulates vLLM's scheduler with and without chunked prefill.
    
    Source reference: vllm/core/scheduler.py
    """
    
    def __init__(self, 
                 max_num_seqs: int = 256,
                 max_num_batched_tokens: int = 8192,
                 enable_chunked_prefill: bool = False,
                 max_prefill_chunk_size: int = 512):
        self.max_num_seqs = max_num_seqs
        self.max_num_batched_tokens = max_num_batched_tokens
        self.enable_chunked_prefill = enable_chunked_prefill
        self.max_prefill_chunk_size = max_prefill_chunk_size
        
        self.waiting_queue: List[Request] = []
        self.running: List[Request] = []
        self.completed: List[Request] = []
        
        self.current_time = 0.0
        self.step_count = 0
        self.step_times: List[float] = []
    
    def add_request(self, request: Request):
        self.waiting_queue.append(request)
    
    def _estimate_step_time(self, num_prefill_tokens: int, num_decode_tokens: int) -> float:
        """Estimate step execution time in ms."""
        # Prefill: compute-bound, ~20 tokens/ms on A100
        prefill_time = num_prefill_tokens * 0.05  # ms
        # Decode: memory-bound, ~10ms base + ~0.02ms per token
        decode_time = 5.0 + num_decode_tokens * 0.02 if num_decode_tokens > 0 else 0
        return max(prefill_time, decode_time) + 1.0  # Overhead
    
    def schedule_step(self) -> dict:
        """
        Schedule one iteration.
        
        Returns dict with scheduled work.
        """
        self.step_count += 1
        
        # Count tokens for existing decode requests
        decode_tokens = len([r for r in self.running if not r.is_prefilling])
        
        # Token budget
        token_budget = self.max_num_batched_tokens - decode_tokens
        
        # Schedule prefills from waiting queue
        prefill_tokens = 0
        new_prefills = []
        
        while self.waiting_queue and token_budget > 0:
            request = self.waiting_queue[0]
            
            if self.enable_chunked_prefill:
                # Chunked: take at most chunk_size tokens
                chunk = min(
                    request.prefill_tokens_remaining,
                    self.max_prefill_chunk_size,
                    token_budget,
                )
            else:
                # Non-chunked: must fit entire prefill
                chunk = request.prefill_tokens_remaining
                if chunk > token_budget:
                    break  # Can't fit this prefill
            
            if chunk <= 0:
                break
            
            request.prefill_tokens_remaining -= chunk
            prefill_tokens += chunk
            token_budget -= chunk
            
            if request.prefill_tokens_remaining == 0:
                # Prefill complete, move to running
                self.waiting_queue.pop(0)
                self.running.append(request)
                new_prefills.append(request)
            elif self.enable_chunked_prefill:
                # Partial prefill, keep in waiting but also "running" for this chunk
                if request not in self.running:
                    self.running.append(request)
                new_prefills.append(request)
            
            if not self.enable_chunked_prefill:
                break  # Only one prefill per step without chunking
        
        # Execute step
        step_time = self._estimate_step_time(prefill_tokens, decode_tokens)
        self.current_time += step_time
        self.step_times.append(step_time)
        
        # Update request states
        completed_this_step = []
        for request in self.running:
            if not request.is_prefilling:
                # Decode: generate one token
                request.tokens_generated += 1
                request.token_times.append(self.current_time)
                
                if request.first_token_time is None:
                    request.first_token_time = self.current_time
                
                if request.is_complete:
                    request.completion_time = self.current_time
                    completed_this_step.append(request)
        
        for request in completed_this_step:
            self.running.remove(request)
            self.completed.append(request)
        
        return {
            'step': self.step_count,
            'time': self.current_time,
            'prefill_tokens': prefill_tokens,
            'decode_tokens': decode_tokens,
            'step_time_ms': step_time,
            'running': len(self.running),
            'waiting': len(self.waiting_queue),
            'completed': len(completed_this_step),
        }
    
    def run_until_complete(self, max_steps: int = 10000):
        """Run scheduler until all requests complete."""
        for _ in range(max_steps):
            if not self.waiting_queue and not self.running:
                break
            self.schedule_step()
    
    def get_metrics(self) -> dict:
        all_ttfts = [r.ttft for r in self.completed if r.ttft is not None]
        all_itl_p99s = [r.itl_stats['p99'] for r in self.completed if r.itl_stats['p99'] > 0]
        
        return {
            'avg_ttft_ms': np.mean(all_ttfts) if all_ttfts else 0,
            'p99_ttft_ms': np.percentile(all_ttfts, 99) if all_ttfts else 0,
            'avg_itl_p99_ms': np.mean(all_itl_p99s) if all_itl_p99s else 0,
            'total_time_ms': self.current_time,
            'total_steps': self.step_count,
            'requests_completed': len(self.completed),
        }

print("SchedulerSimulator defined.")
````

````python
# Compare chunked vs non-chunked prefill

def create_workload(num_requests: int = 50, seed: int = 42) -> List[Request]:
    """Create a realistic mix of requests."""
    random.seed(seed)
    requests = []
    for i in range(num_requests):
        prompt_len = random.choice([128, 256, 512, 1024, 2048, 4096])
        max_output = random.randint(50, 200)
        arrival = i * 20  # 20ms between arrivals
        requests.append(Request(
            id=i, prompt_len=prompt_len,
            max_output_len=max_output, arrival_time=arrival
        ))
    return requests

# Run without chunked prefill
workload1 = create_workload(30)
scheduler_no_chunk = SchedulerSimulator(
    enable_chunked_prefill=False,
    max_num_batched_tokens=8192
)
for r in workload1:
    scheduler_no_chunk.add_request(r)
scheduler_no_chunk.run_until_complete()
metrics_no_chunk = scheduler_no_chunk.get_metrics()

# Run with chunked prefill
workload2 = create_workload(30)
scheduler_chunked = SchedulerSimulator(
    enable_chunked_prefill=True,
    max_prefill_chunk_size=512,
    max_num_batched_tokens=8192
)
for r in workload2:
    scheduler_chunked.add_request(r)
scheduler_chunked.run_until_complete()
metrics_chunked = scheduler_chunked.get_metrics()

print(f"{'Metric':<25} {'No Chunking':>15} {'Chunked (512)':>15}")
print("-" * 58)
for key in ['avg_ttft_ms', 'p99_ttft_ms', 'avg_itl_p99_ms', 'total_time_ms', 'total_steps']:
    v1 = metrics_no_chunk[key]
    v2 = metrics_chunked[key]
    print(f"{key:<25} {v1:>14.1f} {v2:>14.1f}")
````

## 2. Chunked Prefill: Design Details

### Token Budget Allocation

```
Per-iteration token budget = max_num_batched_tokens (e.g., 8192)

Budget allocation:
  1. Reserve tokens for running decode requests (1 per sequence)
  2. Allocate remaining budget to prefill chunks

Example:
  Budget = 8192 tokens
  Running decodes = 200 sequences → 200 tokens
  Available for prefill = 8192 - 200 = 7992 tokens
  
  With chunk_size=512:
    Can fit 7992 / 512 = 15 prefill chunks (from different requests)
    Or fewer if requests have less remaining prefill
```

### Source Code: `vllm/core/scheduler.py`

```python
# From vllm/core/scheduler.py (simplified)
def _schedule_prefills(self):
    budget = self.scheduler_config.max_num_batched_tokens
    
    # Reserve for running decodes
    num_running_decode = len(self.running) - len(self.swapped)
    budget -= num_running_decode
    
    if self.scheduler_config.chunked_prefill_enabled:
        # Can schedule partial prefills
        for seq_group in self.waiting:
            num_tokens = min(
                seq_group.get_seqs()[0].get_len() - seq_group.num_computed,
                self.scheduler_config.max_prefill_tokens,  # chunk size
                budget,
            )
            if num_tokens > 0:
                self._schedule_seq_group(seq_group, num_tokens)
                budget -= num_tokens
    else:
        # Must schedule full prefill
        for seq_group in self.waiting:
            num_tokens = seq_group.get_seqs()[0].get_len()
            if num_tokens <= budget:
                self._schedule_seq_group(seq_group, num_tokens)
                budget -= num_tokens
```

## 3. Mixed Prefill-Decode Batching

### How It Works

```
Mixed batch example:

Batch = [
  Seq 0: PREFILL chunk (512 tokens)    ← Attending to its own tokens
  Seq 1: DECODE (1 token)              ← Attending to cached KV (200 tokens)
  Seq 2: DECODE (1 token)              ← Attending to cached KV (500 tokens)
  Seq 3: PREFILL chunk (256 tokens)    ← Attending to its own tokens
  Seq 4: DECODE (1 token)              ← Attending to cached KV (100 tokens)
]

Total tokens in batch = 512 + 1 + 1 + 256 + 1 = 771

Attention computation is different for each type:
  Prefill: Full self-attention within the chunk
  Decode: Single-query attention against KV-cache
```

### FlashAttention with Mixed Batches

vLLM uses FlashAttention's `varlen` API to handle variable-length sequences in a single kernel call:

```python
# Using flash_attn_varlen_func for mixed batches
from flash_attn import flash_attn_varlen_func

output = flash_attn_varlen_func(
    q=query,                    # [total_tokens, num_heads, head_dim]
    k=key,                      # [total_kv_tokens, num_kv_heads, head_dim]
    v=value,                    # [total_kv_tokens, num_kv_heads, head_dim]
    cu_seqlens_q=cu_seqlens_q,  # Cumulative query lengths
    cu_seqlens_k=cu_seqlens_k,  # Cumulative KV lengths
    max_seqlen_q=max_q_len,
    max_seqlen_k=max_kv_len,
)
```

````python
class MixedBatchAnalyzer:
    """
    Analyzes the composition and efficiency of mixed prefill-decode batches.
    """
    
    @staticmethod
    def create_mixed_batch(
        prefill_chunks: List[int],  # Token counts for each prefill chunk
        decode_context_lens: List[int],  # KV cache lengths for decode seqs
    ) -> dict:
        """
        Create a mixed batch and analyze its properties.
        """
        total_query_tokens = sum(prefill_chunks) + len(decode_context_lens)
        total_kv_tokens = sum(prefill_chunks) + sum(decode_context_lens)
        
        # Compute FLOP counts
        head_dim = 128
        num_heads = 32
        
        # Prefill FLOPs: O(seq_len^2 * head_dim) per head
        prefill_flops = sum(chunk * chunk * head_dim * num_heads * 2 for chunk in prefill_chunks)
        
        # Decode FLOPs: O(context_len * head_dim) per head
        decode_flops = sum(ctx * head_dim * num_heads * 2 for ctx in decode_context_lens)
        
        # Memory reads (KV cache)
        kv_memory_bytes = total_kv_tokens * num_heads * head_dim * 2 * 2  # K+V, FP16
        
        return {
            'num_prefill_seqs': len(prefill_chunks),
            'num_decode_seqs': len(decode_context_lens),
            'total_query_tokens': total_query_tokens,
            'total_kv_tokens': total_kv_tokens,
            'prefill_flops': prefill_flops,
            'decode_flops': decode_flops,
            'total_flops': prefill_flops + decode_flops,
            'kv_memory_MB': kv_memory_bytes / 1024 / 1024,
            'prefill_fraction': prefill_flops / (prefill_flops + decode_flops + 1),
        }

# Example mixed batch
batch = MixedBatchAnalyzer.create_mixed_batch(
    prefill_chunks=[512, 256],
    decode_context_lens=[200, 500, 100, 800, 300, 1000, 150, 400]
)

print("=== Mixed Batch Analysis ===")
for k, v in batch.items():
    if isinstance(v, float):
        if 'fraction' in k:
            print(f"  {k}: {v:.1%}")
        elif 'MB' in k:
            print(f"  {k}: {v:.1f}")
        else:
            print(f"  {k}: {v:.0f}")
    else:
        print(f"  {k}: {v:,}")
````

## 4. Impact on TTFT and ITL

````python
# Compare TTFT and ITL for different chunk sizes

chunk_sizes = [64, 128, 256, 512, 1024, 2048, 4096, 0]  # 0 = no chunking
results = {}

for cs in chunk_sizes:
    workload = create_workload(30)
    enabled = cs > 0
    scheduler = SchedulerSimulator(
        enable_chunked_prefill=enabled,
        max_prefill_chunk_size=cs if enabled else 8192,
        max_num_batched_tokens=8192
    )
    for r in workload:
        scheduler.add_request(r)
    scheduler.run_until_complete()
    
    metrics = scheduler.get_metrics()
    metrics['max_step_time'] = max(scheduler.step_times) if scheduler.step_times else 0
    metrics['step_time_std'] = np.std(scheduler.step_times) if scheduler.step_times else 0
    results[cs] = metrics

print(f"{'Chunk Size':>11} {'TTFT avg':>10} {'TTFT p99':>10} {'ITL p99':>10} {'Max Step':>10} {'Step Std':>10}")
print("-" * 65)
for cs in chunk_sizes:
    m = results[cs]
    label = f"{cs}" if cs > 0 else "No chunk"
    print(f"{label:>11} {m['avg_ttft_ms']:>9.1f} {m['p99_ttft_ms']:>9.1f} "
          f"{m['avg_itl_p99_ms']:>9.1f} {m['max_step_time']:>9.1f} {m['step_time_std']:>9.1f}")
````

## 5. How Attention Computation Changes

### Attention in Mixed Batches

```
For prefill tokens: FULL self-attention within the chunk
  Q = chunk tokens, K = chunk tokens, V = chunk tokens
  Attention mask: causal within chunk
  
  ┌─────────┐
  │ ■ . . . │  ← Token 0 attends to itself
  │ ■ ■ . . │  ← Token 1 attends to 0, 1
  │ ■ ■ ■ . │  ← Token 2 attends to 0, 1, 2
  │ ■ ■ ■ ■ │  ← Token 3 attends to all
  └─────────┘

For decode tokens: SINGLE-QUERY attention against KV-cache
  Q = 1 token, K = cached KV, V = cached KV
  No mask needed (single query attends to all)
  
  [■ ■ ■ ■ ■ ■ ... ■]  ← 1 token attends to all cached tokens

For continued prefill chunks (not first chunk):
  Q = chunk tokens, K = cached_prefix + chunk tokens
  Must attend to both cached prefix AND current chunk
```

### Implementation: Attention Metadata

```python
# vLLM creates different attention metadata for mixed batches
@dataclass
class FlashAttentionMetadata:
    # Prefill metadata

    prefill_seq_lens: List[int]      # Length of each prefill sequence

    decode_seq_lens: List[int]       # KV cache length per decode seq
    
    # Slot mapping for cache writes
    slot_mapping: torch.Tensor       # Where to write new KV entries

    block_tables: torch.Tensor       # Physical block mappings
```

````python
def analyze_attention_costs(
    prefill_chunks: List[int],
    decode_context_lens: List[int],
    head_dim: int = 128,
    num_heads: int = 32,
):
    """
    Break down attention computation cost for a mixed batch.
    """
    print("=== Attention Cost Breakdown ===")
    print(f"\nPrefill chunks: {prefill_chunks}")
    print(f"Decode context lengths: {decode_context_lens}")
    
    total_flops = 0
    total_bytes = 0
    
    # Prefill attention
    print(f"\n  Prefill Attention:")
    for i, chunk in enumerate(prefill_chunks):
        # QK^T: [chunk, head_dim] × [head_dim, chunk] = [chunk, chunk]
        qk_flops = chunk * chunk * head_dim * num_heads * 2
        # Softmax: O(chunk^2)
        # AV: [chunk, chunk] × [chunk, head_dim] = [chunk, head_dim]
        av_flops = chunk * chunk * head_dim * num_heads * 2
        
        flops = qk_flops + av_flops
        total_flops += flops
        
        print(f"    Chunk {i} ({chunk} tokens): {flops/1e6:.1f} MFLOPs")
    
    # Decode attention (PagedAttention)
    print(f"\n  Decode Attention:")
    for i, ctx_len in enumerate(decode_context_lens):
        # QK^T: [1, head_dim] × [head_dim, ctx_len] = [1, ctx_len]
        qk_flops = 1 * ctx_len * head_dim * num_heads * 2
        # AV: [1, ctx_len] × [ctx_len, head_dim] = [1, head_dim]
        av_flops = 1 * ctx_len * head_dim * num_heads * 2
        
        flops = qk_flops + av_flops
        total_flops += flops
        
        # Memory: must read entire KV cache
        kv_bytes = ctx_len * num_heads * head_dim * 2 * 2  # K+V, FP16
        total_bytes += kv_bytes
        
        print(f"    Seq {i} (ctx={ctx_len}): {flops/1e6:.1f} MFLOPs, {kv_bytes/1024:.1f} KB KV read")
    
    print(f"\n  Total attention FLOPs: {total_flops/1e9:.2f} GFLOPs")
    print(f"  Total KV cache reads: {total_bytes/1024/1024:.1f} MB")
    print(f"  Arithmetic intensity: {total_flops/max(total_bytes,1):.1f} FLOPs/byte")

analyze_attention_costs(
    prefill_chunks=[512, 256],
    decode_context_lens=[200, 500, 1000, 300]
)
````

## 6. Optimal Chunk Size Selection

````python
def evaluate_chunk_size_for_workload(
    workload_type: str,
    chunk_size: int,
    num_requests: int = 50,
) -> dict:
    """
    Evaluate a chunk size for a specific workload type.
    """
    random.seed(42)
    requests = []
    
    for i in range(num_requests):
        if workload_type == 'chatbot':
            prompt_len = random.randint(100, 500)
            output_len = random.randint(50, 200)
        elif workload_type == 'long_context':
            prompt_len = random.randint(2000, 8000)
            output_len = random.randint(50, 100)
        elif workload_type == 'code_gen':
            prompt_len = random.randint(500, 2000)
            output_len = random.randint(200, 500)
        elif workload_type == 'mixed':
            prompt_len = random.choice([100, 500, 2000, 5000])
            output_len = random.randint(50, 300)
        
        requests.append(Request(
            id=i, prompt_len=prompt_len,
            max_output_len=output_len, arrival_time=i * 15
        ))
    
    enabled = chunk_size > 0
    scheduler = SchedulerSimulator(
        enable_chunked_prefill=enabled,
        max_prefill_chunk_size=chunk_size if enabled else 16384,
        max_num_batched_tokens=8192
    )
    for r in requests:
        scheduler.add_request(r)
    scheduler.run_until_complete()
    
    metrics = scheduler.get_metrics()
    metrics['max_step_time'] = max(scheduler.step_times) if scheduler.step_times else 0
    return metrics

# Test different workloads
workloads = ['chatbot', 'long_context', 'code_gen', 'mixed']
test_chunks = [128, 256, 512, 1024, 2048, 0]

print(f"{'Workload':<15} {'Chunk':>8} {'TTFT':>8} {'ITL p99':>8} {'Max Step':>10}")
print("-" * 55)

best_chunks = {}
for wl in workloads:
    best_itl = float('inf')
    best_cs = 0
    for cs in test_chunks:
        m = evaluate_chunk_size_for_workload(wl, cs)
        label = str(cs) if cs > 0 else 'None'
        print(f"{wl:<15} {label:>8} {m['avg_ttft_ms']:>7.0f} {m['avg_itl_p99_ms']:>7.1f} {m['max_step_time']:>9.1f}")
        
        if m['avg_itl_p99_ms'] < best_itl and cs > 0:
            best_itl = m['avg_itl_p99_ms']
            best_cs = cs
    
    best_chunks[wl] = best_cs
    print()

print("\nBest chunk sizes:")
for wl, cs in best_chunks.items():
    print(f"  {wl}: {cs}")
````

| Scenario | Enable? | Reason |
|----------|---------|--------|
| Online serving (latency-sensitive) | Yes | Consistent ITL |
| Batch processing (throughput) | Maybe not | Slight overhead |
| Long context inputs | Yes | Prevents blocking |
| Short prompts only | Optional | Minimal benefit |

## 8. Summary

### Key Tradeoffs

| Chunk Size | TTFT | ITL P99 | Throughput | GPU Utilization |
|------------|------|---------|------------|----------------|
| No chunking | Best | Worst | Good | Spiky |
| Large (2048) | Good | Better | Good | Better |
| Medium (512) | OK | Good | Good | Consistent |
| Small (128) | Worst | Best | Lower | Very consistent |

### Best Practices
1. **Enable chunked prefill** for production online serving
2. **Chunk size 512-1024** is a good default
3. **Monitor ITL P99** to validate benefit
4. **Larger chunks** for throughput-oriented workloads
5. **Smaller chunks** for strict latency SLAs

````python
# Exercise 2 Starter: Adaptive chunk sizing

class AdaptiveChunkedScheduler(SchedulerSimulator):
    """
    Scheduler with adaptive chunk sizing based on load.
    """
    
    def __init__(self, min_chunk: int = 128, max_chunk: int = 2048, **kwargs):
        super().__init__(enable_chunked_prefill=True, **kwargs)
        self.min_chunk = min_chunk
        self.max_chunk = max_chunk
    
    def _get_adaptive_chunk_size(self) -> int:
        """
        Adjust chunk size based on current load.
        
        High load (many running decodes) → small chunks
        Low load (few running decodes) → large chunks
        """
        num_running = len(self.running)
        load_fraction = num_running / max(self.max_num_seqs, 1)
        
        # Linear interpolation between max and min chunk
        chunk = int(self.max_chunk - (self.max_chunk - self.min_chunk) * load_fraction)
        chunk = max(self.min_chunk, min(self.max_chunk, chunk))
        
        return chunk
    
    def schedule_step(self) -> dict:
        # Update chunk size before scheduling
        self.max_prefill_chunk_size = self._get_adaptive_chunk_size()
        return super().schedule_step()

# Test adaptive scheduler
workload = create_workload(30)
adaptive = AdaptiveChunkedScheduler(
    min_chunk=128, max_chunk=2048,
    max_num_batched_tokens=8192
)
for r in workload:
    adaptive.add_request(r)
adaptive.run_until_complete()

print(f"Adaptive scheduler results:")
m = adaptive.get_metrics()
for k, v in m.items():
    if isinstance(v, float):
        print(f"  {k}: {v:.1f}")
    else:
        print(f"  {k}: {v}")
````
