## 1. MoE Architecture Recap

In a standard Transformer, every token passes through every parameter.
In MoE, each token is routed to a subset of "expert" FFN layers.

```python
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch
import numpy as np

def draw_moe_architecture():
    """MoE architecture: Input -> Router/Gate -> Experts (top-k) -> Combine -> Output."""
    fig, ax = plt.subplots(figsize=(16, 8))
    ax.set_xlim(0, 18); ax.set_ylim(0, 10); ax.axis('off')
    ax.set_title('Mixture of Experts (MoE) Architecture', fontsize=16, fontweight='bold', pad=15)

    # Input
    inp = FancyBboxPatch((0.5, 4.0), 2.0, 2.0, boxstyle='round,pad=0.12',
                         facecolor='#4A90D9', edgecolor='white', linewidth=2, alpha=0.85)
    ax.add_patch(inp)
    ax.text(1.5, 5.0, 'Input\nToken', ha='center', va='center', fontsize=10, fontweight='bold', color='white')

    # Router / Gate
    gate = FancyBboxPatch((3.5, 3.5), 2.5, 3.0, boxstyle='round,pad=0.12',
                          facecolor='#F39C12', edgecolor='white', linewidth=2, alpha=0.85)
    ax.add_patch(gate)
    ax.text(4.75, 5.0, 'Router\n/ Gate\n(softmax)', ha='center', va='center',
            fontsize=10, fontweight='bold', color='white')

    ax.annotate('', xy=(3.5, 5.0), xytext=(2.5, 5.0),
                arrowprops=dict(arrowstyle='->', color='#333', lw=2.5))

    # Experts
    n_experts = 6
    expert_colors = ['#27AE60', '#27AE60', '#E74C3C', '#27AE60', '#27AE60', '#27AE60']
    expert_alphas = [0.3, 0.3, 0.9, 0.9, 0.3, 0.3]  # Only top-k=2 activated (experts 2,3)
    for i in range(n_experts):
        y = 0.5 + i * 1.5
        color = expert_colors[i]
        alpha = expert_alphas[i]
        expert = FancyBboxPatch((7.5, y), 3.0, 1.2, boxstyle='round,pad=0.1',
                                facecolor=color, edgecolor='white', linewidth=1.5, alpha=alpha)
        ax.add_patch(expert)
        status = 'ACTIVE' if alpha > 0.5 else 'inactive'
        ax.text(9.0, y + 0.6, f'Expert {i+1}\n({status})', ha='center', va='center',
                fontsize=8, fontweight='bold', color='white' if alpha > 0.5 else '#999')

        # Arrows from gate
        lw = 2.5 if alpha > 0.5 else 0.8
        ls = '-' if alpha > 0.5 else '--'
        ax.annotate('', xy=(7.5, y + 0.6), xytext=(6.0, 5.0),
                    arrowprops=dict(arrowstyle='->', color=color, lw=lw,
                                    alpha=alpha, linestyle=ls,
                                    connectionstyle='arc3,rad=0'))

    # Top-k annotation
    ax.text(6.8, 8.5, 'Only top-k=2\nexperts activated\nper token', ha='center', fontsize=9,
            color='#E74C3C', fontweight='bold',
            bbox=dict(boxstyle='round,pad=0.3', facecolor='#FFEBEE', edgecolor='#E74C3C'))

    # Combine
    combine = FancyBboxPatch((11.5, 3.5), 2.5, 3.0, boxstyle='round,pad=0.12',
                             facecolor='#8E44AD', edgecolor='white', linewidth=2, alpha=0.85)
    ax.add_patch(combine)
    ax.text(12.75, 5.0, 'Weighted\nCombine\n(sum)', ha='center', va='center',
            fontsize=10, fontweight='bold', color='white')

    # Arrows from active experts to combine
    for i in [2, 3]:
        y = 0.5 + i * 1.5 + 0.6
        ax.annotate('', xy=(11.5, 5.0), xytext=(10.5, y),
                    arrowprops=dict(arrowstyle='->', color='#E74C3C', lw=2))

    # Output
    out = FancyBboxPatch((15.0, 4.0), 2.0, 2.0, boxstyle='round,pad=0.12',
                         facecolor='#4A90D9', edgecolor='white', linewidth=2, alpha=0.85)
    ax.add_patch(out)
    ax.text(16.0, 5.0, 'Output', ha='center', va='center', fontsize=10, fontweight='bold', color='white')

    ax.annotate('', xy=(15.0, 5.0), xytext=(14.0, 5.0),
                arrowprops=dict(arrowstyle='->', color='#333', lw=2.5))

    plt.tight_layout()
    plt.savefig('moe_architecture.png', dpi=150, bbox_inches='tight')
    plt.show()

draw_moe_architecture()
```

```python
@dataclass
class MoEConfig:
    """Configuration for a Mixture of Experts model."""
    name: str
    hidden_size: int
    num_layers: int
    num_experts: int
    top_k: int                    # Experts activated per token
    expert_intermediate_size: int # FFN hidden dim per expert
    num_attention_heads: int
    num_kv_heads: int
    vocab_size: int
    shared_expert: bool = False   # DeepSeek uses shared experts
    num_shared_experts: int = 0
    
    @property
    def total_params_billion(self) -> float:
        """Total parameters including all experts."""
        # Attention params per layer
        qkv_params = self.hidden_size * (self.num_attention_heads + 2 * self.num_kv_heads) * (self.hidden_size // self.num_attention_heads)
        attn_out = self.hidden_size * self.hidden_size
        attn_per_layer = qkv_params + attn_out
        
        # Expert FFN params (each expert: up + gate + down)
        expert_params = 3 * self.hidden_size * self.expert_intermediate_size
        experts_per_layer = expert_params * self.num_experts
        
        # Shared expert params (if any)
        shared_per_layer = expert_params * self.num_shared_experts if self.shared_expert else 0
        
        # Router params
        router_per_layer = self.hidden_size * self.num_experts
        
        # Embeddings
        embed_params = self.vocab_size * self.hidden_size * 2
        
        total = (attn_per_layer + experts_per_layer + shared_per_layer + router_per_layer) * self.num_layers + embed_params
        return total / 1e9
    
    @property
    def active_params_billion(self) -> float:
        """Parameters active for any given token."""
        qkv_params = self.hidden_size * (self.num_attention_heads + 2 * self.num_kv_heads) * (self.hidden_size // self.num_attention_heads)
        attn_out = self.hidden_size * self.hidden_size
        attn_per_layer = qkv_params + attn_out
        
        expert_params = 3 * self.hidden_size * self.expert_intermediate_size
        active_experts_per_layer = expert_params * self.top_k
        shared = expert_params * self.num_shared_experts if self.shared_expert else 0
        
        router_per_layer = self.hidden_size * self.num_experts
        embed_params = self.vocab_size * self.hidden_size * 2
        
        total = (attn_per_layer + active_experts_per_layer + shared + router_per_layer) * self.num_layers + embed_params
        return total / 1e9
    
    @property
    def fp16_size_gb(self) -> float:
        return self.total_params_billion * 2  # 2 bytes per param in FP16


# Model catalog
MOE_MODELS = {
    "Mixtral-8x7B": MoEConfig(
        name="Mixtral-8x7B", hidden_size=4096, num_layers=32,
        num_experts=8, top_k=2, expert_intermediate_size=14336,
        num_attention_heads=32, num_kv_heads=8, vocab_size=32000,
    ),
    "Mixtral-8x22B": MoEConfig(
        name="Mixtral-8x22B", hidden_size=6144, num_layers=56,
        num_experts=8, top_k=2, expert_intermediate_size=16384,
        num_attention_heads=48, num_kv_heads=8, vocab_size=32768,
    ),
    "DeepSeek-V2": MoEConfig(
        name="DeepSeek-V2", hidden_size=5120, num_layers=60,
        num_experts=160, top_k=6, expert_intermediate_size=1536,
        num_attention_heads=128, num_kv_heads=128, vocab_size=102400,
        shared_expert=True, num_shared_experts=2,
    ),
    "DeepSeek-V3": MoEConfig(
        name="DeepSeek-V3", hidden_size=7168, num_layers=61,
        num_experts=256, top_k=8, expert_intermediate_size=2048,
        num_attention_heads=128, num_kv_heads=128, vocab_size=129280,
        shared_expert=True, num_shared_experts=1,
    ),
}

# Display model comparison
print(f"{'Model':<18} {'Experts':>8} {'Top-K':>6} {'Total Params':>13} {'Active Params':>14} {'FP16 Size':>10} {'Ratio':>7}")
print("=" * 85)
for name, m in MOE_MODELS.items():
    ratio = m.active_params_billion / m.total_params_billion
    print(f"{name:<18} {m.num_experts:>8} {m.top_k:>6} {m.total_params_billion:>11.1f}B "
          f"{m.active_params_billion:>12.1f}B {m.fp16_size_gb:>8.1f}GB {ratio:>6.1%}")
```

## 2. Memory Challenges: All Experts Must Be Loaded

```python
def calculate_moe_memory(
    model: MoEConfig,
    max_batch_size: int = 32,
    max_seq_len: int = 4096,
    quantization_bits: int = 16,
    kv_cache_dtype_bits: int = 16,
) -> dict:
    """Calculate detailed memory breakdown for an MoE model."""
    bytes_per_param = quantization_bits / 8
    
    # Model weights
    total_params = model.total_params_billion * 1e9
    weights_gb = total_params * bytes_per_param / (1024**3)
    
    # Expert weights breakdown
    expert_params_per_expert = 3 * model.hidden_size * model.expert_intermediate_size
    expert_weight_per_expert_gb = expert_params_per_expert * bytes_per_param / (1024**3)
    all_experts_gb = expert_weight_per_expert_gb * model.num_experts * model.num_layers
    
    # KV cache
    kv_bytes = kv_cache_dtype_bits / 8
    head_dim = model.hidden_size // model.num_attention_heads
    kv_per_token = 2 * model.num_kv_heads * head_dim * model.num_layers * kv_bytes
    kv_cache_gb = max_batch_size * max_seq_len * kv_per_token / (1024**3)
    
    # Activation memory (rough estimate)
    activation_gb = max_batch_size * max_seq_len * model.hidden_size * 2 / (1024**3) * 2
    
    total_gb = weights_gb + kv_cache_gb + activation_gb
    
    return {
        "model": model.name,
        "total_memory_gb": round(total_gb, 1),
        "weights_gb": round(weights_gb, 1),
        "expert_weights_gb": round(all_experts_gb, 1),
        "expert_weights_pct": round(all_experts_gb / weights_gb * 100, 1),
        "kv_cache_gb": round(kv_cache_gb, 1),
        "activation_gb": round(activation_gb, 1),
        "per_expert_gb": round(expert_weight_per_expert_gb * model.num_layers, 2),
        "min_gpus_80gb": math.ceil(total_gb / (80 * 0.85)),
        "min_gpus_40gb": math.ceil(total_gb / (40 * 0.85)),
    }


print(f"{'Model':<18} {'Total':>8} {'Weights':>8} {'Experts':>8} {'Exp%':>6} {'KV$':>6} {'80GB':>5} {'40GB':>5}")
print("=" * 70)
for name, model in MOE_MODELS.items():
    mem = calculate_moe_memory(model)
    print(f"{name:<18} {mem['total_memory_gb']:>7.1f}G {mem['weights_gb']:>7.1f}G "
          f"{mem['expert_weights_gb']:>7.1f}G {mem['expert_weights_pct']:>5.1f}% "
          f"{mem['kv_cache_gb']:>5.1f}G {mem['min_gpus_80gb']:>5} {mem['min_gpus_40gb']:>5}")

# Quantization impact
print("\nWith 4-bit quantization (AWQ):")
for name, model in MOE_MODELS.items():
    mem = calculate_moe_memory(model, quantization_bits=4)
    print(f"  {name:<18} {mem['total_memory_gb']:>7.1f}GB -> {mem['min_gpus_80gb']} x A100-80GB")
```

## 3. Expert Parallelism Deployment

```python
# Expert parallelism: distribute experts across GPUs

def compute_expert_placement(
    model: MoEConfig,
    num_gpus: int,
    strategy: str = "balanced",  # "balanced", "group", "replicated"
) -> dict:
    """Compute expert placement across GPUs."""
    if strategy == "balanced":
        # Distribute experts evenly across GPUs
        experts_per_gpu = math.ceil(model.num_experts / num_gpus)
        placement = {}
        for i in range(num_gpus):
            start = i * experts_per_gpu
            end = min(start + experts_per_gpu, model.num_experts)
            placement[f"GPU-{i}"] = list(range(start, end))
    
    elif strategy == "group":
        # Group frequently co-activated experts together
        # (Simplified: group by top_k chunks)
        placement = {}
        group_size = model.top_k
        gpu_idx = 0
        for i in range(0, model.num_experts, group_size):
            gpu_key = f"GPU-{gpu_idx % num_gpus}"
            if gpu_key not in placement:
                placement[gpu_key] = []
            placement[gpu_key].extend(range(i, min(i + group_size, model.num_experts)))
            gpu_idx += 1
    
    elif strategy == "replicated":
        # Replicate all experts on each GPU (requires enough memory)
        placement = {
            f"GPU-{i}": list(range(model.num_experts))
            for i in range(num_gpus)
        }
    
    # Calculate communication requirements
    # In EP, tokens need to be routed to the GPU holding the target expert
    # This is an all-to-all communication pattern
    comm_per_token = model.hidden_size * 2  # FP16 bytes
    
    return {
        "strategy": strategy,
        "num_gpus": num_gpus,
        "experts_per_gpu": {k: len(v) for k, v in placement.items()},
        "placement": placement,
        "all_to_all_msg_bytes": comm_per_token,
        "balanced": max(len(v) for v in placement.values()) - min(len(v) for v in placement.values()) <= 1,
    }


# Demo: Mixtral-8x7B expert placement
mixtral = MOE_MODELS["Mixtral-8x7B"]
for strategy in ["balanced", "group"]:
    for num_gpus in [2, 4, 8]:
        result = compute_expert_placement(mixtral, num_gpus, strategy)
        per_gpu = result["experts_per_gpu"]
        print(f"{strategy:<10} {num_gpus} GPUs: {per_gpu} (balanced: {result['balanced']})")

# DeepSeek-V2 with many experts
print("\nDeepSeek-V2 (160 experts):")
deepseek = MOE_MODELS["DeepSeek-V2"]
for num_gpus in [4, 8, 16]:
    result = compute_expert_placement(deepseek, num_gpus, "balanced")
    per_gpu = list(result["experts_per_gpu"].values())
    print(f"  {num_gpus} GPUs: {per_gpu[0]} experts/GPU")
```

## 4. Expert Offloading Strategies

```python
class ExpertOffloader:
    """Simulate expert offloading between GPU and CPU memory."""
    
    def __init__(self, model: MoEConfig, gpu_memory_gb: float = 24,
                 cpu_memory_gb: float = 256, pcie_bw_gb_s: float = 32):
        self.model = model
        self.gpu_memory_gb = gpu_memory_gb
        self.pcie_bw = pcie_bw_gb_s
        
        # Calculate per-expert size
        expert_params = 3 * model.hidden_size * model.expert_intermediate_size
        self.expert_size_gb = expert_params * 2 / (1024**3)  # FP16
        self.expert_size_all_layers = self.expert_size_gb * model.num_layers
        
        # GPU budget for experts (after attention and other overhead)
        overhead_gb = 5  # Attention, embeddings, KV cache
        self.gpu_expert_budget_gb = gpu_memory_gb * 0.85 - overhead_gb
        self.max_experts_on_gpu = int(self.gpu_expert_budget_gb / self.expert_size_all_layers)
        
        # Cache state
        self.gpu_experts: set = set()  # Expert IDs on GPU
        self.cache_hits = 0
        self.cache_misses = 0
    
    def access_experts(self, expert_ids: list[int]) -> float:
        """Access experts, loading from CPU if needed. Returns overhead in ms."""
        overhead_ms = 0
        
        for eid in expert_ids:
            if eid in self.gpu_experts:
                self.cache_hits += 1
            else:
                self.cache_misses += 1
                # Need to load from CPU
                transfer_time = self.expert_size_all_layers / self.pcie_bw * 1000  # ms
                overhead_ms += transfer_time
                
                # Evict LRU if needed
                if len(self.gpu_experts) >= self.max_experts_on_gpu:
                    self.gpu_experts.pop()  # Simple eviction
                self.gpu_experts.add(eid)
        
        return overhead_ms
    
    def simulate_workload(self, num_tokens: int = 10000, seed: int = 42) -> dict:
        """Simulate a workload with expert routing."""
        rng = random.Random(seed)
        total_overhead_ms = 0
        self.cache_hits = 0
        self.cache_misses = 0
        
        # Pre-load most popular experts
        popular = list(range(min(self.max_experts_on_gpu, self.model.num_experts)))
        self.gpu_experts = set(popular)
        
        for _ in range(num_tokens):
            # Route to top_k experts (with realistic skewed distribution)
            # Most tokens go to a few popular experts (power law)
            selected = []
            for _ in range(self.model.top_k):
                # Zipf-like distribution
                expert_id = int(rng.paretovariate(1.5)) % self.model.num_experts
                selected.append(expert_id)
            
            overhead = self.access_experts(selected)
            total_overhead_ms += overhead
        
        total_accesses = self.cache_hits + self.cache_misses
        return {
            "tokens_processed": num_tokens,
            "experts_on_gpu": len(self.gpu_experts),
            "max_experts_on_gpu": self.max_experts_on_gpu,
            "cache_hit_rate": self.cache_hits / total_accesses if total_accesses > 0 else 0,
            "total_overhead_ms": round(total_overhead_ms, 1),
            "avg_overhead_per_token_ms": round(total_overhead_ms / num_tokens, 3),
            "expert_size_gb": round(self.expert_size_all_layers, 3),
        }


# Simulate offloading for different GPU sizes
print("Expert Offloading Analysis: Mixtral-8x7B")
print(f"{'GPU Memory':>10} {'On GPU':>7} {'Hit Rate':>10} {'Overhead/tok':>13}")
print("=" * 45)

for gpu_mem in [16, 24, 40, 48, 80]:
    offloader = ExpertOffloader(MOE_MODELS["Mixtral-8x7B"], gpu_memory_gb=gpu_mem)
    result = offloader.simulate_workload(10000)
    print(f"{gpu_mem:>8}GB {result['experts_on_gpu']:>5}/{MOE_MODELS['Mixtral-8x7B'].num_experts} "
          f"{result['cache_hit_rate']:>9.1%} {result['avg_overhead_per_token_ms']:>11.3f}ms")

print("\nExpert Offloading Analysis: DeepSeek-V2")
for gpu_mem in [24, 40, 80]:
    offloader = ExpertOffloader(MOE_MODELS["DeepSeek-V2"], gpu_memory_gb=gpu_mem)
    result = offloader.simulate_workload(10000)
    print(f"  {gpu_mem}GB: {result['experts_on_gpu']}/{MOE_MODELS['DeepSeek-V2'].num_experts} experts, "
          f"hit rate: {result['cache_hit_rate']:.1%}, overhead: {result['avg_overhead_per_token_ms']:.3f}ms/tok")
```

```python
# Simulate expert utilization patterns

def simulate_expert_utilization(
    model: MoEConfig,
    num_tokens: int = 50000,
    distribution: str = "uniform",  # "uniform", "skewed", "domain"
    seed: int = 42,
) -> dict:
    """Simulate and analyze expert routing patterns."""
    rng = random.Random(seed)
    expert_counts = defaultdict(int)
    
    for _ in range(num_tokens):
        if distribution == "uniform":
            selected = rng.sample(range(model.num_experts), model.top_k)
        elif distribution == "skewed":
            # Some experts are much more popular (power law)
            selected = []
            for _ in range(model.top_k):
                eid = int(rng.paretovariate(1.2)) % model.num_experts
                selected.append(eid)
        elif distribution == "domain":
            # Simulate domain-specific routing
            domain = rng.choice(["code", "math", "text"])
            domain_experts = {
                "code": range(0, model.num_experts // 3),
                "math": range(model.num_experts // 3, 2 * model.num_experts // 3),
                "text": range(2 * model.num_experts // 3, model.num_experts),
            }
            pool = list(domain_experts[domain])
            selected = rng.sample(pool, min(model.top_k, len(pool)))
        
        for eid in selected:
            expert_counts[eid] += 1
    
    counts = list(expert_counts.values())
    total_activations = sum(counts)
    
    # Load balance factor: max/avg (1.0 = perfect balance)
    avg_count = total_activations / model.num_experts
    load_factor = max(counts) / avg_count if avg_count > 0 else 0
    
    # Experts never used
    unused = model.num_experts - len(expert_counts)
    
    return {
        "distribution": distribution,
        "total_activations": total_activations,
        "avg_activations": round(avg_count, 0),
        "max_activations": max(counts),
        "min_activations": min(counts) if counts else 0,
        "load_balance_factor": round(load_factor, 2),
        "unused_experts": unused,
        "top5_experts": sorted(expert_counts.items(), key=lambda x: x[1], reverse=True)[:5],
    }


print("Expert Utilization Analysis: Mixtral-8x7B")
print("=" * 80)
for dist in ["uniform", "skewed", "domain"]:
    result = simulate_expert_utilization(MOE_MODELS["Mixtral-8x7B"], distribution=dist)
    print(f"\n  Distribution: {dist}")
    print(f"    Load balance factor: {result['load_balance_factor']}x (1.0 = perfect)")
    print(f"    Unused experts: {result['unused_experts']}")
    print(f"    Top 5 experts: {result['top5_experts']}")
    print(f"    Min/Max activations: {result['min_activations']}/{result['max_activations']}")
```

```python
# ASSIGNMENT 1 - SOLUTION

def compare_serving_efficiency_solution(
    moe_model, dense_params_b, gpu_memory_gb=80,
    gpu_tflops=312, gpu_memory_bw_gb_s=2039,
):
    results = {}
    
    # MoE
    moe_total_params = moe_model.total_params_billion * 1e9
    moe_active_params = moe_model.active_params_billion * 1e9
    moe_memory_gb = moe_total_params * 2 / (1024**3)
    moe_flops_per_token = 2 * moe_active_params  # 2 FLOPs per param
    moe_bytes_accessed = moe_total_params * 2  # Must load all weights from HBM
    moe_arithmetic_intensity = moe_flops_per_token / moe_bytes_accessed
    
    # Bottleneck analysis (batch_size=1 is memory-bound)
    moe_compute_time = moe_flops_per_token / (gpu_tflops * 1e12)
    moe_memory_time = moe_bytes_accessed / (gpu_memory_bw_gb_s * 1e9)
    moe_time_per_token = max(moe_compute_time, moe_memory_time)
    moe_tokens_per_s = 1 / moe_time_per_token
    
    # Dense
    dense_params = dense_params_b * 1e9
    dense_memory_gb = dense_params * 2 / (1024**3)
    dense_flops = 2 * dense_params
    dense_bytes = dense_params * 2
    dense_ai = dense_flops / dense_bytes
    
    dense_compute_time = dense_flops / (gpu_tflops * 1e12)
    dense_memory_time = dense_bytes / (gpu_memory_bw_gb_s * 1e9)
    dense_time = max(dense_compute_time, dense_memory_time)
    dense_tps = 1 / dense_time
    
    num_gpus_moe = math.ceil(moe_memory_gb / (gpu_memory_gb * 0.85))
    num_gpus_dense = math.ceil(dense_memory_gb / (gpu_memory_gb * 0.85))
    
    return {
        "MoE": {
            "model": moe_model.name,
            "total_params_B": round(moe_model.total_params_billion, 1),
            "active_params_B": round(moe_model.active_params_billion, 1),
            "memory_gb": round(moe_memory_gb, 1),
            "num_gpus": num_gpus_moe,
            "arithmetic_intensity": round(moe_arithmetic_intensity, 3),
            "bottleneck": "memory" if moe_memory_time > moe_compute_time else "compute",
            "theoretical_tok_s": round(moe_tokens_per_s, 0),
        },
        "Dense": {
            "model": f"{dense_params_b}B dense",
            "total_params_B": dense_params_b,
            "active_params_B": dense_params_b,
            "memory_gb": round(dense_memory_gb, 1),
            "num_gpus": num_gpus_dense,
            "arithmetic_intensity": round(dense_ai, 3),
            "bottleneck": "memory" if dense_memory_time > dense_compute_time else "compute",
            "theoretical_tok_s": round(dense_tps, 0),
        },
        "MoE_advantage": {
            "speed_ratio": round(moe_tokens_per_s / dense_tps, 2),
            "memory_ratio": round(moe_memory_gb / dense_memory_gb, 2),
            "compute_ratio": round(moe_active_params / dense_params, 2),
        }
    }

result = compare_serving_efficiency_solution(MOE_MODELS["Mixtral-8x7B"], 70)
print("MoE vs Dense Serving Efficiency")
print("=" * 60)
for category in ["MoE", "Dense"]:
    print(f"\n{category}:")
    for k, v in result[category].items():
        print(f"  {k}: {v}")
print(f"\nMoE Advantage:")
for k, v in result["MoE_advantage"].items():
    print(f"  {k}: {v}x")
```

```python
# ASSIGNMENT 2 - SOLUTION

class ExpertUtilizationProfilerSolution:
    def __init__(self, model):
        self.model = model
        self.activations = defaultdict(lambda: defaultdict(int))
        self.total_tokens = 0
    
    def record(self, layer, expert_ids):
        self.total_tokens += 1
        for eid in expert_ids:
            self.activations[layer][eid] += 1
    
    def analyze(self):
        all_counts = []
        per_layer_stats = {}
        
        for layer in range(self.model.num_layers):
            counts = [self.activations[layer].get(e, 0) for e in range(self.model.num_experts)]
            all_counts.extend(counts)
            
            if any(c > 0 for c in counts):
                avg = statistics.mean(counts)
                std = statistics.stdev(counts) if len(counts) > 1 else 0
                cv = std / avg if avg > 0 else 0
                per_layer_stats[layer] = {
                    "avg": round(avg, 1),
                    "cv": round(cv, 3),
                    "max_expert": counts.index(max(counts)),
                    "min_expert": counts.index(min(counts)),
                }
        
        # Global stats
        global_per_expert = defaultdict(int)
        for layer in self.activations:
            for eid, count in self.activations[layer].items():
                global_per_expert[eid] += count
        
        global_counts = [global_per_expert.get(e, 0) for e in range(self.model.num_experts)]
        
        # Hot/cold experts
        avg_global = statistics.mean(global_counts) if global_counts else 0
        hot = [e for e in range(self.model.num_experts) if global_per_expert[e] > avg_global * 1.5]
        cold = [e for e in range(self.model.num_experts) if global_per_expert[e] < avg_global * 0.5]
        prunable = [e for e in range(self.model.num_experts) if global_per_expert[e] < avg_global * 0.1]
        
        return {
            "total_tokens": self.total_tokens,
            "global_load_balance_cv": round(statistics.stdev(global_counts) / max(1, statistics.mean(global_counts)), 3) if global_counts else 0,
            "hot_experts": hot,
            "cold_experts": cold,
            "prunable_experts": prunable,
            "sample_layer_stats": {k: v for k, v in list(per_layer_stats.items())[:5]},
        }
    
    def print_heatmap(self, max_layers=8):
        print(f"\nExpert Activation Heatmap ({self.model.name})")
        print(f"{'Layer':<6}", end="")
        for e in range(min(16, self.model.num_experts)):
            print(f" E{e:>2}", end="")
        print()
        
        layers_to_show = list(range(min(max_layers, self.model.num_layers)))
        
        for layer in layers_to_show:
            counts = [self.activations[layer].get(e, 0) for e in range(min(16, self.model.num_experts))]
            max_count = max(counts) if counts else 1
            
            print(f"L{layer:<4}", end=" ")
            for c in counts:
                if max_count == 0:
                    char = "."
                else:
                    intensity = c / max_count
                    if intensity > 0.8: char = "#"
                    elif intensity > 0.5: char = "*"
                    elif intensity > 0.2: char = "."
                    else: char = " "
                print(f"  {char} ", end="")
            print()
        
        print("\nLegend: # = high activation, * = medium, . = low, (space) = near zero")


# Simulate and profile
profiler = ExpertUtilizationProfilerSolution(MOE_MODELS["Mixtral-8x7B"])
rng = random.Random(42)

for _ in range(10000):
    for layer in range(MOE_MODELS["Mixtral-8x7B"].num_layers):
        # Skewed routing
        experts = []
        for _ in range(2):
            eid = int(rng.paretovariate(1.5)) % 8
            experts.append(eid)
        profiler.record(layer, experts)

analysis = profiler.analyze()
print("Expert Utilization Analysis:")
print(f"  Load Balance CV: {analysis['global_load_balance_cv']} (0 = perfect, >1 = poor)")
print(f"  Hot experts: {analysis['hot_experts']}")
print(f"  Cold experts: {analysis['cold_experts']}")
print(f"  Prunable: {analysis['prunable_experts']}")

profiler.print_heatmap(max_layers=8)
```

### Assignment 3: Optimize MoE Serving for Your Hardware

```python
# ASSIGNMENT 3: Find the optimal serving configuration for MoE models

def optimize_moe_serving(
    model: MoEConfig,
    available_gpus: list[dict],  # [{"name": "A100-80GB", "count": 4, "memory_gb": 80}]
    target_throughput: float = 1000,  # tokens/s target
    max_latency_ms: float = 5000,
) -> list[dict]:
    """Find optimal MoE serving configurations.
    
    TODO: Evaluate combinations of:
    - Quantization (FP16, FP8, INT4)
    - TP/EP configuration
    - Expert offloading vs all-on-GPU
    
    Rank by: cost efficiency (throughput per dollar)
    """
    quant_options = [
        {"name": "FP16", "size_ratio": 1.0, "speed_ratio": 1.0},
        {"name": "FP8", "size_ratio": 0.5, "speed_ratio": 1.3},
        {"name": "AWQ-4bit", "size_ratio": 0.28, "speed_ratio": 1.6},
    ]
    
    configs = []
    for gpu_spec in available_gpus:
        for quant in quant_options:
            model_mem = model.fp16_size_gb * quant["size_ratio"]
            num_gpus_needed = math.ceil(model_mem / (gpu_spec["memory_gb"] * 0.85))
            
            if num_gpus_needed > gpu_spec["count"]:
                continue
            
            # Estimate throughput
            base_throughput = 2000  # Rough baseline for A100
            throughput = base_throughput * quant["speed_ratio"]
            
            # Estimate latency
            latency = 200 / quant["speed_ratio"] * (1 + 0.1 * num_gpus_needed)
            
            if throughput >= target_throughput and latency <= max_latency_ms:
                configs.append({
                    "gpu": gpu_spec["name"],
                    "num_gpus": num_gpus_needed,
                    "quantization": quant["name"],
                    "model_size_gb": round(model_mem, 1),
                    "throughput": round(throughput, 0),
                    "latency_ms": round(latency, 0),
                })
    
    configs.sort(key=lambda c: c["num_gpus"])
    return configs


gpus = [
    {"name": "A100-80GB", "count": 8, "memory_gb": 80},
    {"name": "L40S-48GB", "count": 4, "memory_gb": 48},
    {"name": "A10G-24GB", "count": 4, "memory_gb": 24},
]

configs = optimize_moe_serving(MOE_MODELS["Mixtral-8x7B"], gpus)
print(f"Optimal configs for Mixtral-8x7B:")
for c in configs:
    print(f"  {c['gpu']} x{c['num_gpus']} | {c['quantization']:<10} | "
          f"{c['model_size_gb']:>5.1f}GB | {c['throughput']:>5.0f} tok/s | {c['latency_ms']:>4.0f}ms")
```

## Summary

Key takeaways for MoE serving:

1. **Memory is the bottleneck**: All experts must fit in GPU memory even though only top-K are active
2. **Quantization is highly effective**: 4-bit quantization reduces MoE memory by 4x with minimal quality loss
3. **Expert parallelism** distributes experts across GPUs with all-to-all communication
4. **Expert offloading** to CPU is viable when cache hit rates are high (>90%)
5. **Load balance matters**: Skewed expert routing wastes GPU resources
6. **MoE models compute less per token** than equally-sized dense models, leading to better throughput
7. **DeepSeek's fine-grained MoE** (160+ experts) requires more GPUs but has better expert specialization
