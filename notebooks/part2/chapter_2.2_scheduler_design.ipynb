## 1. Why Is the Scheduler the Brain of LLM Serving?

In traditional web servers, a request uses CPU for a few milliseconds and returns. In LLM serving, a single request can occupy **gigabytes of GPU memory** (KV-cache) and require **seconds of continuous GPU computation**. The scheduler must answer hard questions every few milliseconds:

- **Which requests get GPU time NOW?** (out of potentially hundreds waiting)
- **Should we start a new request** or continue existing ones? (prefill vs decode priority)
- **Is there enough memory** for this new request's KV-cache?
- **Should we evict a running request** to make room for a higher-priority one?
- **How do we meet SLO guarantees** while maximizing throughput?

### The Scheduling Loop

The scheduler runs in a tight loop, making decisions at every iteration:

```
while True:
    # 1. Check for completed requests (hit EOS or max_tokens)
    # 2. Free memory from completed requests
    # 3. Check if waiting requests can be admitted (memory check)
    # 4. Decide the batch for this iteration
    # 5. Execute the batch on GPU
    # 6. Process results (sampling, stopping criteria)
```

````python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np
from dataclasses import dataclass, field
from typing import Optional, List, Dict, Tuple
from enum import Enum
import heapq
import random
import time
import warnings
warnings.filterwarnings('ignore')

class RequestState(Enum):
    WAITING = "waiting"      # In queue, waiting to be scheduled
    RUNNING = "running"      # Currently being processed on GPU
    PREEMPTED = "preempted"  # Evicted from GPU, waiting to resume
    COMPLETED = "completed"  # Finished generation

@dataclass
class SchedulerRequest:
    """A request from the scheduler's perspective."""
    request_id: str
    arrival_time: float
    prompt_tokens: int
    max_gen_tokens: int
    priority: int = 0  # Higher = more important
    slo_ttft_ms: float = 5000  # TTFT SLO in milliseconds
    slo_tpot_ms: float = 100   # TPOT SLO in milliseconds
    
    # Runtime state
    state: RequestState = RequestState.WAITING
    generated_tokens: int = 0
    kv_cache_blocks: int = 0  # Number of KV-cache blocks allocated
    
    # Timing
    first_scheduled_time: Optional[float] = None
    first_token_time: Optional[float] = None
    completion_time: Optional[float] = None
    total_preemptions: int = 0
    
    # For timeline visualization
    timeline: List[Tuple[float, float, str]] = field(default_factory=list)  # (start, end, state)
    
    @property
    def remaining_tokens(self):
        return self.max_gen_tokens - self.generated_tokens
    
    @property
    def total_tokens(self):
        return self.prompt_tokens + self.generated_tokens
    
    def __lt__(self, other):
        return self.arrival_time < other.arrival_time

print("Request and state models defined.")
print(f"States: {[s.value for s in RequestState]}")
````

## 2. FCFS (First Come First Served) Scheduler

The simplest scheduling policy: serve requests in the order they arrive.

### Algorithm
1. Maintain a FIFO queue of waiting requests
2. Each iteration, admit as many requests from the queue as memory allows
3. No preemption: once a request starts, it runs to completion

### Pros
- Simple to implement and reason about
- Fair in a strict ordering sense
- No starvation

### Cons
- **Head-of-line blocking**: A long request delays all subsequent ones
- **No SLO awareness**: Cannot prioritize urgent requests
- **Poor memory utilization**: Cannot preempt to reclaim memory

````python
class FCFSScheduler:
    """First Come First Served scheduler."""
    
    def __init__(self, max_batch_size: int = 8, 
                 total_gpu_blocks: int = 1000,
                 block_size: int = 16):  # tokens per block
        self.max_batch_size = max_batch_size
        self.total_gpu_blocks = total_gpu_blocks
        self.block_size = block_size
        
        self.waiting_queue: List[SchedulerRequest] = []
        self.running_batch: List[SchedulerRequest] = []
        self.completed: List[SchedulerRequest] = []
        self.free_blocks = total_gpu_blocks
        self.current_time = 0.0
        
        # Metrics
        self.schedule_log: List[dict] = []
    
    def add_request(self, req: SchedulerRequest):
        """Add a new request to the waiting queue."""
        req.state = RequestState.WAITING
        self.waiting_queue.append(req)
    
    def _blocks_needed(self, req: SchedulerRequest) -> int:
        """Calculate KV-cache blocks needed for a request."""
        total_tokens = req.prompt_tokens + req.max_gen_tokens
        return (total_tokens + self.block_size - 1) // self.block_size
    
    def _blocks_used(self, req: SchedulerRequest) -> int:
        """Blocks currently used by a running request."""
        return (req.total_tokens + self.block_size - 1) // self.block_size
    
    def schedule(self) -> dict:
        """Make scheduling decision for current iteration.
        Returns dict with prefill and decode batches."""
        
        # Step 1: Check for completed requests
        newly_completed = []
        for req in self.running_batch:
            if req.generated_tokens >= req.max_gen_tokens:
                req.state = RequestState.COMPLETED
                req.completion_time = self.current_time
                self.free_blocks += req.kv_cache_blocks
                newly_completed.append(req)
        
        for req in newly_completed:
            self.running_batch.remove(req)
            self.completed.append(req)
        
        # Step 2: Admit new requests from waiting queue (FCFS order)
        prefill_batch = []
        while (self.waiting_queue and 
               len(self.running_batch) + len(prefill_batch) < self.max_batch_size):
            req = self.waiting_queue[0]
            blocks_needed = self._blocks_needed(req)
            
            if blocks_needed <= self.free_blocks:
                self.waiting_queue.pop(0)
                req.state = RequestState.RUNNING
                req.kv_cache_blocks = blocks_needed
                self.free_blocks -= blocks_needed
                req.first_scheduled_time = self.current_time
                prefill_batch.append(req)
            else:
                break  # FCFS: can't skip ahead
        
        # Step 3: Build decode batch (all running requests)
        decode_batch = list(self.running_batch)
        
        # Step 4: Move prefill requests to running
        self.running_batch.extend(prefill_batch)
        
        decision = {
            'time': self.current_time,
            'prefill': [r.request_id for r in prefill_batch],
            'decode': [r.request_id for r in decode_batch],
            'completed': [r.request_id for r in newly_completed],
            'waiting': len(self.waiting_queue),
            'free_blocks': self.free_blocks,
        }
        self.schedule_log.append(decision)
        return decision
    
    def step(self):
        """Simulate one iteration: schedule + execute."""
        decision = self.schedule()
        
        # Simulate execution: each running request generates one token
        for req in self.running_batch:
            req.generated_tokens += 1
            if req.generated_tokens == 1:
                req.first_token_time = self.current_time
        
        self.current_time += 0.025  # 25ms per iteration
        return decision

# Simulate FCFS scheduling
fcfs = FCFSScheduler(max_batch_size=4, total_gpu_blocks=200, block_size=16)

# Create requests with varying sizes
requests_fcfs = [
    SchedulerRequest("A", 0.0, prompt_tokens=64, max_gen_tokens=20),
    SchedulerRequest("B", 0.01, prompt_tokens=128, max_gen_tokens=50),
    SchedulerRequest("C", 0.02, prompt_tokens=32, max_gen_tokens=10),
    SchedulerRequest("D", 0.03, prompt_tokens=256, max_gen_tokens=30),
    SchedulerRequest("E", 0.05, prompt_tokens=64, max_gen_tokens=15),
    SchedulerRequest("F", 0.1, prompt_tokens=128, max_gen_tokens=25),
]

for req in requests_fcfs:
    fcfs.add_request(req)

# Run simulation
print("=== FCFS Scheduling Simulation ===")
for i in range(80):
    decision = fcfs.step()
    if decision['prefill'] or decision['completed']:
        print(f"  t={decision['time']:.3f}s: prefill={decision['prefill']}, "
              f"decode={decision['decode']}, completed={decision['completed']}, "
              f"waiting={decision['waiting']}")
    if not fcfs.running_batch and not fcfs.waiting_queue:
        break

print(f"\n=== Results ===")
for req in fcfs.completed:
    ttft = (req.first_token_time - req.arrival_time) * 1000
    total = (req.completion_time - req.arrival_time) * 1000
    print(f"  Req {req.request_id}: TTFT={ttft:.0f}ms, Total={total:.0f}ms, "
          f"Tokens={req.generated_tokens}")
````

## 3. Priority-Based Scheduler

In production, not all requests are equal. Some users pay more, some tasks are time-sensitive, some requests are about to miss their SLO deadlines.

### Priority Sources

| Priority Factor | Example | Weight |
|----------------|---------|--------|
| User tier | Premium vs free | High |
| SLO urgency | Time until SLO violation | Critical |
| Request age | Prevent starvation | Medium |
| Estimated cost | Prefer short requests | Low |

### SLO-Aware Priority

The most sophisticated approach calculates priority dynamically based on how close a request is to violating its SLO:

$$\text{priority} = \frac{\text{time\_remaining\_before\_SLO\_violation}}{\text{estimated\_remaining\_work}}$$

Lower ratio = higher urgency = should be scheduled first.

````python
class PriorityScheduler:
    """Priority-based scheduler with SLO awareness."""
    
    def __init__(self, max_batch_size: int = 8,
                 total_gpu_blocks: int = 1000,
                 block_size: int = 16):
        self.max_batch_size = max_batch_size
        self.total_gpu_blocks = total_gpu_blocks
        self.block_size = block_size
        
        self.waiting_queue: List[SchedulerRequest] = []
        self.running_batch: List[SchedulerRequest] = []
        self.completed: List[SchedulerRequest] = []
        self.free_blocks = total_gpu_blocks
        self.current_time = 0.0
        self.iteration_time = 0.025  # 25ms per step
        self.schedule_log = []
    
    def _calculate_priority(self, req: SchedulerRequest) -> float:
        """Calculate dynamic priority score. Higher = should be scheduled sooner."""
        wait_time = self.current_time - req.arrival_time
        
        if req.state == RequestState.WAITING:
            # For waiting requests: urgency based on TTFT SLO
            slo_remaining = (req.slo_ttft_ms / 1000) - wait_time
            estimated_work = req.prompt_tokens * 0.0001  # prefill time estimate
            urgency = max(0.001, slo_remaining) / max(0.001, estimated_work)
            # Lower urgency ratio = more urgent = higher priority score
            slo_priority = 100.0 / max(0.01, urgency)
        else:
            slo_priority = 50.0  # Running requests get medium base priority
        
        # Combine with static priority and age
        age_bonus = wait_time * 10  # Prevent starvation
        return req.priority * 20 + slo_priority + age_bonus
    
    def _blocks_needed(self, req: SchedulerRequest) -> int:
        total_tokens = req.prompt_tokens + req.max_gen_tokens
        return (total_tokens + self.block_size - 1) // self.block_size
    
    def add_request(self, req: SchedulerRequest):
        req.state = RequestState.WAITING
        self.waiting_queue.append(req)
    
    def schedule(self) -> dict:
        # Step 1: Remove completed requests
        newly_completed = []
        for req in self.running_batch:
            if req.generated_tokens >= req.max_gen_tokens:
                req.state = RequestState.COMPLETED
                req.completion_time = self.current_time
                self.free_blocks += req.kv_cache_blocks
                newly_completed.append(req)
        
        for req in newly_completed:
            self.running_batch.remove(req)
            self.completed.append(req)
        
        # Step 2: Sort waiting queue by priority (highest first)
        self.waiting_queue.sort(key=lambda r: self._calculate_priority(r), reverse=True)
        
        # Step 3: Admit requests by priority
        prefill_batch = []
        still_waiting = []
        for req in self.waiting_queue:
            if len(self.running_batch) + len(prefill_batch) >= self.max_batch_size:
                still_waiting.append(req)
                continue
            
            blocks_needed = self._blocks_needed(req)
            if blocks_needed <= self.free_blocks:
                req.state = RequestState.RUNNING
                req.kv_cache_blocks = blocks_needed
                self.free_blocks -= blocks_needed
                req.first_scheduled_time = self.current_time
                prefill_batch.append(req)
            else:
                still_waiting.append(req)
        
        self.waiting_queue = still_waiting
        decode_batch = list(self.running_batch)
        self.running_batch.extend(prefill_batch)
        
        decision = {
            'time': self.current_time,
            'prefill': [(r.request_id, f"pri={self._calculate_priority(r):.1f}") for r in prefill_batch],
            'decode': [r.request_id for r in decode_batch],
            'completed': [r.request_id for r in newly_completed],
            'waiting': len(self.waiting_queue),
        }
        self.schedule_log.append(decision)
        return decision
    
    def step(self):
        decision = self.schedule()
        for req in self.running_batch:
            req.generated_tokens += 1
            if req.generated_tokens == 1:
                req.first_token_time = self.current_time
        self.current_time += self.iteration_time
        return decision

# Simulate priority scheduling
priority_sched = PriorityScheduler(max_batch_size=3, total_gpu_blocks=200, block_size=16)

# Same requests but with different priorities and SLOs
requests_pri = [
    SchedulerRequest("A", 0.0, 64, 20, priority=1, slo_ttft_ms=500),    # Low priority, tight SLO
    SchedulerRequest("B", 0.01, 128, 50, priority=3, slo_ttft_ms=2000),  # High priority
    SchedulerRequest("C", 0.02, 32, 10, priority=1, slo_ttft_ms=5000),   # Low priority, loose SLO
    SchedulerRequest("D", 0.03, 256, 30, priority=5, slo_ttft_ms=1000),  # Highest priority
    SchedulerRequest("E", 0.05, 64, 15, priority=2, slo_ttft_ms=3000),   # Medium priority
    SchedulerRequest("F", 0.1, 128, 25, priority=4, slo_ttft_ms=1500),   # High priority, medium SLO
]

for req in requests_pri:
    priority_sched.add_request(req)

print("=== Priority Scheduling Simulation ===")
for i in range(80):
    decision = priority_sched.step()
    if decision['prefill'] or decision['completed']:
        print(f"  t={decision['time']:.3f}s: prefill={decision['prefill']}, "
              f"completed={decision['completed']}, waiting={decision['waiting']}")
    if not priority_sched.running_batch and not priority_sched.waiting_queue:
        break

print(f"\n=== Results ===")
for req in sorted(priority_sched.completed, key=lambda r: r.request_id):
    ttft = (req.first_token_time - req.arrival_time) * 1000
    total = (req.completion_time - req.arrival_time) * 1000
    slo_met = ttft <= req.slo_ttft_ms
    print(f"  Req {req.request_id} (pri={req.priority}): TTFT={ttft:.0f}ms "
          f"(SLO={req.slo_ttft_ms}ms, {'MET' if slo_met else 'VIOLATED'}), "
          f"Total={total:.0f}ms")
````

## 4. Preemption: When and How to Preempt Running Requests

Sometimes, even with priority scheduling, we need to **stop a running request** to make room for a more important one. This is called **preemption**.

### Why Preempt?

1. A high-priority request arrives but GPU memory is full
2. A request is about to violate its SLO and needs more GPU resources
3. The batch is too large, causing unacceptable per-token latency for all requests

### Two Preemption Strategies

#### Strategy 1: Swap to CPU
- Save the KV-cache from GPU to CPU memory
- When the request is re-scheduled, copy KV-cache back to GPU
- **Pro**: No redundant computation
- **Con**: PCIe bandwidth is limited; large KV-caches take time to transfer

#### Strategy 2: Recomputation
- Discard the KV-cache entirely
- When re-scheduled, recompute from scratch (run prefill again)
- **Pro**: No CPU memory needed, no transfer overhead
- **Con**: Wastes the compute already done

````python
# When to choose Swap vs Recomputation

def analyze_preemption_strategy(kv_cache_size_mb: float, 
                                 prompt_tokens: int,
                                 pcie_bandwidth_gbps: float = 32.0,
                                 prefill_speed_tokens_per_sec: float = 10000):
    """Compare swap vs recomputation cost for a given request."""
    
    # Swap cost: transfer KV-cache to CPU and back
    swap_time_ms = 2 * (kv_cache_size_mb / 1024) / pcie_bandwidth_gbps * 1000  # round trip
    
    # Recomputation cost: re-run prefill
    recompute_time_ms = prompt_tokens / prefill_speed_tokens_per_sec * 1000
    
    return {
        'kv_cache_mb': kv_cache_size_mb,
        'prompt_tokens': prompt_tokens,
        'swap_time_ms': swap_time_ms,
        'recompute_time_ms': recompute_time_ms,
        'recommendation': 'swap' if swap_time_ms < recompute_time_ms else 'recompute',
    }

# Analyze for different scenarios
print("=== Preemption Strategy Analysis ===")
print(f"{'KV-Cache (MB)':>14} {'Prompt Tokens':>14} {'Swap (ms)':>10} {'Recompute (ms)':>15} {'Recommendation':>15}")
print("-" * 70)

scenarios = [
    (10, 128),    # Small cache, short prompt
    (50, 512),    # Medium cache, medium prompt
    (200, 2048),  # Large cache, long prompt
    (500, 4096),  # Very large cache, very long prompt
    (1000, 8192), # Huge cache, huge prompt
    (5, 1024),    # Small cache but long prompt (recompute wins)
    (500, 256),   # Large cache but short prompt (swap might lose)
]

swap_times = []
recomp_times = []

for kv_mb, prompt_tok in scenarios:
    result = analyze_preemption_strategy(kv_mb, prompt_tok)
    swap_times.append(result['swap_time_ms'])
    recomp_times.append(result['recompute_time_ms'])
    print(f"{kv_mb:>14} {prompt_tok:>14} {result['swap_time_ms']:>10.2f} "
          f"{result['recompute_time_ms']:>15.2f} {result['recommendation']:>15}")
````

## 5. Preemptive Scheduler Implementation

Let's implement a full preemptive scheduler that can evict running requests to make room for higher-priority ones.

````python
class PreemptiveScheduler:
    """Scheduler with preemption support (swap or recompute)."""
    
    def __init__(self, max_batch_size=4, total_gpu_blocks=100,
                 block_size=16, preemption_mode='swap'):
        self.max_batch_size = max_batch_size
        self.total_gpu_blocks = total_gpu_blocks
        self.block_size = block_size
        self.preemption_mode = preemption_mode
        
        self.waiting_queue = []
        self.running_batch = []
        self.preempted_queue = []
        self.completed = []
        self.free_blocks = total_gpu_blocks
        self.current_time = 0.0
        self.iteration_time = 0.025
        self.preemption_events = []
    
    def _blocks_needed(self, req):
        return (req.prompt_tokens + req.max_gen_tokens + self.block_size - 1) // self.block_size
    
    def _priority(self, req):
        return req.priority * 100 + (self.current_time - req.arrival_time) * 10
    
    def add_request(self, req):
        req.state = RequestState.WAITING
        self.waiting_queue.append(req)
    
    def _try_preempt(self, needed_blocks: int, min_priority: float) -> List[SchedulerRequest]:
        """Try to preempt running requests to free up blocks.
        Returns list of preempted requests, or empty list if not possible."""
        # Sort running batch by priority (lowest first - preempt least important)
        candidates = sorted(self.running_batch, key=lambda r: self._priority(r))
        
        preempted = []
        freed = 0
        for req in candidates:
            if self._priority(req) >= min_priority:
                break  # Don't preempt requests with higher priority
            
            preempted.append(req)
            freed += req.kv_cache_blocks
            if freed >= needed_blocks:
                break
        
        if freed >= needed_blocks:
            return preempted
        return []  # Can't free enough
    
    def schedule(self):
        # Step 1: Complete finished requests
        newly_completed = []
        for req in self.running_batch:
            if req.generated_tokens >= req.max_gen_tokens:
                req.state = RequestState.COMPLETED
                req.completion_time = self.current_time
                self.free_blocks += req.kv_cache_blocks
                newly_completed.append(req)
        for req in newly_completed:
            self.running_batch.remove(req)
            self.completed.append(req)
        
        # Step 2: Try to admit preempted requests first (they already did work)
        readmitted = []
        still_preempted = []
        for req in sorted(self.preempted_queue, key=lambda r: self._priority(r), reverse=True):
            blocks = self._blocks_needed(req)
            if (len(self.running_batch) + len(readmitted) < self.max_batch_size and
                blocks <= self.free_blocks):
                req.state = RequestState.RUNNING
                req.kv_cache_blocks = blocks
                self.free_blocks -= blocks
                readmitted.append(req)
            else:
                still_preempted.append(req)
        self.preempted_queue = still_preempted
        self.running_batch.extend(readmitted)
        
        # Step 3: Admit new requests, with preemption if needed
        prefill_batch = []
        self.waiting_queue.sort(key=lambda r: self._priority(r), reverse=True)
        still_waiting = []
        
        for req in self.waiting_queue:
            if len(self.running_batch) + len(prefill_batch) >= self.max_batch_size:
                still_waiting.append(req)
                continue
            
            blocks = self._blocks_needed(req)
            if blocks <= self.free_blocks:
                req.state = RequestState.RUNNING
                req.kv_cache_blocks = blocks
                self.free_blocks -= blocks
                req.first_scheduled_time = self.current_time
                prefill_batch.append(req)
            else:
                # Try preemption
                preemptable = self._try_preempt(blocks - self.free_blocks, self._priority(req))
                if preemptable:
                    for p in preemptable:
                        p.state = RequestState.PREEMPTED
                        p.total_preemptions += 1
                        self.free_blocks += p.kv_cache_blocks
                        self.running_batch.remove(p)
                        
                        if self.preemption_mode == 'recompute':
                            p.generated_tokens = 0  # Must redo all work
                        
                        self.preempted_queue.append(p)
                        self.preemption_events.append({
                            'time': self.current_time,
                            'preempted': p.request_id,
                            'by': req.request_id,
                            'mode': self.preemption_mode,
                        })
                    
                    req.state = RequestState.RUNNING
                    req.kv_cache_blocks = blocks
                    self.free_blocks -= blocks
                    req.first_scheduled_time = self.current_time
                    prefill_batch.append(req)
                else:
                    still_waiting.append(req)
        
        self.waiting_queue = still_waiting
        self.running_batch.extend(prefill_batch)
        
        return {
            'time': self.current_time,
            'prefill': [r.request_id for r in prefill_batch],
            'decode': [r.request_id for r in self.running_batch if r not in prefill_batch],
            'completed': [r.request_id for r in newly_completed],
            'preempted': [e['preempted'] for e in self.preemption_events if e['time'] == self.current_time],
            'readmitted': [r.request_id for r in readmitted],
        }
    
    def step(self):
        decision = self.schedule()
        for req in self.running_batch:
            req.generated_tokens += 1
            if req.generated_tokens == 1:
                req.first_token_time = self.current_time
        self.current_time += self.iteration_time
        return decision

# Test preemptive scheduling
print("=== Preemptive Scheduling (Swap Mode) ===")
preemptive = PreemptiveScheduler(max_batch_size=3, total_gpu_blocks=50, 
                                  block_size=16, preemption_mode='swap')

# Low priority request arrives first and uses lots of memory
preemptive.add_request(SchedulerRequest("LP1", 0.0, 128, 40, priority=1))
preemptive.add_request(SchedulerRequest("LP2", 0.01, 64, 30, priority=1))
preemptive.add_request(SchedulerRequest("LP3", 0.02, 96, 35, priority=1))

# Run a few steps
for _ in range(5):
    d = preemptive.step()

# High priority request arrives!
preemptive.add_request(SchedulerRequest("HP1", preemptive.current_time, 64, 20, priority=10))
print(f"\n--- High-priority request HP1 arrives at t={preemptive.current_time:.3f}s ---\n")

for i in range(60):
    d = preemptive.step()
    if d['prefill'] or d['completed'] or d['preempted'] or d.get('readmitted'):
        parts = [f"t={d['time']:.3f}s:"]
        if d['preempted']: parts.append(f"PREEMPTED={d['preempted']}")
        if d['prefill']: parts.append(f"prefill={d['prefill']}")
        if d.get('readmitted'): parts.append(f"readmit={d['readmitted']}")
        if d['completed']: parts.append(f"completed={d['completed']}")
        print("  " + " | ".join(parts))
    if not preemptive.running_batch and not preemptive.waiting_queue and not preemptive.preempted_queue:
        break

print(f"\n=== Preemption Events ===")
for event in preemptive.preemption_events:
    print(f"  t={event['time']:.3f}s: {event['preempted']} preempted by {event['by']} ({event['mode']})")

print(f"\n=== Final Results ===")
for req in sorted(preemptive.completed, key=lambda r: r.request_id):
    ttft = (req.first_token_time - req.arrival_time) * 1000 if req.first_token_time else float('inf')
    total = (req.completion_time - req.arrival_time) * 1000
    print(f"  {req.request_id} (pri={req.priority}): TTFT={ttft:.0f}ms, Total={total:.0f}ms, "
          f"Preemptions={req.total_preemptions}")
````

## 6. Admission Control: Scheduling with Limited GPU Memory

GPU memory is the most precious resource in LLM serving. The scheduler must perform **admission control** to prevent out-of-memory errors.

### Memory Budget Components

1. **Model weights**: Fixed cost, loaded once
2. **KV-cache**: Variable, proportional to (batch_size * seq_len * num_layers * hidden_dim)
3. **Activation memory**: Variable, needed during forward pass
4. **CUDA overhead**: Driver allocations, fragmentation

### KV-Cache Memory Formula

For a single request with sequence length $s$:

$$\text{KV\_cache\_bytes} = 2 \times n_{\text{layers}} \times s \times d_{\text{model}} \times \text{sizeof}(\text{dtype})$$

For Llama-3-8B (32 layers, 4096 hidden, FP16):
- Per token: $2 \times 32 \times 4096 \times 2 = 524,288$ bytes = **0.5 MB per token!**
- A 4096-token sequence: ~2 GB of KV-cache

````python
# KV-Cache Memory Calculator

def kv_cache_memory(num_layers, hidden_dim, seq_len, dtype_bytes=2, 
                     num_kv_heads=None):
    """Calculate KV-cache memory for one request.
    
    If num_kv_heads is set (GQA), uses that instead of full hidden_dim.
    """
    if num_kv_heads:
        head_dim = hidden_dim // 32  # assuming 32 total heads
        kv_dim = num_kv_heads * head_dim
    else:
        kv_dim = hidden_dim
    
    # 2 for K and V
    return 2 * num_layers * seq_len * kv_dim * dtype_bytes

# Calculate for popular models
models = [
    ("Llama-3-8B", 32, 4096, None, 8192),
    ("Llama-3-70B", 80, 8192, 8, 8192),  # GQA with 8 KV heads
    ("Llama-3-405B", 126, 16384, 8, 131072),
    ("Qwen2.5-72B", 80, 8192, 8, 131072),
]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: KV-cache per token
ax = axes[0]
model_names = []
per_token_mb = []

print("=== KV-Cache Memory Analysis ===")
print(f"{'Model':<20} {'Per Token (KB)':<16} {'1K tokens (MB)':<16} {'4K tokens (MB)':<16} {'Max Context (GB)':<16}")
print("-" * 84)

for name, layers, hidden, kv_heads, max_ctx in models:
    mem_1 = kv_cache_memory(layers, hidden, 1, num_kv_heads=kv_heads)
    mem_1k = kv_cache_memory(layers, hidden, 1024, num_kv_heads=kv_heads)
    mem_4k = kv_cache_memory(layers, hidden, 4096, num_kv_heads=kv_heads)
    mem_max = kv_cache_memory(layers, hidden, max_ctx, num_kv_heads=kv_heads)
    
    model_names.append(name)
    per_token_mb.append(mem_1 / 1024)
    
    print(f"{name:<20} {mem_1/1024:<16.1f} {mem_1k/1024/1024:<16.1f} "
          f"{mem_4k/1024/1024:<16.1f} {mem_max/1024/1024/1024:<16.2f}")

ax.bar(model_names, per_token_mb, color=['#3498db', '#e74c3c', '#2ecc71', '#f39c12'], alpha=0.8)
ax.set_ylabel('KV-Cache per Token (KB)')
ax.set_title('KV-Cache Memory per Token by Model', fontweight='bold')
ax.grid(axis='y', alpha=0.3)
plt.setp(ax.get_xticklabels(), rotation=15)

# Plot 2: How many concurrent requests fit?
ax = axes[1]
gpu_memory_gb = 80  # A100-80GB
model_weight_gb = [16, 140, 810, 144]  # approximate FP16 weights

for i, (name, layers, hidden, kv_heads, max_ctx) in enumerate(models):
    if model_weight_gb[i] > gpu_memory_gb:
        continue  # Skip models that don't fit on single GPU
    
    available_gb = gpu_memory_gb - model_weight_gb[i] - 2  # 2GB overhead
    seq_lens = np.arange(128, min(max_ctx, 8192), 128)
    max_concurrent = []
    for sl in seq_lens:
        mem_per_req = kv_cache_memory(layers, hidden, sl, num_kv_heads=kv_heads)
        max_concurrent.append(available_gb * 1024**3 / mem_per_req)
    
    ax.plot(seq_lens, max_concurrent, '-', label=name, linewidth=2)

ax.set_xlabel('Sequence Length (tokens)')
ax.set_ylabel('Max Concurrent Requests')
ax.set_title(f'Max Concurrent Requests on A100-80GB', fontweight='bold')
ax.legend()
ax.grid(alpha=0.3)
ax.set_ylim(0)

plt.tight_layout()
plt.show()
````

## 7. Fairness vs Efficiency Tradeoff

A fundamental tension in scheduler design:

- **Fairness**: Every request should get reasonable service, regardless of priority
- **Efficiency**: Maximize GPU utilization and total throughput
- **SLO compliance**: Meet latency guarantees for paying customers

These goals often conflict:
- Batching more requests improves throughput but increases per-request latency
- Prioritizing high-priority requests may starve low-priority ones
- Preemption recovers resources but wastes the work already done

### Strategies to Balance

1. **Aging**: Increase priority of waiting requests over time to prevent starvation
2. **Weighted fair queueing**: Allocate GPU time proportionally to priority weights
3. **Reservation**: Reserve some batch slots for each priority tier
4. **Deadline-aware**: Only preempt when SLO violation is imminent

````python
# Compare FCFS vs Priority vs Preemptive scheduling

def run_simulation(scheduler_class, requests_data, **kwargs):
    """Run a scheduling simulation and return metrics."""
    sched = scheduler_class(**kwargs)
    
    # Add all requests
    requests = []
    for r in requests_data:
        req = SchedulerRequest(**r)
        requests.append(req)
        sched.add_request(req)
    
    # Run until all complete
    for _ in range(500):
        sched.step()
        has_work = bool(sched.running_batch) or bool(sched.waiting_queue)
        if hasattr(sched, 'preempted_queue'):
            has_work = has_work or bool(sched.preempted_queue)
        if not has_work:
            break
    
    # Collect metrics
    metrics = []
    for req in sched.completed:
        ttft = (req.first_token_time - req.arrival_time) * 1000 if req.first_token_time else float('inf')
        total = (req.completion_time - req.arrival_time) * 1000 if req.completion_time else float('inf')
        metrics.append({
            'id': req.request_id,
            'priority': req.priority,
            'ttft_ms': ttft,
            'total_ms': total,
            'preemptions': req.total_preemptions,
            'slo_met': ttft <= req.slo_ttft_ms,
        })
    return metrics

# Create a mix of high and low priority requests
np.random.seed(42)
test_requests = []
for i in range(12):
    pri = 5 if i % 3 == 0 else 1  # Every 3rd request is high priority
    test_requests.append({
        'request_id': f"R{i:02d}",
        'arrival_time': i * 0.02,
        'prompt_tokens': random.randint(32, 256),
        'max_gen_tokens': random.randint(10, 40),
        'priority': pri,
        'slo_ttft_ms': 500 if pri == 5 else 2000,
    })

common_kwargs = {'max_batch_size': 4, 'total_gpu_blocks': 100, 'block_size': 16}

fcfs_metrics = run_simulation(FCFSScheduler, test_requests, **common_kwargs)
priority_metrics = run_simulation(PriorityScheduler, test_requests, **common_kwargs)
preemptive_metrics = run_simulation(PreemptiveScheduler, test_requests, 
                                    preemption_mode='swap', **common_kwargs)

# Visualize comparison
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

schedulers = ['FCFS', 'Priority', 'Preemptive']
all_metrics = [fcfs_metrics, priority_metrics, preemptive_metrics]

# Plot 1: Average TTFT by priority
ax = axes[0]
for sched_name, metrics in zip(schedulers, all_metrics):
    high_ttft = np.mean([m['ttft_ms'] for m in metrics if m['priority'] == 5])
    low_ttft = np.mean([m['ttft_ms'] for m in metrics if m['priority'] == 1])
    x = schedulers.index(sched_name)
    ax.bar(x - 0.15, high_ttft, 0.3, color='#e74c3c', alpha=0.8, 
           label='High Priority' if x == 0 else '')
    ax.bar(x + 0.15, low_ttft, 0.3, color='#3498db', alpha=0.8,
           label='Low Priority' if x == 0 else '')

ax.set_xticks(range(len(schedulers)))
ax.set_xticklabels(schedulers)
ax.set_ylabel('Average TTFT (ms)')
ax.set_title('TTFT by Priority & Scheduler', fontweight='bold')
ax.legend()
ax.grid(axis='y', alpha=0.3)

# Plot 2: SLO compliance rate
ax = axes[1]
for i, (sched_name, metrics) in enumerate(zip(schedulers, all_metrics)):
    slo_rate = sum(1 for m in metrics if m['slo_met']) / len(metrics) * 100
    high_slo = sum(1 for m in metrics if m['slo_met'] and m['priority'] == 5) / \
               max(1, sum(1 for m in metrics if m['priority'] == 5)) * 100
    ax.bar(i, slo_rate, color='#2ecc71', alpha=0.8)
    ax.text(i, slo_rate + 1, f'{slo_rate:.0f}%', ha='center', fontweight='bold')

ax.set_xticks(range(len(schedulers)))
ax.set_xticklabels(schedulers)
ax.set_ylabel('SLO Compliance Rate (%)')
ax.set_title('SLO Compliance by Scheduler', fontweight='bold')
ax.set_ylim(0, 110)
ax.grid(axis='y', alpha=0.3)

# Plot 3: Total latency distribution
ax = axes[2]
for i, (sched_name, metrics) in enumerate(zip(schedulers, all_metrics)):
    totals = [m['total_ms'] for m in metrics]
    bp = ax.boxplot(totals, positions=[i], widths=0.5)

ax.set_xticks(range(len(schedulers)))
ax.set_xticklabels(schedulers)
ax.set_ylabel('Total Latency (ms)')
ax.set_title('Total Latency Distribution', fontweight='bold')
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()
````

## 8. Scheduling Timeline Visualization

Let's create a detailed Gantt-chart style visualization showing exactly what happens at each iteration.

````python
def visualize_schedule_timeline(scheduler_class, requests_data, title, **kwargs):
    """Visualize scheduling decisions as a timeline."""
    sched = scheduler_class(**kwargs)
    
    for r in requests_data:
        sched.add_request(SchedulerRequest(**r))
    
    # Track state over time for each request
    request_ids = [r['request_id'] for r in requests_data]
    timelines = {rid: [] for rid in request_ids}  # (start_time, end_time, state)
    
    prev_states = {rid: None for rid in request_ids}
    
    for _ in range(200):
        t = sched.current_time
        sched.step()
        
        # Record states
        for rid in request_ids:
            req = None
            for r in sched.running_batch + sched.waiting_queue + sched.completed:
                if r.request_id == rid:
                    req = r
                    break
            if hasattr(sched, 'preempted_queue'):
                for r in sched.preempted_queue:
                    if r.request_id == rid:
                        req = r
                        break
            
            if req:
                state = req.state.value
                if state != prev_states[rid]:
                    if prev_states[rid] is not None:
                        timelines[rid][-1] = (timelines[rid][-1][0], t, prev_states[rid])
                    timelines[rid].append((t, t + 0.025, state))
                    prev_states[rid] = state
                else:
                    if timelines[rid]:
                        timelines[rid][-1] = (timelines[rid][-1][0], t + 0.025, state)
        
        has_work = bool(sched.running_batch) or bool(sched.waiting_queue)
        if hasattr(sched, 'preempted_queue'):
            has_work = has_work or bool(sched.preempted_queue)
        if not has_work:
            break
    
    # Plot
    fig, ax = plt.subplots(figsize=(16, max(4, len(request_ids) * 0.8)))
    
    state_colors = {
        'waiting': '#95a5a6',
        'running': '#2ecc71',
        'preempted': '#e74c3c',
        'completed': '#3498db',
    }
    
    for i, rid in enumerate(reversed(request_ids)):
        for start, end, state in timelines[rid]:
            ax.barh(i, end - start, left=start, height=0.6, 
                    color=state_colors.get(state, 'gray'), alpha=0.8, 
                    edgecolor='black', linewidth=0.5)
    
    ax.set_yticks(range(len(request_ids)))
    ax.set_yticklabels(list(reversed(request_ids)))
    ax.set_xlabel('Time (seconds)')
    ax.set_title(title, fontsize=13, fontweight='bold')
    
    patches = [mpatches.Patch(color=c, label=s.capitalize()) for s, c in state_colors.items()]
    ax.legend(handles=patches, loc='upper right')
    ax.grid(axis='x', alpha=0.3)
    
    plt.tight_layout()
    plt.show()

# Visualize all three schedulers
simple_requests = [
    {'request_id': 'A', 'arrival_time': 0.0, 'prompt_tokens': 64, 'max_gen_tokens': 25, 'priority': 1},
    {'request_id': 'B', 'arrival_time': 0.01, 'prompt_tokens': 128, 'max_gen_tokens': 15, 'priority': 3},
    {'request_id': 'C', 'arrival_time': 0.02, 'prompt_tokens': 32, 'max_gen_tokens': 30, 'priority': 1},
    {'request_id': 'D', 'arrival_time': 0.1, 'prompt_tokens': 64, 'max_gen_tokens': 20, 'priority': 5},
    {'request_id': 'E', 'arrival_time': 0.2, 'prompt_tokens': 96, 'max_gen_tokens': 10, 'priority': 2},
]

small_kwargs = {'max_batch_size': 3, 'total_gpu_blocks': 60, 'block_size': 16}

visualize_schedule_timeline(FCFSScheduler, simple_requests, 'FCFS Scheduler Timeline', **small_kwargs)
visualize_schedule_timeline(PriorityScheduler, simple_requests, 'Priority Scheduler Timeline', **small_kwargs)
visualize_schedule_timeline(PreemptiveScheduler, simple_requests, 'Preemptive Scheduler Timeline',
                           preemption_mode='swap', **small_kwargs)
````

## 9. How vLLM and SGLang Schedule

### vLLM Scheduler

vLLM uses a **three-queue scheduler**:
1. **Waiting queue**: New requests that haven't been prefilled yet
2. **Running queue**: Requests currently generating tokens
3. **Swapped queue**: Preempted requests with KV-cache on CPU

Each iteration, the scheduler:
1. First tries to swap in preempted requests
2. Then admits new requests from the waiting queue
3. If memory is insufficient, preempts lowest-priority running requests

### SGLang Scheduler

SGLang separates prefill and decode into different scheduling phases:
1. **Prefill phase**: Process a batch of new requests (compute-intensive)
2. **Decode phase**: Continue generating for running requests (memory-bound)
3. Uses **RadixAttention** to share KV-cache across requests with common prefixes

### Key Insight: Chunked Prefill

Both systems support **chunked prefill**: splitting long prompts into chunks and interleaving prefill with decode steps. This prevents a single long prompt from blocking all decode operations.

## 11. Summary

In this chapter, we explored the scheduler -- the brain of an LLM serving system:

1. **FCFS Scheduler**: Simple but suffers from head-of-line blocking. No priority awareness.
2. **Priority Scheduler**: Dynamically prioritizes based on user tier, SLO urgency, and request age. Better SLO compliance.
3. **Preemptive Scheduler**: Can evict running requests to make room for higher-priority ones.
   - **Swap**: Save KV-cache to CPU. Good when KV-cache is small relative to prefill cost.
   - **Recompute**: Discard KV-cache and re-prefill. Good when KV-cache is large and prompt is short.
4. **Admission Control**: Must carefully manage GPU memory to avoid OOM.
5. **Chunked Prefill**: Prevents long prefills from blocking decode operations.
6. **Fairness vs Efficiency**: A fundamental tradeoff; aging and weighted fair queueing help balance.

### Key Design Principles

- The scheduler runs at **iteration granularity** (every ~25ms)
- **Memory is the binding constraint**, not compute
- **Continuous batching** is essential for high throughput
- **SLO awareness** differentiates production from toy schedulers
