## 1. Why Serve Multiple Models?

Production deployments rarely serve a single model. Common scenarios:

| Scenario | Example | Models |
|----------|---------|--------|
| **Quality tiers** | Route simple queries to small model, complex to large | 8B + 70B |
| **Specialized models** | Code generation, translation, summarization | CodeLlama + Llama |
| **A/B testing** | Compare new model version against production | v1.0 + v1.1 |
| **LoRA variants** | Same base model with different fine-tunes | base + N adapters |
| **Cost optimization** | Cheaper model for internal, premium for external | Small + Large |
| **Migration** | Gradually shift traffic from old to new model | Old + New |

### Challenges

- **Memory management**: GPU memory is finite; not all models fit simultaneously
- **Loading time**: Loading a model can take 30-120 seconds
- **Request routing**: Which model should handle each request?
- **Resource isolation**: One model's traffic surge shouldn't starve others

```python
# Model registry

class ModelStatus(Enum):
    UNLOADED = "unloaded"
    LOADING = "loading"
    READY = "ready"
    WARMING_UP = "warming_up"
    COOLING_DOWN = "cooling_down"
    ERROR = "error"

@dataclass
class ModelInfo:
    """Information about a registered model."""
    name: str
    size_gb: float  # GPU memory footprint
    max_context: int = 8192
    capabilities: List[str] = field(default_factory=list)  # e.g., ["chat", "code", "translation"]
    is_lora: bool = False
    base_model: Optional[str] = None  # For LoRA: name of base model
    lora_size_gb: float = 0.0  # LoRA adapter size
    
    # Runtime state
    status: ModelStatus = ModelStatus.UNLOADED
    load_time_s: float = 0.0
    last_used: float = 0.0
    total_requests: int = 0
    active_requests: int = 0
    avg_latency_ms: float = 0.0
    
    @property
    def effective_size_gb(self):
        """Actual GPU memory needed (LoRA only needs adapter memory if base is loaded)."""
        if self.is_lora:
            return self.lora_size_gb
        return self.size_gb

# Register models
model_catalog = {
    'llama-3-8b': ModelInfo('llama-3-8b', 16, 8192, ['chat', 'general']),
    'llama-3-70b': ModelInfo('llama-3-70b', 140, 8192, ['chat', 'general', 'reasoning']),
    'codellama-34b': ModelInfo('codellama-34b', 68, 16384, ['code', 'chat']),
    'llama-3-8b-sql': ModelInfo('llama-3-8b-sql', 16, 8192, ['sql', 'data'], 
                                is_lora=True, base_model='llama-3-8b', lora_size_gb=0.5),
    'llama-3-8b-medical': ModelInfo('llama-3-8b-medical', 16, 8192, ['medical', 'chat'],
                                    is_lora=True, base_model='llama-3-8b', lora_size_gb=0.5),
}

print("=== Model Catalog ===")
print(f"{'Model':<25} {'Size (GB)':<12} {'Context':<10} {'Capabilities':<30} {'LoRA?':<8}")
print("-" * 85)
for name, info in model_catalog.items():
    caps = ', '.join(info.capabilities)
    lora = f"Yes (base: {info.base_model})" if info.is_lora else "No"
    print(f"{name:<25} {info.size_gb:<12.1f} {info.max_context:<10} {caps:<30} {lora:<8}")
```

## 2. Model Routing Strategies

The router decides which model handles each request. Several strategies exist.

```python
class RoutingStrategy(Enum):
    DIRECT = "direct"             # Client specifies model explicitly
    ROUND_ROBIN = "round_robin"   # Rotate between models
    LOAD_BASED = "load_based"     # Route to least loaded model
    CAPABILITY = "capability"     # Route based on task type
    COST_BASED = "cost_based"     # Route based on cost optimization
    AB_TEST = "ab_test"           # A/B testing with traffic splitting


class ModelRouter:
    """Routes requests to appropriate models."""
    
    def __init__(self, models: Dict[str, ModelInfo]):
        self.models = models
        self.rr_index = 0  # For round-robin
        self.ab_configs: Dict[str, Dict[str, float]] = {}  # experiment -> {model: weight}
        self.routing_log: List[dict] = []
    
    def route_direct(self, model_name: str) -> Optional[str]:
        """Direct routing: use the specified model."""
        if model_name in self.models:
            return model_name
        return None
    
    def route_round_robin(self, eligible_models: List[str] = None) -> str:
        """Round-robin across eligible models."""
        if eligible_models is None:
            eligible_models = [m for m, info in self.models.items() 
                              if info.status == ModelStatus.READY]
        if not eligible_models:
            return None
        model = eligible_models[self.rr_index % len(eligible_models)]
        self.rr_index += 1
        return model
    
    def route_load_based(self) -> Optional[str]:
        """Route to the model with the least active requests."""
        ready_models = [(name, info) for name, info in self.models.items() 
                        if info.status == ModelStatus.READY]
        if not ready_models:
            return None
        return min(ready_models, key=lambda x: x[1].active_requests)[0]
    
    def route_capability(self, task_type: str) -> Optional[str]:
        """Route based on required capability."""
        matching = [(name, info) for name, info in self.models.items()
                    if task_type in info.capabilities and info.status == ModelStatus.READY]
        if not matching:
            # Fallback to general-purpose model
            matching = [(name, info) for name, info in self.models.items()
                        if 'general' in info.capabilities and info.status == ModelStatus.READY]
        if not matching:
            return None
        # Prefer smallest capable model (cost optimization)
        return min(matching, key=lambda x: x[1].size_gb)[0]
    
    def route_ab_test(self, experiment: str) -> Optional[str]:
        """A/B test routing with weighted random selection."""
        if experiment not in self.ab_configs:
            return None
        
        config = self.ab_configs[experiment]
        models = list(config.keys())
        weights = list(config.values())
        
        return random.choices(models, weights=weights, k=1)[0]
    
    def setup_ab_test(self, experiment: str, config: Dict[str, float]):
        """Configure A/B test traffic splitting."""
        assert abs(sum(config.values()) - 1.0) < 0.01, "Weights must sum to 1.0"
        self.ab_configs[experiment] = config
    
    def route(self, request: dict, strategy: RoutingStrategy = RoutingStrategy.DIRECT) -> Optional[str]:
        """Main routing method."""
        model = request.get('model')
        
        if strategy == RoutingStrategy.DIRECT:
            result = self.route_direct(model)
        elif strategy == RoutingStrategy.ROUND_ROBIN:
            result = self.route_round_robin()
        elif strategy == RoutingStrategy.LOAD_BASED:
            result = self.route_load_based()
        elif strategy == RoutingStrategy.CAPABILITY:
            task = request.get('task_type', 'general')
            result = self.route_capability(task)
        elif strategy == RoutingStrategy.AB_TEST:
            experiment = request.get('experiment', 'default')
            result = self.route_ab_test(experiment)
        else:
            result = self.route_direct(model)
        
        self.routing_log.append({
            'strategy': strategy.value,
            'requested': model,
            'routed_to': result,
            'time': time.time(),
        })
        
        return result

# Demo routing strategies
# Set some models as ready
for name in ['llama-3-8b', 'llama-3-70b', 'codellama-34b']:
    model_catalog[name].status = ModelStatus.READY
model_catalog['llama-3-8b'].active_requests = 10
model_catalog['llama-3-70b'].active_requests = 3
model_catalog['codellama-34b'].active_requests = 7

router = ModelRouter(model_catalog)

print("=== Routing Strategy Demos ===")

# Direct
print(f"\n1. Direct: 'llama-3-8b' -> {router.route({'model': 'llama-3-8b'}, RoutingStrategy.DIRECT)}")

# Round-robin
print(f"\n2. Round-Robin:")
for i in range(6):
    result = router.route({}, RoutingStrategy.ROUND_ROBIN)
    print(f"   Request {i+1} -> {result}")

# Load-based
print(f"\n3. Load-Based: -> {router.route({}, RoutingStrategy.LOAD_BASED)}")
print(f"   (Chose llama-3-70b with only 3 active requests)")

# Capability-based
print(f"\n4. Capability-Based:")
for task in ['code', 'reasoning', 'general', 'translation']:
    result = router.route({'task_type': task}, RoutingStrategy.CAPABILITY)
    print(f"   Task '{task}' -> {result}")

# A/B Test
router.setup_ab_test('model_upgrade', {'llama-3-8b': 0.7, 'llama-3-70b': 0.3})
print(f"\n5. A/B Test (70/30 split):")
ab_results = {}
for _ in range(100):
    result = router.route({'experiment': 'model_upgrade'}, RoutingStrategy.AB_TEST)
    ab_results[result] = ab_results.get(result, 0) + 1
for model, count in ab_results.items():
    print(f"   {model}: {count}% of traffic")
```

## 3. LoRA Adapter Serving: Weight Sharing

LoRA (Low-Rank Adaptation) allows serving multiple model variants from a **single base model** by hot-swapping small adapter weights.

### Memory Savings

Instead of loading N full models:
- **Without LoRA**: N * model_size (e.g., 3 * 16GB = 48GB)
- **With LoRA**: 1 * model_size + N * adapter_size (e.g., 16GB + 3 * 0.5GB = 17.5GB)

### How LoRA Serving Works

For each linear layer with LoRA:
$$Y = XW + X \cdot B \cdot A \cdot \alpha$$

Where $B$ (down-projection) and $A$ (up-projection) are the small LoRA matrices, and $\alpha$ is the scaling factor.

```python
@dataclass
class LoRAAdapter:
    """Represents a LoRA adapter that can be hot-swapped."""
    name: str
    base_model: str
    rank: int = 16  # LoRA rank
    alpha: float = 32.0
    target_modules: List[str] = field(default_factory=lambda: ['q_proj', 'v_proj'])
    size_mb: float = 0.0
    loaded: bool = False
    
    def __post_init__(self):
        # Estimate size: 2 * rank * hidden_dim * num_target_layers * dtype
        # Simplified estimate
        self.size_mb = self.rank * 4096 * 2 * len(self.target_modules) * 32 * 2 / 1024 / 1024

class LoRAModelManager:
    """Manages base models and their LoRA adapters."""
    
    def __init__(self, gpu_memory_gb: float = 80):
        self.gpu_memory_gb = gpu_memory_gb
        self.used_memory_gb = 0
        self.base_models: Dict[str, ModelInfo] = {}
        self.adapters: Dict[str, LoRAAdapter] = {}
        self.loaded_base_models: set = set()
        self.loaded_adapters: set = set()
    
    def load_base_model(self, model: ModelInfo) -> bool:
        """Load a base model into GPU memory."""
        if model.name in self.loaded_base_models:
            return True
        if self.used_memory_gb + model.size_gb > self.gpu_memory_gb:
            return False
        
        self.base_models[model.name] = model
        self.loaded_base_models.add(model.name)
        self.used_memory_gb += model.size_gb
        model.status = ModelStatus.READY
        print(f"  Loaded base model '{model.name}' ({model.size_gb}GB). "
              f"GPU: {self.used_memory_gb:.1f}/{self.gpu_memory_gb}GB")
        return True
    
    def load_adapter(self, adapter: LoRAAdapter) -> bool:
        """Load a LoRA adapter (requires base model to be loaded)."""
        if adapter.base_model not in self.loaded_base_models:
            print(f"  Error: Base model '{adapter.base_model}' not loaded")
            return False
        
        adapter_gb = adapter.size_mb / 1024
        if self.used_memory_gb + adapter_gb > self.gpu_memory_gb:
            return False
        
        self.adapters[adapter.name] = adapter
        self.loaded_adapters.add(adapter.name)
        self.used_memory_gb += adapter_gb
        adapter.loaded = True
        print(f"  Loaded adapter '{adapter.name}' ({adapter.size_mb:.0f}MB). "
              f"GPU: {self.used_memory_gb:.1f}/{self.gpu_memory_gb}GB")
        return True
    
    def unload_adapter(self, name: str):
        if name in self.adapters:
            adapter = self.adapters[name]
            self.used_memory_gb -= adapter.size_mb / 1024
            self.loaded_adapters.discard(name)
            adapter.loaded = False
            print(f"  Unloaded adapter '{name}'")
    
    def get_available_models(self) -> List[str]:
        """List all serveable models (base + adapters)."""
        models = list(self.loaded_base_models)
        models.extend(self.loaded_adapters)
        return models
    
    def memory_report(self) -> dict:
        return {
            'total_gb': self.gpu_memory_gb,
            'used_gb': self.used_memory_gb,
            'free_gb': self.gpu_memory_gb - self.used_memory_gb,
            'base_models': list(self.loaded_base_models),
            'adapters': list(self.loaded_adapters),
            'num_serveable': len(self.get_available_models()),
        }

# Demo: LoRA serving
print("=== LoRA Model Manager Demo ===")
lora_mgr = LoRAModelManager(gpu_memory_gb=80)

# Load base model
base = ModelInfo('llama-3-8b', 16, 8192, ['chat', 'general'])
lora_mgr.load_base_model(base)

# Load multiple LoRA adapters on same base
adapters = [
    LoRAAdapter('llama-3-8b-sql', 'llama-3-8b', rank=16, target_modules=['q_proj', 'v_proj']),
    LoRAAdapter('llama-3-8b-medical', 'llama-3-8b', rank=32, target_modules=['q_proj', 'k_proj', 'v_proj', 'o_proj']),
    LoRAAdapter('llama-3-8b-finance', 'llama-3-8b', rank=8, target_modules=['q_proj', 'v_proj']),
    LoRAAdapter('llama-3-8b-legal', 'llama-3-8b', rank=16, target_modules=['q_proj', 'v_proj', 'up_proj', 'down_proj']),
]

for adapter in adapters:
    lora_mgr.load_adapter(adapter)

print(f"\n=== Memory Report ===")
report = lora_mgr.memory_report()
print(json.dumps(report, indent=2))

# Visualize memory
fig, ax = plt.subplots(figsize=(12, 4))
# Base model
ax.barh(0, base.size_gb, left=0, height=0.6, color='#3498db', alpha=0.8, edgecolor='black')
ax.text(base.size_gb/2, 0, f'Base: {base.name}\n{base.size_gb}GB', 
        ha='center', va='center', fontweight='bold', color='white')

# Adapters
offset = base.size_gb
adapter_colors = ['#e74c3c', '#2ecc71', '#f39c12', '#9b59b6']
for i, adapter in enumerate(adapters):
    size = adapter.size_mb / 1024
    ax.barh(0, size, left=offset, height=0.6, color=adapter_colors[i], alpha=0.8, edgecolor='black')
    offset += size

# Free space
free = report['free_gb']
ax.barh(0, free, left=offset, height=0.6, color='lightgray', alpha=0.3, edgecolor='gray')
ax.text(offset + free/2, 0, f'Free: {free:.1f}GB', ha='center', va='center')

ax.set_xlim(0, 80)
ax.set_yticks([])
ax.set_xlabel('GPU Memory (GB)')
ax.set_title('GPU Memory: 1 Base + 4 LoRA Adapters vs 5 Full Models', fontweight='bold')

# Add comparison bar
ax.barh(-1, 5 * 16, left=0, height=0.6, color='#e74c3c', alpha=0.3, edgecolor='black', hatch='//')
ax.text(40, -1, f'Without LoRA: 5 full models = {5*16}GB (DOES NOT FIT!)', 
        ha='center', va='center', fontweight='bold', color='red')

ax.set_ylim(-1.8, 0.8)
plt.tight_layout()
plt.show()
```

## 4. Dynamic Model Loading/Unloading

In production, models must be loaded and unloaded dynamically based on demand. This requires:

1. **Memory budget tracking**: Know how much GPU memory is available
2. **LRU eviction**: Unload least recently used models when memory is needed
3. **Graceful unloading**: Wait for active requests to complete before unloading
4. **Preloading**: Anticipate which models will be needed

```python
class DynamicModelManager:
    """Manages dynamic loading/unloading of models with LRU eviction."""
    
    def __init__(self, gpu_memory_gb: float = 80, reserved_gb: float = 5):
        self.total_memory = gpu_memory_gb
        self.reserved_memory = reserved_gb  # For KV-cache and overhead
        self.available_memory = gpu_memory_gb - reserved_gb
        self.used_memory = 0.0
        
        self.model_registry: Dict[str, ModelInfo] = {}
        self.loaded_models: OrderedDict = OrderedDict()  # name -> ModelInfo (LRU order)
        self.load_history: List[dict] = []
    
    def register_model(self, model: ModelInfo):
        """Register a model (doesn't load it)."""
        self.model_registry[model.name] = model
    
    def _simulate_load(self, model: ModelInfo):
        """Simulate model loading time."""
        load_time = model.size_gb * 2  # ~2 seconds per GB (simulated)
        model.load_time_s = load_time
        return load_time
    
    def load_model(self, name: str) -> dict:
        """Load a model, evicting LRU models if necessary."""
        if name in self.loaded_models:
            # Move to end (most recently used)
            self.loaded_models.move_to_end(name)
            model = self.loaded_models[name]
            model.last_used = time.time()
            return {'status': 'already_loaded', 'model': name}
        
        if name not in self.model_registry:
            return {'status': 'error', 'message': f'Model {name} not registered'}
        
        model = self.model_registry[name]
        needed = model.effective_size_gb
        
        # Evict models if needed
        evicted = []
        while self.used_memory + needed > self.available_memory:
            if not self.loaded_models:
                return {'status': 'error', 'message': 'Cannot fit model even after evictions'}
            
            # Evict LRU (first item in OrderedDict)
            lru_name, lru_model = next(iter(self.loaded_models.items()))
            
            if lru_model.active_requests > 0:
                # Can't evict model with active requests - try next
                # In practice, we'd wait or try another model
                break
            
            self.unload_model(lru_name)
            evicted.append(lru_name)
        
        if self.used_memory + needed > self.available_memory:
            return {'status': 'error', 'message': 'Insufficient memory after evictions'}
        
        # Load the model
        load_time = self._simulate_load(model)
        model.status = ModelStatus.READY
        model.last_used = time.time()
        self.loaded_models[name] = model
        self.used_memory += needed
        
        result = {
            'status': 'loaded',
            'model': name,
            'load_time_s': load_time,
            'evicted': evicted,
            'memory_used': self.used_memory,
            'memory_available': self.available_memory - self.used_memory,
        }
        self.load_history.append(result)
        return result
    
    def unload_model(self, name: str):
        """Unload a model from GPU memory."""
        if name in self.loaded_models:
            model = self.loaded_models.pop(name)
            self.used_memory -= model.effective_size_gb
            model.status = ModelStatus.UNLOADED
            print(f"  Evicted '{name}' ({model.effective_size_gb}GB). "
                  f"Memory: {self.used_memory:.1f}/{self.available_memory:.1f}GB")
    
    def request_model(self, name: str) -> dict:
        """Request a model for serving. Loads if necessary."""
        if name in self.loaded_models:
            self.loaded_models.move_to_end(name)
            self.loaded_models[name].last_used = time.time()
            return {'status': 'ready', 'cold_start': False}
        else:
            result = self.load_model(name)
            if result['status'] == 'loaded':
                return {'status': 'ready', 'cold_start': True, 
                        'load_time_s': result['load_time_s'],
                        'evicted': result.get('evicted', [])}
            return result

# Demo: Dynamic loading with LRU eviction
print("=== Dynamic Model Manager Demo ===")
dmm = DynamicModelManager(gpu_memory_gb=80, reserved_gb=10)

# Register models
for name, info in model_catalog.items():
    info.status = ModelStatus.UNLOADED
    dmm.register_model(info)

# Simulate requests for different models
request_sequence = [
    'llama-3-8b',      # Load (16GB)
    'codellama-34b',   # Load (68GB) -- tight fit!
    'llama-3-8b',      # Already loaded
    'llama-3-70b',     # Need 140GB -- must evict!
    'llama-3-8b',      # Was evicted, reload
]

for model_name in request_sequence:
    print(f"\nRequesting '{model_name}'...")
    result = dmm.request_model(model_name)
    print(f"  Result: {result}")
    print(f"  Loaded models: {list(dmm.loaded_models.keys())}")
    print(f"  Memory: {dmm.used_memory:.1f}/{dmm.available_memory:.1f}GB")
```

## 5. Model Warm-Up and Cool-Down

### Warm-Up
After loading, a model's first few inferences may be slower due to:
- CUDA kernel compilation (torch.compile, Triton)
- Memory allocation patterns not yet optimized
- CPU cache cold

**Strategy**: Run a few dummy requests before marking the model as ready.

### Cool-Down
Before unloading, we should:
- Stop accepting new requests
- Wait for active requests to complete
- Persist any cached state (e.g., prefix cache)
- Then free GPU memory

```python
class ModelLifecycleManager:
    """Manages the full lifecycle: load -> warm-up -> serve -> cool-down -> unload."""
    
    def __init__(self):
        self.models: Dict[str, ModelInfo] = {}
        self.lifecycle_log: List[dict] = []
    
    def lifecycle_load(self, model: ModelInfo) -> List[dict]:
        """Full lifecycle: load with warm-up."""
        events = []
        
        # Phase 1: Loading weights
        model.status = ModelStatus.LOADING
        load_time = model.size_gb * 2  # Simulated
        events.append({'phase': 'loading', 'duration_s': load_time, 
                       'status': ModelStatus.LOADING.value})
        
        # Phase 2: Warm-up (run dummy requests)
        model.status = ModelStatus.WARMING_UP
        warmup_requests = 3
        warmup_times = []
        for i in range(warmup_requests):
            # First request is slowest (kernel compilation)
            warmup_time = 2.0 if i == 0 else 0.5
            warmup_times.append(warmup_time)
        
        events.append({'phase': 'warmup', 'requests': warmup_requests,
                       'times_s': warmup_times, 'status': ModelStatus.WARMING_UP.value})
        
        # Phase 3: Ready
        model.status = ModelStatus.READY
        events.append({'phase': 'ready', 'status': ModelStatus.READY.value})
        
        self.models[model.name] = model
        self.lifecycle_log.extend(events)
        return events
    
    def lifecycle_unload(self, name: str) -> List[dict]:
        """Full lifecycle: cool-down then unload."""
        events = []
        model = self.models.get(name)
        if not model:
            return [{'phase': 'error', 'message': 'Model not found'}]
        
        # Phase 1: Stop accepting requests
        model.status = ModelStatus.COOLING_DOWN
        events.append({'phase': 'cooling_down', 'status': ModelStatus.COOLING_DOWN.value,
                       'active_requests': model.active_requests})
        
        # Phase 2: Drain active requests (simulated)
        drain_time = model.active_requests * 0.5  # Average 0.5s per request
        events.append({'phase': 'draining', 'duration_s': drain_time})
        
        # Phase 3: Unload
        model.status = ModelStatus.UNLOADED
        events.append({'phase': 'unloaded', 'status': ModelStatus.UNLOADED.value})
        
        del self.models[name]
        self.lifecycle_log.extend(events)
        return events

# Demo lifecycle
print("=== Model Lifecycle Demo ===")
lcm = ModelLifecycleManager()

model = ModelInfo('llama-3-8b', 16, 8192, ['chat'])
print("\n--- Loading ---")
for event in lcm.lifecycle_load(model):
    print(f"  {event}")

model.active_requests = 5
print("\n--- Unloading (with 5 active requests) ---")
for event in lcm.lifecycle_unload('llama-3-8b'):
    print(f"  {event}")
```

## 6. A/B Testing with Traffic Splitting

A/B testing allows comparing model versions in production with real traffic.

```python
class ABTestManager:
    """Manages A/B testing experiments."""
    
    def __init__(self):
        self.experiments: Dict[str, dict] = {}
        self.results: Dict[str, Dict[str, List[dict]]] = {}  # experiment -> model -> [metrics]
    
    def create_experiment(self, name: str, variants: Dict[str, float],
                          metric: str = 'latency_ms'):
        """Create A/B test.
        variants: {'model_a': 0.5, 'model_b': 0.5}
        """
        self.experiments[name] = {
            'variants': variants,
            'metric': metric,
            'created': time.time(),
            'active': True,
        }
        self.results[name] = {model: [] for model in variants}
    
    def get_variant(self, experiment: str, user_id: str = None) -> str:
        """Get variant for a request. Uses consistent hashing for user stickiness."""
        if experiment not in self.experiments:
            return None
        
        exp = self.experiments[experiment]
        variants = list(exp['variants'].keys())
        weights = list(exp['variants'].values())
        
        if user_id:
            # Consistent assignment: same user always gets same variant
            hash_val = hash(user_id + experiment) % 1000 / 1000
            cumulative = 0
            for variant, weight in zip(variants, weights):
                cumulative += weight
                if hash_val < cumulative:
                    return variant
        
        return random.choices(variants, weights=weights, k=1)[0]
    
    def record_result(self, experiment: str, model: str, metrics: dict):
        """Record result for a variant."""
        if experiment in self.results and model in self.results[experiment]:
            self.results[experiment][model].append(metrics)
    
    def get_experiment_results(self, experiment: str) -> dict:
        """Get aggregated results for an experiment."""
        if experiment not in self.results:
            return None
        
        summary = {}
        for model, metrics_list in self.results[experiment].items():
            if not metrics_list:
                continue
            latencies = [m.get('latency_ms', 0) for m in metrics_list]
            qualities = [m.get('quality_score', 0) for m in metrics_list]
            
            summary[model] = {
                'count': len(metrics_list),
                'avg_latency_ms': np.mean(latencies),
                'p50_latency_ms': np.percentile(latencies, 50),
                'p99_latency_ms': np.percentile(latencies, 99),
                'avg_quality': np.mean(qualities) if any(qualities) else None,
            }
        return summary

# Demo: A/B testing
print("=== A/B Testing Demo ===")
ab = ABTestManager()

# Create experiment: compare 8B vs 70B
ab.create_experiment('model_size_test', {
    'llama-3-8b': 0.5,
    'llama-3-70b': 0.5,
})

# Simulate 200 requests
np.random.seed(42)
for i in range(200):
    user = f"user_{i % 20}"  # 20 unique users
    variant = ab.get_variant('model_size_test', user)
    
    # Simulate metrics (70B is slower but higher quality)
    if variant == 'llama-3-8b':
        latency = np.random.normal(50, 15)
        quality = np.random.normal(7.0, 1.0)
    else:
        latency = np.random.normal(150, 40)
        quality = np.random.normal(8.5, 0.8)
    
    ab.record_result('model_size_test', variant, {
        'latency_ms': max(10, latency),
        'quality_score': min(10, max(1, quality)),
    })

results = ab.get_experiment_results('model_size_test')
print("\nExperiment: model_size_test")
for model, stats in results.items():
    print(f"\n  {model}:")
    for k, v in stats.items():
        if v is not None:
            print(f"    {k}: {v:.1f}")

# Visualize
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

for model, color in [('llama-3-8b', '#3498db'), ('llama-3-70b', '#e74c3c')]:
    latencies = [m['latency_ms'] for m in ab.results['model_size_test'][model]]
    qualities = [m['quality_score'] for m in ab.results['model_size_test'][model]]
    
    axes[0].hist(latencies, bins=20, alpha=0.6, color=color, label=model, edgecolor='black')
    axes[1].hist(qualities, bins=20, alpha=0.6, color=color, label=model, edgecolor='black')

axes[0].set_xlabel('Latency (ms)')
axes[0].set_title('Latency Distribution', fontweight='bold')
axes[0].legend()

axes[1].set_xlabel('Quality Score')
axes[1].set_title('Quality Score Distribution', fontweight='bold')
axes[1].legend()

plt.suptitle('A/B Test: Llama-3-8B vs 70B', fontsize=14, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()
```

## 7. Complete Multi-Model Router

Let's build a complete multi-model serving system combining all components.

```python
class MultiModelServer:
    """Complete multi-model serving system."""
    
    def __init__(self, gpu_memory_gb: float = 80):
        self.model_manager = DynamicModelManager(gpu_memory_gb)
        self.router = None  # Initialized after models registered
        self.ab_manager = ABTestManager()
        self.request_count = 0
        self.metrics: List[dict] = []
    
    def register_model(self, model: ModelInfo):
        self.model_manager.register_model(model)
    
    def initialize_router(self):
        self.router = ModelRouter(self.model_manager.model_registry)
    
    def handle_request(self, request: dict) -> dict:
        """Handle a request with automatic model loading and routing."""
        self.request_count += 1
        start_time = time.time()
        
        # 1. Determine target model
        model_name = request.get('model')
        if not model_name:
            # Auto-route based on capability
            task = request.get('task_type', 'general')
            model_name = self.router.route(request, RoutingStrategy.CAPABILITY)
        
        if not model_name:
            return {'error': 'No suitable model found'}
        
        # 2. Ensure model is loaded
        load_result = self.model_manager.request_model(model_name)
        if load_result['status'] != 'ready':
            return {'error': f'Failed to load model: {load_result}'}
        
        # 3. Simulate inference
        model = self.model_manager.loaded_models.get(model_name)
        if model:
            model.active_requests += 1
            model.total_requests += 1
        
        # Simulated latency
        latency = np.random.exponential(50) + 10
        if load_result.get('cold_start'):
            latency += load_result.get('load_time_s', 0) * 1000
        
        if model:
            model.active_requests -= 1
        
        result = {
            'model': model_name,
            'latency_ms': latency,
            'cold_start': load_result.get('cold_start', False),
            'evicted': load_result.get('evicted', []),
        }
        self.metrics.append(result)
        return result

# Demo
print("=== Multi-Model Server Demo ===")
mms = MultiModelServer(gpu_memory_gb=80)

for name, info in model_catalog.items():
    info.status = ModelStatus.UNLOADED
    info.active_requests = 0
    mms.register_model(info)

mms.initialize_router()

# Simulate varied workload
workload = [
    {'model': 'llama-3-8b', 'task_type': 'chat'},
    {'model': 'llama-3-8b', 'task_type': 'chat'},
    {'model': 'codellama-34b', 'task_type': 'code'},
    {'task_type': 'general'},  # Auto-route
    {'model': 'llama-3-8b', 'task_type': 'chat'},
    {'task_type': 'code'},  # Auto-route to codellama
]

for i, req in enumerate(workload):
    print(f"\nRequest {i+1}: {req}")
    result = mms.handle_request(req)
    cold = ' (COLD START!)' if result.get('cold_start') else ''
    evicted = f" evicted={result['evicted']}" if result.get('evicted') else ''
    print(f"  -> {result['model']}, latency={result['latency_ms']:.0f}ms{cold}{evicted}")
```

## 9. Summary

In this chapter, we built a complete multi-model serving system:

1. **Model Routing**: Direct, round-robin, load-based, capability-based, and A/B test strategies
2. **LoRA Serving**: Share base model weights, hot-swap small adapters (saves 60-90% memory)

4. **Lifecycle Management**: Warm-up, serve, cool-down, unload with graceful transitions
5. **A/B Testing**: Traffic splitting with consistent user assignment and metrics tracking

### Key Design Principles

- **Lazy loading**: Don't load models until they're needed
- **LRU eviction**: Keep frequently used models warm, evict cold ones
- **Graceful degradation**: Serve from smaller model if preferred model can't load
- **Consistent routing**: Same user should get same model in A/B tests
- **LoRA sharing**: Amortize base model cost across many variants
