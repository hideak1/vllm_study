## 1. The Problem: KV Cache Memory Waste

Before PagedAttention, LLM serving systems allocated **contiguous memory** for each sequence's KV cache. This caused massive memory waste.

```python
# Demonstrating the memory waste problem

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 8))

# Traditional: contiguous allocation with max_seq_len
ax1.set_title('Traditional: Contiguous Allocation (Static)', fontsize=14, fontweight='bold')
ax1.set_xlim(0, 100)
ax1.set_ylim(0, 4)
ax1.set_xlabel('GPU Memory (tokens)')
ax1.set_yticks([0.5, 1.5, 2.5, 3.5])
ax1.set_yticklabels(['Seq 3', 'Seq 2', 'Seq 1', 'Seq 0'])

max_len = 25  # max_seq_len per sequence
actual_lens = [8, 15, 5, 20]  # actual lengths
colors_used = ['#4CAF50', '#2196F3', '#FF9800', '#9C27B0']

for i, (actual, color) in enumerate(zip(actual_lens, colors_used)):
    y = 3 - i
    # Allocated (max_len)
    ax1.barh(y, max_len, left=i*max_len, height=0.8, color='#EEEEEE', edgecolor='black')
    # Used (actual)
    ax1.barh(y, actual, left=i*max_len, height=0.8, color=color, edgecolor='black')
    # Waste label
    waste = max_len - actual
    ax1.text(i*max_len + max_len - waste/2, y, f'waste\n{waste}', 
             ha='center', va='center', fontsize=8, color='red')

total_allocated = max_len * 4
total_used = sum(actual_lens)
total_waste = total_allocated - total_used
ax1.text(50, -0.5, f'Total: {total_allocated} allocated, {total_used} used, '
         f'{total_waste} wasted ({total_waste/total_allocated:.0%} waste)',
         ha='center', fontsize=10, fontweight='bold', color='red')

# PagedAttention: paged allocation
ax2.set_title('PagedAttention: Paged Allocation (Dynamic)', fontsize=14, fontweight='bold')
ax2.set_xlim(0, 100)
ax2.set_ylim(-1, 4)
ax2.set_xlabel('GPU Memory (blocks, 4 tokens each)')
ax2.set_yticks([0.5, 1.5, 2.5, 3.5])
ax2.set_yticklabels(['Seq 3', 'Seq 2', 'Seq 1', 'Seq 0'])

block_size = 4
# Calculate blocks needed for each sequence
block_counts = [(actual + block_size - 1) // block_size for actual in actual_lens]
total_blocks = sum(block_counts)
internal_waste = sum(block_counts) * block_size - sum(actual_lens)

# Draw blocks for each sequence
block_pos = 0
for i, (bc, actual, color) in enumerate(zip(block_counts, actual_lens, colors_used)):
    y = 3 - i
    remaining = actual
    for b in range(bc):
        tokens_in_block = min(remaining, block_size)
        # Full block
        ax2.barh(y, block_size, left=block_pos*block_size, height=0.8, 
                 color=color, edgecolor='black', alpha=0.8)
        # Waste in last block
        if tokens_in_block < block_size:
            waste_start = block_pos * block_size + tokens_in_block
            ax2.barh(y, block_size - tokens_in_block, left=waste_start, height=0.8,
                     color='#EEEEEE', edgecolor='black', alpha=0.5)
        remaining -= tokens_in_block
        block_pos += 1

# Free blocks
free_blocks = 25 - total_blocks  # Same total memory
for fb in range(free_blocks):
    ax2.barh(0.5, block_size, left=(block_pos + fb)*block_size, height=3.8,
             color='white', edgecolor='gray', linestyle='--', alpha=0.3)

ax2.text(50, -0.5, f'Total: {total_blocks} blocks used ({total_blocks*block_size} tokens), '
         f'{internal_waste} internal waste ({internal_waste/(total_blocks*block_size):.0%}), '
         f'{free_blocks} blocks free',
         ha='center', fontsize=10, fontweight='bold', color='green')

plt.tight_layout()
plt.show()
```

## 2. The Virtual Memory Analogy

PagedAttention draws a direct analogy from operating system virtual memory management.

```python
analogy = """
OS Virtual Memory  ←→  PagedAttention
═══════════════════════════════════════════════════════

Process             ←→  Sequence (request)
Virtual address     ←→  Logical block index
Physical address    ←→  Physical block number (GPU memory)
Page                ←→  Block (group of KV cache slots)
Page table          ←→  Block table
Page fault          ←→  Block allocation (new block needed)
Swap space          ←→  CPU RAM for swapped KV cache
Copy-on-Write       ←→  Copy-on-Write for beam search
Page size           ←→  Block size (e.g., 16 tokens)
Internal fragment.  ←→  Waste in last block
External fragment.  ←→  Eliminated! (non-contiguous blocks)

Key Insight:
  Just as OS virtual memory allows processes to use non-contiguous
  physical memory (through page tables), PagedAttention allows
  sequences to use non-contiguous KV cache blocks (through block tables).

  This eliminates EXTERNAL fragmentation entirely.
  Only INTERNAL fragmentation remains (waste in the last block),
  which is bounded by block_size - 1 tokens per sequence.
"""
print(analogy)
```

## 3. Logical Blocks vs Physical Blocks

```python
blocks_explanation = """
Logical Blocks vs Physical Blocks
══════════════════════════════════

LOGICAL BLOCK:
  - Belongs to a specific sequence
  - Has a logical index (0, 1, 2, ...)
  - Represents a contiguous chunk of the sequence's KV cache
  - Logical block 0 = first block_size tokens of the sequence
  - Logical block 1 = next block_size tokens, etc.
  
  Example (block_size=4):
  Sequence: "The quick brown fox jumps over the lazy dog"
  Tokens:   [The] [quick] [brown] [fox] [jumps] [over] [the] [lazy] [dog]
  
  Logical Block 0: [The, quick, brown, fox]     (tokens 0-3)
  Logical Block 1: [jumps, over, the, lazy]      (tokens 4-7)
  Logical Block 2: [dog, _, _, _]                 (tokens 8, rest empty)
                          ↑ internal fragmentation

PHYSICAL BLOCK:
  - A fixed-size chunk of GPU memory
  - Holds block_size KV cache entries
  - Each entry = key tensor + value tensor for all layers and heads
  - Size per block = 2 * num_layers * num_heads * head_dim * block_size * dtype_size
  - Physical blocks are pre-allocated at startup
  - Managed by the BlockAllocator (free list)

BLOCK TABLE:
  - Per-sequence mapping: logical_index → physical_block_number
  - Example: [3, 7, 1] means:
    Logical block 0 → Physical block 3
    Logical block 1 → Physical block 7
    Logical block 2 → Physical block 1
  - This is what the attention kernel uses to find KV cache data
"""
print(blocks_explanation)
```

```python
# Calculate physical block memory size

def calculate_block_memory(num_layers, num_kv_heads, head_dim, block_size, dtype_bytes=2):
    """Calculate memory per physical block.
    
    Each block stores K and V for all layers, all KV heads.
    Shape per block: [2, num_layers, block_size, num_kv_heads, head_dim]
    """
    # 2 for K and V
    elements = 2 * num_layers * block_size * num_kv_heads * head_dim
    bytes_per_block = elements * dtype_bytes
    return bytes_per_block

# Example: LLaMA-7B
models = {
    'LLaMA-7B': {'layers': 32, 'kv_heads': 32, 'head_dim': 128},
    'LLaMA-13B': {'layers': 40, 'kv_heads': 40, 'head_dim': 128},
    'LLaMA-70B (GQA)': {'layers': 80, 'kv_heads': 8, 'head_dim': 128},
    'Mistral-7B (GQA)': {'layers': 32, 'kv_heads': 8, 'head_dim': 128},
}

block_size = 16
print(f"Block size: {block_size} tokens")
print(f"{'Model':<25} {'KV Heads':<10} {'Block Size (KB)':<18} {'Blocks in 10GB':<15}")
print("=" * 70)

for name, config in models.items():
    block_bytes = calculate_block_memory(
        config['layers'], config['kv_heads'], config['head_dim'], block_size
    )
    block_kb = block_bytes / 1024
    blocks_in_10gb = int(10 * 1024 * 1024 * 1024 / block_bytes)  # 10GB
    max_tokens = blocks_in_10gb * block_size
    print(f"{name:<25} {config['kv_heads']:<10} {block_kb:<18.1f} {blocks_in_10gb:<15} ({max_tokens:,} tokens)")

print("\nNote: GQA (Grouped Query Attention) dramatically reduces KV cache size!")
print("LLaMA-70B with 8 KV heads uses 10x less KV cache per token than LLaMA-7B with 32 heads.")
```

## 4. PagedAttention v1 Algorithm

Let's understand how the PagedAttention kernel works.

```python
paged_attn_v1 = """
PagedAttention v1 Algorithm
═══════════════════════════

Standard Attention:
  output = softmax(Q * K^T / sqrt(d)) * V
  where Q is the query (current token), K and V are cached

Challenge with paging:
  K and V are NOT contiguous in memory!
  They are scattered across physical blocks.

PagedAttention v1 Solution:
  Process attention block by block.

  For each query token:
    1. Look up the block table for this sequence
    2. For each block in the block table:
       a. Load K block from physical memory
       b. Compute partial attention: Q * K_block^T / sqrt(d)
       c. Track the running maximum for numerical stability
    3. Compute softmax using the tracked maximum
    4. For each block in the block table:
       a. Load V block from physical memory
       b. Multiply by softmax weights
       c. Accumulate output

Pseudocode:
  # q: [num_heads, head_dim] - query for one token
  # block_table: [num_logical_blocks] - physical block indices
  # kv_cache: [num_physical_blocks, 2, block_size, num_heads, head_dim]

  max_score = -inf
  exp_sum = 0
  output = zeros(num_heads, head_dim)

  for logical_idx, physical_block in enumerate(block_table):
      k_block = kv_cache[physical_block, 0]  # [block_size, num_heads, head_dim]
      v_block = kv_cache[physical_block, 1]  # [block_size, num_heads, head_dim]
      
      # Compute attention scores for this block
      scores = q @ k_block.T / sqrt(head_dim)  # [num_heads, block_size]
      
      # Mask out invalid positions in the last block
      if logical_idx == len(block_table) - 1:
          valid_tokens = seq_len - logical_idx * block_size
          scores[:, valid_tokens:] = -inf
      
      # Online softmax update (numerically stable)
      block_max = max(scores)
      new_max = max(max_score, block_max)
      
      # Rescale previous accumulator
      scale = exp(max_score - new_max)
      output *= scale
      exp_sum *= scale
      
      # Add this block's contribution
      exp_scores = exp(scores - new_max)
      output += exp_scores @ v_block
      exp_sum += sum(exp_scores)
      
      max_score = new_max
  
  # Normalize
  output /= exp_sum
"""
print(paged_attn_v1)
```

## 5. PagedAttention v2 Algorithm

```python
paged_attn_v2 = """
PagedAttention v2: Improved Parallelism
═══════════════════════════════════════

Problem with v1:
  In v1, each query token is processed by one GPU thread block.
  That thread block iterates over ALL KV blocks sequentially.
  For long sequences (many KV blocks), this is slow because
  the work is not parallelized across blocks.

v2 Solution: Two-pass approach
  Pass 1 (Parallel): Each KV block is processed by a SEPARATE thread block
    - Each thread block computes partial attention for its KV block
    - Stores: partial output, partial max score, partial exp sum
    - All KV blocks are processed IN PARALLEL

  Pass 2 (Reduce): Combine all partial results
    - One thread block per query token
    - Reads all partial results from Pass 1
    - Uses the log-sum-exp trick to combine them correctly
    - Produces the final attention output

When to use v1 vs v2:
  v1: Short sequences (few KV blocks)
      - Less overhead (single pass)
      - Default for most cases
  
  v2: Long sequences (many KV blocks)
      - Better GPU utilization through parallelism
      - Extra memory for intermediate results
      - Beneficial when seq_len > ~8192

The selection is automatic: v1 is used when the sequence fits in a single
KV partition (max_num_partitions == 1), otherwise v2 is used with its
partition-based reduction. This is determined by max_context_len / PARTITION_SIZE.
"""
print(paged_attn_v2)
```

## 6. Simplified PagedAttention Implementation

```python
import numpy as np

class SimplifiedPagedAttention:
    """A simplified Python implementation of PagedAttention.
    
    This demonstrates the core concepts without CUDA complexity.
    Real vLLM uses CUDA kernels for this.
    """
    
    def __init__(self, num_blocks, block_size, num_heads, head_dim):
        self.num_blocks = num_blocks
        self.block_size = block_size
        self.num_heads = num_heads
        self.head_dim = head_dim
        
        # Pre-allocate KV cache: [num_blocks, 2(K,V), block_size, num_heads, head_dim]
        self.kv_cache = np.zeros(
            (num_blocks, 2, block_size, num_heads, head_dim),
            dtype=np.float32
        )
        
        # Free block list
        self.free_blocks = list(range(num_blocks))
        
        # Block tables: seq_id → list of physical block indices
        self.block_tables = {}
        
        # Track how many tokens are in each sequence
        self.seq_lens = {}
    
    def allocate_blocks(self, seq_id, num_tokens):
        """Allocate blocks for a new sequence."""
        num_blocks_needed = (num_tokens + self.block_size - 1) // self.block_size
        
        if len(self.free_blocks) < num_blocks_needed:
            raise RuntimeError("Not enough free blocks!")
        
        # Allocate physical blocks (they may be non-contiguous!)
        allocated = []
        for _ in range(num_blocks_needed):
            block = self.free_blocks.pop(0)
            allocated.append(block)
        
        self.block_tables[seq_id] = allocated
        self.seq_lens[seq_id] = num_tokens
        
        print(f"  Allocated {num_blocks_needed} blocks for seq {seq_id}: "
              f"physical blocks = {allocated}")
        return allocated
    
    def write_kv_cache(self, seq_id, position, key, value):
        """Write K,V for a token at a given position."""
        logical_block = position // self.block_size
        block_offset = position % self.block_size
        
        physical_block = self.block_tables[seq_id][logical_block]
        
        self.kv_cache[physical_block, 0, block_offset] = key   # K
        self.kv_cache[physical_block, 1, block_offset] = value # V
    
    def paged_attention(self, seq_id, query):
        """Compute attention using paged KV cache.
        
        query: [num_heads, head_dim] - the query for one token
        Returns: [num_heads, head_dim] - the attention output
        """
        block_table = self.block_tables[seq_id]
        seq_len = self.seq_lens[seq_id]
        scale = 1.0 / np.sqrt(self.head_dim)
        
        # Online softmax variables
        max_score = np.full(self.num_heads, -np.inf)
        exp_sum = np.zeros(self.num_heads)
        output = np.zeros((self.num_heads, self.head_dim))
        
        # Process each block
        for logical_idx, physical_block in enumerate(block_table):
            # How many valid tokens in this block?
            start_pos = logical_idx * self.block_size
            valid_tokens = min(self.block_size, seq_len - start_pos)
            
            if valid_tokens <= 0:
                break
            
            # Load K and V for this block
            k_block = self.kv_cache[physical_block, 0, :valid_tokens]  # [valid, heads, dim]
            v_block = self.kv_cache[physical_block, 1, :valid_tokens]  # [valid, heads, dim]
            
            # Compute attention scores: Q @ K^T for each head
            # query: [heads, dim], k_block: [valid, heads, dim]
            scores = np.zeros((self.num_heads, valid_tokens))
            for h in range(self.num_heads):
                scores[h] = k_block[:, h, :] @ query[h, :] * scale
            
            # Online softmax update
            block_max = np.max(scores, axis=1)  # [heads]
            new_max = np.maximum(max_score, block_max)
            
            # Rescale previous accumulator
            old_scale = np.exp(max_score - new_max)
            output = output * old_scale[:, np.newaxis]
            exp_sum = exp_sum * old_scale
            
            # Add this block's contribution
            exp_scores = np.exp(scores - new_max[:, np.newaxis])
            for h in range(self.num_heads):
                output[h] += exp_scores[h] @ v_block[:, h, :]
            exp_sum += np.sum(exp_scores, axis=1)
            
            max_score = new_max
        
        # Normalize
        output = output / exp_sum[:, np.newaxis]
        return output
    
    def free_blocks_for_seq(self, seq_id):
        """Free all blocks for a sequence."""
        blocks = self.block_tables.pop(seq_id)
        self.free_blocks.extend(blocks)
        del self.seq_lens[seq_id]
        print(f"  Freed {len(blocks)} blocks for seq {seq_id}: {blocks}")


# Demo
print("=" * 60)
print("Simplified PagedAttention Demo")
print("=" * 60)

pa = SimplifiedPagedAttention(
    num_blocks=8, block_size=4, num_heads=2, head_dim=4
)

print(f"\nTotal blocks: {pa.num_blocks}, Block size: {pa.block_size}")
print(f"Free blocks: {pa.free_blocks}")

# Allocate for sequence 0 (6 tokens → 2 blocks)
print("\nAllocating for sequence 0 (6 tokens):")
pa.allocate_blocks(seq_id=0, num_tokens=6)
print(f"Free blocks after: {pa.free_blocks}")

# Allocate for sequence 1 (10 tokens → 3 blocks)
print("\nAllocating for sequence 1 (10 tokens):")
pa.allocate_blocks(seq_id=1, num_tokens=10)
print(f"Free blocks after: {pa.free_blocks}")

# Write some KV cache data for seq 0
print("\nWriting KV cache for sequence 0...")
np.random.seed(42)
for pos in range(6):
    key = np.random.randn(pa.num_heads, pa.head_dim).astype(np.float32)
    value = np.random.randn(pa.num_heads, pa.head_dim).astype(np.float32)
    pa.write_kv_cache(seq_id=0, position=pos, key=key, value=value)

# Run paged attention for seq 0
print("\nRunning PagedAttention for sequence 0...")
query = np.random.randn(pa.num_heads, pa.head_dim).astype(np.float32)
output = pa.paged_attention(seq_id=0, query=query)
print(f"  Query shape: {query.shape}")
print(f"  Output shape: {output.shape}")
print(f"  Output[0][:3]: {output[0][:3]}")

# Free sequence 0
print("\nFreeing sequence 0:")
pa.free_blocks_for_seq(seq_id=0)
print(f"Free blocks after: {pa.free_blocks}")
```

## 7. CUDA Kernel Interface

```python
cuda_interface = '''
CUDA Kernel Interface for PagedAttention
═════════════════════════════════════════

Source: csrc/attention/attention_kernels.cu
Python binding: vllm/attention/ops/paged_attn.py

Python Interface:
─────────────────

class PagedAttention:
    @staticmethod
    def forward_decode(
        output: torch.Tensor,           # [num_tokens, num_heads, head_dim]
        query: torch.Tensor,            # [num_tokens, num_heads, head_dim]
        key_cache: torch.Tensor,        # [num_blocks, block_size, num_kv_heads, head_dim]
        value_cache: torch.Tensor,      # [num_blocks, block_size, num_kv_heads, head_dim]
        head_mapping: torch.Tensor,     # [num_heads] → kv_head (for GQA/MQA)
        scale: float,                   # 1/sqrt(head_dim)
        block_tables: torch.Tensor,     # [num_seqs, max_num_blocks] physical block IDs
        seq_lens: torch.Tensor,         # [num_seqs] sequence lengths
        max_seq_len: int,               # Maximum sequence length in batch
        alibi_slopes: Optional[torch.Tensor],  # ALiBi attention bias slopes
    ) -> None:
        # Calls the CUDA kernel:
        # vllm._C.ops.paged_attention_v1(
        #     output, query, key_cache, value_cache, ...
        # )

CUDA Kernel Signature:
──────────────────────

// V1: Sequential block processing
template <typename scalar_t, int HEAD_SIZE, int BLOCK_SIZE, int NUM_THREADS>
__global__ void paged_attention_v1_kernel(
    scalar_t* __restrict__ out,         // Output tensor
    const scalar_t* __restrict__ q,      // Query tensor
    const scalar_t* __restrict__ k_cache, // Key cache (all blocks)
    const scalar_t* __restrict__ v_cache, // Value cache (all blocks)
    const int* __restrict__ head_mapping, // Query head → KV head
    const float scale,
    const int* __restrict__ block_tables, // Block tables for all seqs
    const int* __restrict__ seq_lens,     // Sequence lengths
    const int max_num_blocks_per_seq,
    ...
) {
    // Grid: one thread block per (sequence, head) pair
    // Block: NUM_THREADS threads (typically 128 or 256)
    
    const int seq_idx = blockIdx.x;   // Which sequence
    const int head_idx = blockIdx.y;  // Which attention head
    const int thread_idx = threadIdx.x;
    
    // Load query for this head
    float q_vec[HEAD_SIZE];
    load_query(q, seq_idx, head_idx, q_vec);
    
    // Iterate over KV blocks
    float max_score = -FLT_MAX;
    float exp_sum = 0.0f;
    float output[HEAD_SIZE] = {0};
    
    const int num_blocks = CEIL_DIV(seq_lens[seq_idx], BLOCK_SIZE);
    const int* block_table = block_tables + seq_idx * max_num_blocks_per_seq;
    
    for (int block_idx = 0; block_idx < num_blocks; block_idx++) {
        int physical_block = block_table[block_idx];
        
        // Load K block and compute Q*K^T
        // (parallelized across threads within the thread block)
        compute_qk(q_vec, k_cache, physical_block, scores, ...);
        
        // Online softmax update
        update_softmax(scores, &max_score, &exp_sum, output, ...);
        
        // Accumulate V contribution
        accumulate_v(v_cache, physical_block, scores, output, ...);
    }
    
    // Normalize and write output
    normalize_and_store(output, exp_sum, out, seq_idx, head_idx);
}
'''
print(cuda_interface)
```

## 8. Copy-on-Write for Beam Search

```python
cow_explanation = """
Copy-on-Write (CoW) in PagedAttention
══════════════════════════════════════

Problem:
  In beam search, multiple beams share the same prompt.
  Their KV cache for the prompt tokens is IDENTICAL.
  Without CoW: each beam gets its own copy → wastes memory.

Solution: Share physical blocks, copy only when modified.

Example: Beam search with beam_width=3, prompt="Hello world"

After prefill:
  Beam 0: block_table = [P5]  ← all three share physical block P5
  Beam 1: block_table = [P5]  ← ref_count(P5) = 3
  Beam 2: block_table = [P5]  ←
  
  Memory used: 1 block (instead of 3!)

After first decode step:
  All beams generate different tokens.
  Now they CANNOT share the last block (it has different tokens).
  
  CoW triggers:
  - Beam 0 writes to P5 (keeps original, ref_count → 2)
  - Wait, actually: the system copies P5 → P8 for Beam 0
  - Beam 0: block_table = [P5, P8]   (new block for new token)
  - Beam 1: block_table = [P5, P9]   (new block for its new token)
  - Beam 2: block_table = [P5, P10]  (new block for its new token)
  
  The prompt block P5 is STILL shared (ref_count = 3)!
  Only the new decode blocks are unique per beam.

Copy-on-Write Logic:
  when append_slots(seq) is called:
    if the current last block has ref_count > 1:
      # Another sequence shares this block!
      # We can't modify it in place.
      1. Allocate a new physical block
      2. Copy the data from old block to new block
      3. Decrease ref_count of old block
      4. Update block table to point to new block
      5. Record the copy in blocks_to_copy
    else:
      # We own this block exclusively, write directly
"""
print(cow_explanation)
```

## 9. Memory Fragmentation Analysis

```python
fragmentation = """
Memory Fragmentation in KV Cache Management
═════════════════════════════════════════════

EXTERNAL FRAGMENTATION:
  Definition: Free memory exists but is in non-contiguous chunks,
  so a large contiguous allocation cannot be satisfied.
  
  Traditional (contiguous allocation):
    [Seq0][Free][Seq1][Free][Seq2][Free][Free]
    Total free: 4 blocks, but largest contiguous: 2 blocks
    A new sequence needing 3 blocks CANNOT be served!
  
  PagedAttention:
    ELIMINATED! Blocks are non-contiguous by design.
    Any free block can be used by any sequence.
    If there are 4 free blocks, ANY sequence needing ≤4 blocks can be served.

INTERNAL FRAGMENTATION:
  Definition: Memory is allocated but not fully used within a block.
  
  In PagedAttention:
    Only the LAST block of each sequence may have waste.
    Waste = block_size - (seq_len % block_size)
    Maximum waste per sequence: block_size - 1 tokens
    
    Example (block_size=16):
    Sequence with 33 tokens:
      Block 0: 16/16 tokens (full)
      Block 1: 16/16 tokens (full)
      Block 2: 1/16 tokens (15 wasted!)
      Internal fragmentation: 15/48 = 31%
    
    Sequence with 48 tokens:
      Block 0: 16/16 (full)
      Block 1: 16/16 (full)  
      Block 2: 16/16 (full)
      Internal fragmentation: 0%

RESERVATION WASTE:
  Because we don't know how long generation will be,
  traditional systems reserve max_seq_len for every sequence.
  Average generation < max_seq_len → most reserved memory wasted.
  
  PagedAttention: Blocks allocated ON DEMAND.
  No reservation waste at all!
"""
print(fragmentation)
```

## 10. Block Size Trade-offs

```python
# Analyze the impact of block size

block_sizes = [1, 4, 8, 16, 32, 64, 128]
seq_lengths = actual_lengths  # From previous cell

results = []
for bs in block_sizes:
    blocks_needed = np.sum(np.ceil(seq_lengths / bs).astype(int))
    total_allocated = blocks_needed * bs
    internal_frag = total_allocated - np.sum(seq_lengths)
    frag_pct = internal_frag / total_allocated * 100
    
    results.append({
        'block_size': bs,
        'total_blocks': blocks_needed,
        'allocated': total_allocated,
        'frag_pct': frag_pct,
        'overhead': blocks_needed,  # Block table entries
    })

print(f"{'Block Size':<12} {'Blocks':<10} {'Allocated':<12} {'Frag %':<10} {'Block Table Size':<18}")
print("=" * 62)
for r in results:
    print(f"{r['block_size']:<12} {r['total_blocks']:<10} {r['allocated']:<12,} "
          f"{r['frag_pct']:<10.1f} {r['overhead'] * num_sequences:<18}")

print("\nTrade-off Analysis:")
print("  Small blocks (1-4): Near-zero fragmentation, but huge block tables")
print("                      and more metadata overhead")
print("  Medium blocks (16): Good balance, vLLM default")
print("  Large blocks (64+): Less metadata, but more internal fragmentation")
print("                      Also affects GPU kernel efficiency")
```

### Exercise 2: Implement Block Table Operations

```python
# Exercise 2: Implement a simple block allocator with reference counting

class BlockAllocator:
    """Simple block allocator with reference counting for CoW."""
    
    def __init__(self, num_blocks):
        self.num_blocks = num_blocks
        self.free_blocks = list(range(num_blocks))
        self.ref_counts = [0] * num_blocks
    
    def allocate(self) -> int:
        if not self.free_blocks:
            raise RuntimeError("Out of blocks!")
        block = self.free_blocks.pop(0)
        self.ref_counts[block] = 1
        return block
    
    def free(self, block: int):
        self.ref_counts[block] -= 1
        if self.ref_counts[block] == 0:
            self.free_blocks.append(block)
    
    def increase_ref(self, block: int):
        """Increase reference count (for sharing)."""
        self.ref_counts[block] += 1
    
    def cow_copy(self, block: int) -> int:
        """Copy-on-Write: create a copy if shared."""
        if self.ref_counts[block] == 1:
            return block  # Not shared, no copy needed
        
        # Need to copy
        new_block = self.allocate()
        self.ref_counts[block] -= 1  # Old block loses one reference
        print(f"    CoW: copied block {block} → {new_block} "
              f"(old ref_count now {self.ref_counts[block]})")
        return new_block
    
    @property
    def num_free(self):
        return len(self.free_blocks)

# Demo: Beam search with CoW
print("Beam Search with Copy-on-Write")
print("=" * 50)

allocator = BlockAllocator(num_blocks=10)
print(f"Initial free blocks: {allocator.num_free}")

# Step 1: Allocate prompt block (shared by all beams)
prompt_block = allocator.allocate()
print(f"\n1. Allocated prompt block: P{prompt_block}")

# Step 2: Create 3 beams sharing the prompt block
beam_tables = {
    0: [prompt_block],
    1: [prompt_block],
    2: [prompt_block],
}
allocator.increase_ref(prompt_block)  # ref: 1 → 2
allocator.increase_ref(prompt_block)  # ref: 2 → 3
print(f"\n2. Shared prompt block among 3 beams")
print(f"   ref_count(P{prompt_block}) = {allocator.ref_counts[prompt_block]}")
print(f"   Free blocks: {allocator.num_free}")

# Step 3: Each beam generates a token (needs new block or CoW)
print(f"\n3. Each beam generates a different token:")
for beam_id in range(3):
    new_block = allocator.allocate()
    beam_tables[beam_id].append(new_block)
    print(f"   Beam {beam_id}: block_table = {beam_tables[beam_id]}, "
          f"new block P{new_block}")

print(f"\n   Free blocks: {allocator.num_free}")
print(f"   Total blocks used: {10 - allocator.num_free}")
print(f"   Without CoW would need: {3 + 3} = 6 blocks (3 prompt copies + 3 decode)")
print(f"   With CoW: {1 + 3} = 4 blocks (1 shared prompt + 3 decode)")
print(f"   Savings: {6 - 4} blocks ({(6-4)/6:.0%})")
```

## Summary

Key takeaways:

1. **PagedAttention uses virtual memory concepts** to manage KV cache, eliminating external fragmentation
2. **Logical blocks** (per-sequence) map to **physical blocks** (GPU memory) via **block tables**
3. **Memory savings** are dramatic: typically 60-80% less memory than static allocation
4. **Copy-on-Write** enables efficient beam search by sharing prompt KV cache blocks
5. **PagedAttention v1** processes blocks sequentially; **v2** parallelizes across blocks
6. **Block size** is a trade-off: smaller = less fragmentation but more overhead
7. **Internal fragmentation** is the only remaining waste, bounded by block_size - 1 per sequence
