## 1. The Scheduler's Role

The Scheduler is the **brain** of vLLM. It decides:
- **Which** requests to run in each step
- **When** to admit new requests
- **What** to preempt when GPU memory is full
- **How** to manage the KV cache block budget

The scheduler runs entirely on the CPU and makes decisions that the GPU workers then execute.

```
Scheduler's Decision Space:
═══════════════════════════

Input: Three queues of SequenceGroups
  ┌─────────┐   ┌─────────┐   ┌─────────┐
  │ WAITING  │   │ RUNNING │   │ SWAPPED │
  │ (new)    │   │ (active)│   │ (on CPU)│
  └─────────┘   └─────────┘   └─────────┘

Output: SchedulerOutputs
  - Which sequences to process this step
  - Which blocks to swap in/out/copy
  - Token budget accounting
```

## 2. Sequence and SequenceGroup Data Structures

````python
# Source: vllm/sequence.py (detailed walkthrough)

sequence_structures = '''
class SequenceData:
    """The actual token data for a sequence."""
    
    def __init__(self, prompt_token_ids: List[int]):
        self.prompt_token_ids = prompt_token_ids
        self.output_token_ids: List[int] = []
        self.cumulative_logprob: float = 0.0
    
    def get_len(self) -> int:
        """Total length: prompt + output tokens."""
        return len(self.prompt_token_ids) + len(self.output_token_ids)
    
    def get_prompt_len(self) -> int:
        return len(self.prompt_token_ids)
    
    def get_output_len(self) -> int:
        return len(self.output_token_ids)
    
    def append_token_id(self, token_id: int, logprob: float):
        """Append a generated token."""
        self.output_token_ids.append(token_id)
        self.cumulative_logprob += logprob


class Sequence:
    """Represents a single sequence in the system."""
    
    def __init__(self, seq_id, prompt, prompt_token_ids, block_size, ...):
        self.seq_id = seq_id
        self.prompt = prompt
        self.block_size = block_size
        
        # Token data
        self.data = SequenceData(prompt_token_ids)
        
        # Status tracking
        self.status = SequenceStatus.WAITING
        
        # Logical token blocks
        # These are mapped to physical blocks by the BlockManager
        self.logical_token_blocks: List[LogicalTokenBlock] = []
        self._append_tokens_to_blocks(prompt_token_ids)
    
    def get_num_new_tokens(self) -> int:
        """How many tokens need to be computed in the next step."""
        if self.data.stage == SequenceStage.PREFILL:
            return self.data.get_prompt_len()  # All prompt tokens
        return 1  # One new token in decode phase


class SequenceGroup:
    """A group of sequences from the same request.
    
    Why groups?
    - Beam search: multiple candidate sequences from one prompt
    - Best-of-N: sample N times, return best
    - Simple generation: group with 1 sequence
    """
    
    def __init__(self, request_id, seqs, sampling_params, arrival_time, ...):
        self.request_id = request_id
        self.seqs_dict = {seq.seq_id: seq for seq in seqs}
        self.sampling_params = sampling_params
        self.arrival_time = arrival_time  # CRITICAL for FCFS scheduling
        self.lora_request = lora_request
    
    def get_seqs(self, status=None) -> List[Sequence]:
        """Get sequences, optionally filtered by status."""
        if status is None:
            return list(self.seqs_dict.values())
        return [s for s in self.seqs_dict.values() if s.status == status]
    
    def num_seqs(self, status=None) -> int:
        return len(self.get_seqs(status))
    
    def is_finished(self) -> bool:
        return all(seq.is_finished() for seq in self.seqs_dict.values())
'''
print(sequence_structures)
````

````python
queue_details = """
Queue Details:
══════════════

1. WAITING Queue:
   - Contains: SequenceGroups that have been added but not yet started
   - Sorted by: arrival_time (FCFS) or priority
   - Entry: add_request() or preemption via recompute
   - Exit: moved to RUNNING when scheduled
   - Sequences have status: SequenceStatus.WAITING
   - No KV cache blocks allocated

2. RUNNING Queue:
   - Contains: SequenceGroups currently being processed
   - These have KV cache blocks allocated on GPU
   - Each step, all running sequences get one decode step
   - Entry: from WAITING (new) or SWAPPED (swap-in)
   - Exit: FINISHED (done) or SWAPPED/WAITING (preempted)
   - Sequences have status: SequenceStatus.RUNNING

3. SWAPPED Queue:
   - Contains: SequenceGroups whose KV cache was swapped to CPU
   - Their KV cache blocks exist on CPU RAM, not GPU
   - Entry: from RUNNING when preempted (swap strategy)
   - Exit: back to RUNNING when space becomes available
   - Sequences have status: SequenceStatus.SWAPPED
   - Swapping back in requires GPU block allocation + data transfer
"""
print(queue_details)
````

## 4. Scheduling Policies

````python
# Source: vllm/core/policy.py

policy_code = '''
class Policy:
    """Base class for scheduling policies."""
    
    def get_priority(
        self,
        now: float,
        seq_group: SequenceGroup,
    ) -> float:
        """Return priority value. Lower = higher priority."""
        raise NotImplementedError
    
    def sort_by_priority(
        self,
        now: float,
        seq_groups: List[SequenceGroup],
    ) -> List[SequenceGroup]:
        """Sort sequence groups by priority."""
        return sorted(
            seq_groups,
            key=lambda sg: self.get_priority(now, sg),
        )


class FCFS(Policy):
    """First-Come-First-Served policy.
    
    Priority is simply the arrival time.
    Earlier arrival = lower value = higher priority.
    """
    
    def get_priority(
        self,
        now: float,
        seq_group: SequenceGroup,
    ) -> float:
        return seq_group.arrival_time


# The default policy is FCFS
# This means:
# - Requests are processed in order of arrival
# - If we need to preempt, we preempt the LATEST arrivals
# - This provides fairness: no request starves
'''
print(policy_code)
````

## 5. The _schedule() Method — Detailed Walkthrough

This is the most complex and important method in the scheduler. Let's walk through it step by step.

````python
# Source: vllm/core/scheduler.py
# The schedule() public method calls _schedule() internally

schedule_overview = '''
def schedule(self) -> Tuple[List[SequenceGroupMetadata], SchedulerOutputs]:
    """Schedule sequence groups for the current step.
    
    Returns:
        - seq_group_metadata_list: metadata for the executor
        - scheduler_outputs: scheduling decisions
    """
    scheduler_outputs = self._schedule()
    
    # Build metadata for each scheduled sequence group
    seq_group_metadata_list = []
    for scheduled in scheduler_outputs.scheduled_seq_groups:
        seq_group = scheduled.seq_group
        
        # Get the block tables from the block manager
        # Block tables map logical block index → physical block number
        seq_data = {}
        block_tables = {}
        for seq in seq_group.get_seqs(status=SequenceStatus.RUNNING):
            seq_data[seq.seq_id] = seq.data
            block_tables[seq.seq_id] = (
                self.block_manager.get_block_table(seq)
            )
        
        metadata = SequenceGroupMetadata(
            request_id=seq_group.request_id,
            is_prompt=(scheduled.token_chunk_size > 1),
            seq_data=seq_data,
            sampling_params=seq_group.sampling_params,
            block_tables=block_tables,
            token_chunk_size=scheduled.token_chunk_size,
        )
        seq_group_metadata_list.append(metadata)
    
    return seq_group_metadata_list, scheduler_outputs
'''
print(schedule_overview)
````

````python
# The internal _schedule() method (simplified but detailed)

schedule_internal = '''
def _schedule(self) -> SchedulerOutputs:
    """The core scheduling logic.
    
    Priority order:
    1. Running sequences (keep them going)
    2. Swapped sequences (resume them)
    3. Waiting sequences (start new ones)
    
    This order ensures that we don't starve running sequences
    by admitting too many new ones.
    """
    
    # Initialize the budget
    budget = SchedulingBudget(
        token_budget=self.scheduler_config.max_num_batched_tokens,
        max_num_seqs=self.scheduler_config.max_num_seqs,
    )
    
    # Track scheduling decisions
    blocks_to_swap_in = {}   # CPU block → GPU block
    blocks_to_swap_out = {}  # GPU block → CPU block
    blocks_to_copy = {}      # GPU block → GPU block (CoW)
    scheduled_seq_groups = []
    preempted_seq_groups = []
    
    # ═══════════════════════════════════════════
    # PHASE 1: Handle RUNNING sequences
    # ═══════════════════════════════════════════
    # These are sequences already in progress.
    # We MUST handle them first to ensure continuity.
    
    running_scheduled, running_preempted = (
        self._schedule_running(budget, blocks_to_swap_out, blocks_to_copy)
    )
    scheduled_seq_groups.extend(running_scheduled)
    preempted_seq_groups.extend(running_preempted)
    
    # ═══════════════════════════════════════════
    # PHASE 2: Handle SWAPPED sequences
    # ═══════════════════════════════════════════
    # Try to swap back in sequences that were previously preempted.
    # Only if there's remaining budget.
    
    if not preempted_seq_groups:  # Don't swap in if we just preempted
        swapped_in = self._schedule_swapped(
            budget, blocks_to_swap_in
        )
        scheduled_seq_groups.extend(swapped_in)
    
    # ═══════════════════════════════════════════
    # PHASE 3: Handle WAITING sequences
    # ═══════════════════════════════════════════
    # Admit new sequences if there's remaining budget.
    # Only if we didn't just preempt or swap in.
    
    if not preempted_seq_groups and not swapped_in:
        waiting_scheduled = self._schedule_waiting(
            budget
        )
        scheduled_seq_groups.extend(waiting_scheduled)
    
    return SchedulerOutputs(
        scheduled_seq_groups=scheduled_seq_groups,
        blocks_to_swap_in=blocks_to_swap_in,
        blocks_to_swap_out=blocks_to_swap_out,
        blocks_to_copy=blocks_to_copy,
        num_batched_tokens=budget.num_batched_tokens,
        preempted=len(preempted_seq_groups),
    )
'''
print(schedule_internal)
````

### 5.1 Phase 1: Scheduling Running Sequences

````python
schedule_running = '''
def _schedule_running(
    self,
    budget: SchedulingBudget,
    blocks_to_swap_out: Dict[int, int],
    blocks_to_copy: Dict[int, int],
) -> Tuple[List, List]:
    """Schedule the running sequences.
    
    For each running sequence:
    1. Try to allocate one more block (if needed) for the next token
    2. If allocation fails → preempt something
    3. Handle copy-on-write for beam search
    """
    scheduled = []
    preempted = []
    
    # Sort running queue by priority (FCFS: earliest first)
    # We process highest-priority first, and preempt lowest-priority
    self.running = self.policy.sort_by_priority(
        time.monotonic(), self.running
    )
    
    while self.running:
        seq_group = self.running[0]
        
        # Can we allocate blocks for the next token?
        num_running_seqs = seq_group.num_seqs(status=SequenceStatus.RUNNING)
        
        # Check if we can append a slot (one more token)
        while not self._can_append_slots(seq_group):
            # NOT ENOUGH GPU MEMORY!
            # We need to preempt something.
            
            if self.running:
                # Preempt the LOWEST priority running sequence
                victim = self.running.pop()  # Last = lowest priority
                self._preempt(victim, blocks_to_swap_out)
                preempted.append(victim)
            else:
                # Nothing left to preempt!
                # This shouldn't happen in practice
                break
        
        # Allocate the new slot
        self._append_slots(seq_group, blocks_to_copy)
        
        # Budget check
        if not budget.can_schedule(
            num_new_tokens=num_running_seqs,  # 1 token per seq
            num_new_seqs=0,  # Already running, not new
        ):
            break
        
        budget.add(
            num_batched_tokens=num_running_seqs,
            num_curr_seqs=num_running_seqs,
        )
        
        scheduled.append(ScheduledSequenceGroup(
            seq_group=seq_group,
            token_chunk_size=1,  # Decode: 1 token per sequence
        ))
        self.running.pop(0)
    
    return scheduled, preempted
'''
print(schedule_running)
````

### 5.2 Phase 2: Scheduling Swapped Sequences

````python
schedule_swapped = '''
def _schedule_swapped(
    self,
    budget: SchedulingBudget,
    blocks_to_swap_in: Dict[int, int],
) -> List:
    """Try to resume swapped-out sequences.
    
    A sequence is swapped when its KV cache was moved from GPU to CPU.
    To resume it, we need to:
    1. Allocate new GPU blocks
    2. Swap the KV cache data from CPU → GPU
    3. Move the sequence back to the running queue
    """
    scheduled = []
    
    # Sort swapped queue by priority
    self.swapped = self.policy.sort_by_priority(
        time.monotonic(), self.swapped
    )
    
    while self.swapped:
        seq_group = self.swapped[0]
        
        # Can we allocate GPU blocks for this sequence?
        if not self._can_swap_in(seq_group):
            break  # Not enough GPU memory
        
        # Budget check
        num_seqs = seq_group.num_seqs(status=SequenceStatus.SWAPPED)
        if not budget.can_schedule(
            num_new_tokens=num_seqs,  # 1 decode token per seq
            num_new_seqs=num_seqs,
        ):
            break
        
        # Swap in: allocate GPU blocks and record the mapping
        swap_mapping = self._swap_in(seq_group)
        blocks_to_swap_in.update(swap_mapping)
        # swap_mapping: {cpu_block_id: gpu_block_id}
        # The worker will use this to copy data from CPU to GPU
        
        # Move from swapped → running
        self.swapped.pop(0)
        self._append_slots(seq_group, {})  # Allocate slot for next token
        
        # Update status
        for seq in seq_group.get_seqs(status=SequenceStatus.SWAPPED):
            seq.status = SequenceStatus.RUNNING
        
        budget.add(num_batched_tokens=num_seqs, num_curr_seqs=num_seqs)
        scheduled.append(ScheduledSequenceGroup(
            seq_group=seq_group,
            token_chunk_size=1,
        ))
    
    return scheduled
'''
print(schedule_swapped)
````

### 5.3 Phase 3: Scheduling Waiting Sequences

````python
schedule_waiting = '''
def _schedule_waiting(
    self,
    budget: SchedulingBudget,
) -> List:
    """Admit new sequences from the waiting queue.
    
    This is where new requests get their first chance to run.
    A waiting sequence needs a PREFILL step first.
    """
    scheduled = []
    
    # Sort waiting queue by priority (FCFS)
    self.waiting = self.policy.sort_by_priority(
        time.monotonic(), self.waiting
    )
    
    while self.waiting:
        seq_group = self.waiting[0]
        
        # How many tokens does this sequence need to process?
        # For prefill: all prompt tokens
        num_new_tokens = seq_group.get_seqs()[0].get_num_new_tokens()
        # Example: prompt "Hello world" = 2 tokens
        
        # Chunked prefill: may process only part of the prompt
        if self.scheduler_config.enable_chunked_prefill:
            remaining_budget = budget.remaining_token_budget()
            num_new_tokens = min(num_new_tokens, remaining_budget)
        
        # Budget check
        num_seqs = seq_group.num_seqs()
        if not budget.can_schedule(
            num_new_tokens=num_new_tokens,
            num_new_seqs=num_seqs,
        ):
            break  # Budget exhausted
        
        # Can we allocate blocks for this sequence?
        if not self._can_allocate(seq_group):
            break  # Not enough GPU memory
        
        # Allocate blocks
        self._allocate(seq_group)
        # This allocates physical blocks on the GPU
        # for the entire prompt length
        
        # Move from waiting → running
        self.waiting.pop(0)
        self.running.append(seq_group)
        
        # Update status
        for seq in seq_group.get_seqs():
            seq.status = SequenceStatus.RUNNING
        
        budget.add(
            num_batched_tokens=num_new_tokens,
            num_curr_seqs=num_seqs,
        )
        scheduled.append(ScheduledSequenceGroup(
            seq_group=seq_group,
            token_chunk_size=num_new_tokens,  # All prompt tokens
        ))
    
    return scheduled
'''
print(schedule_waiting)
````

## 7. Preemption Logic

````python
preemption_code = '''
def _preempt(
    self,
    seq_group: SequenceGroup,
    blocks_to_swap_out: Dict[int, int],
) -> PreemptionMode:
    """Preempt a sequence group to free GPU memory.
    
    Two preemption strategies:
    
    1. RECOMPUTE: Simply discard the KV cache.
       - Free all GPU blocks immediately
       - Move sequence back to WAITING queue
       - The sequence will need to be fully recomputed later
       - Pros: Frees memory instantly, no CPU swap space needed
       - Cons: Wasted computation (must redo prefill)
       - Used when: Sequence has only 1 sequence (no beam search)
    
    2. SWAP: Move KV cache from GPU to CPU.
       - Allocate CPU blocks
       - Copy KV cache data GPU → CPU
       - Free GPU blocks
       - Move sequence to SWAPPED queue
       - Later: swap back in (CPU → GPU)
       - Pros: No wasted computation
       - Cons: Requires CPU memory, swap overhead
       - Used when: Beam search (multiple sequences share blocks)
    """
    
    # Choose preemption mode
    if seq_group.num_seqs() == 1:
        # Simple case: single sequence
        # RECOMPUTE is simpler and doesn't need CPU swap space
        preemption_mode = PreemptionMode.RECOMPUTE
    else:
        # Beam search: multiple sequences share block prefixes
        # Must SWAP to preserve the shared state
        preemption_mode = PreemptionMode.SWAP
    
    if preemption_mode == PreemptionMode.RECOMPUTE:
        # Free all blocks and move to waiting
        self._preempt_by_recompute(seq_group)
    else:
        # Swap out and move to swapped
        self._preempt_by_swap(seq_group, blocks_to_swap_out)
    
    return preemption_mode


def _preempt_by_recompute(self, seq_group):
    """Preempt by discarding KV cache."""
    seqs = seq_group.get_seqs(status=SequenceStatus.RUNNING)
    assert len(seqs) == 1  # Only for single-sequence groups
    
    seq = seqs[0]
    # Free GPU blocks
    self.block_manager.free(seq)
    # Reset sequence to initial state
    seq.status = SequenceStatus.WAITING
    # Discard generated tokens (will need recomputation)
    seq.reset_state_for_recompute()
    # Move back to waiting queue
    self.waiting.insert(0, seq_group)  # Insert at front (high priority)


def _preempt_by_swap(self, seq_group, blocks_to_swap_out):
    """Preempt by swapping KV cache to CPU."""
    # Allocate CPU blocks and get the mapping
    swap_mapping = self.block_manager.swap_out(seq_group)
    blocks_to_swap_out.update(swap_mapping)
    # swap_mapping: {gpu_block_id: cpu_block_id}
    
    # Update status
    for seq in seq_group.get_seqs(status=SequenceStatus.RUNNING):
        seq.status = SequenceStatus.SWAPPED
    
    # Move to swapped queue
    self.swapped.append(seq_group)
'''
print(preemption_code)
````

## 9. Chunked Prefill

Chunked prefill is an important optimization that prevents long prompts from blocking other requests.

````python
chunked_prefill = """
Chunked Prefill Explained
═════════════════════════

WITHOUT Chunked Prefill:
  If a request has a 4000-token prompt and max_num_batched_tokens=4096:
  - The entire prompt must be processed in ONE step
  - This step uses 4000/4096 of the token budget
  - Only 96 tokens left for other decode steps
  - All running decode sequences effectively stall!

  Step 1: [────── 4000 token prefill ──────][96 decode]
  → Very long step, running sequences delayed

WITH Chunked Prefill:
  The 4000-token prompt is split into chunks:
  Step 1: [── 1000 tokens ──][────── 3096 decode tokens ──────]
  Step 2: [── 1000 tokens ──][────── 3096 decode tokens ──────]
  Step 3: [── 1000 tokens ──][────── 3096 decode tokens ──────]
  Step 4: [── 1000 tokens ──][────── 3096 decode tokens ──────]
  → Running sequences keep making progress!

Trade-off:
  + Lower latency for running sequences (not blocked)
  + More predictable step times
  - Higher TTFT for the chunked prefill request (4 steps vs 1)
  - Slightly lower prefill throughput (context fragmentation)
"""
print(chunked_prefill)
````

## 10. Scheduler Simulation

````python
import random
from dataclasses import dataclass, field
from typing import List, Dict, Tuple, Optional
from enum import Enum
import time

class Status(Enum):
    WAITING = 0
    RUNNING = 1
    SWAPPED = 2
    FINISHED = 3

@dataclass
class SimSequence:
    seq_id: int
    prompt_len: int
    max_tokens: int
    generated: int = 0
    status: Status = Status.WAITING
    blocks_allocated: int = 0
    arrival_time: float = 0.0
    
    @property
    def total_tokens(self):
        return self.prompt_len + self.generated
    
    @property
    def blocks_needed(self):
        """How many blocks needed for current state."""
        return (self.total_tokens + 15) // 16  # block_size = 16


class SimScheduler:
    def __init__(self, total_gpu_blocks=100, max_num_seqs=8, max_tokens_per_step=512):
        self.total_gpu_blocks = total_gpu_blocks
        self.used_blocks = 0
        self.max_num_seqs = max_num_seqs
        self.max_tokens_per_step = max_tokens_per_step
        
        self.waiting: List[SimSequence] = []
        self.running: List[SimSequence] = []
        self.swapped: List[SimSequence] = []
        self.finished: List[SimSequence] = []
        
        self.preemption_count = 0
        self.seq_counter = 0
    
    @property
    def free_blocks(self):
        return self.total_gpu_blocks - self.used_blocks
    
    def add_request(self, prompt_len: int, max_tokens: int):
        seq = SimSequence(
            seq_id=self.seq_counter,
            prompt_len=prompt_len,
            max_tokens=max_tokens,
            arrival_time=time.monotonic(),
        )
        self.seq_counter += 1
        self.waiting.append(seq)
        return seq.seq_id
    
    def schedule(self) -> Dict:
        """Simplified three-phase scheduling."""
        token_budget = self.max_tokens_per_step
        seq_budget = self.max_num_seqs
        scheduled = []
        preempted = []
        
        # Phase 1: Schedule running sequences
        still_running = []
        for seq in sorted(self.running, key=lambda s: s.arrival_time):
            needed = seq.blocks_needed - seq.blocks_allocated
            if needed > 0 and self.free_blocks < needed:
                # Need to preempt!
                self.used_blocks -= seq.blocks_allocated
                seq.blocks_allocated = 0
                seq.status = Status.WAITING
                seq.generated = 0  # Recompute mode
                self.waiting.insert(0, seq)
                preempted.append(seq)
                self.preemption_count += 1
                continue
            
            if needed > 0:
                self.used_blocks += needed
                seq.blocks_allocated += needed
            
            if token_budget >= 1 and seq_budget >= 1:
                token_budget -= 1
                seq_budget -= 1
                scheduled.append(('decode', seq))
                still_running.append(seq)
            else:
                still_running.append(seq)
        self.running = still_running
        
        # Phase 2: Schedule waiting (skip if preempted)
        if not preempted:
            still_waiting = []
            for seq in self.waiting:
                blocks_needed = seq.blocks_needed
                if (token_budget >= seq.prompt_len 
                    and seq_budget >= 1
                    and self.free_blocks >= blocks_needed):
                    self.used_blocks += blocks_needed
                    seq.blocks_allocated = blocks_needed
                    seq.status = Status.RUNNING
                    token_budget -= seq.prompt_len
                    seq_budget -= 1
                    scheduled.append(('prefill', seq))
                    self.running.append(seq)
                else:
                    still_waiting.append(seq)
            self.waiting = still_waiting
        
        return {
            'scheduled': scheduled,
            'preempted': preempted,
            'token_count': self.max_tokens_per_step - token_budget,
        }
    
    def process_outputs(self, scheduled):
        """Process outputs: generate tokens and check completion."""
        newly_finished = []
        for phase, seq in scheduled:
            seq.generated += 1
            if seq.generated >= seq.max_tokens:
                seq.status = Status.FINISHED
                self.running.remove(seq)
                self.used_blocks -= seq.blocks_allocated
                seq.blocks_allocated = 0
                self.finished.append(seq)
                newly_finished.append(seq)
        return newly_finished


# Run simulation
scheduler = SimScheduler(total_gpu_blocks=20, max_num_seqs=4, max_tokens_per_step=256)

# Add several requests with different sizes
scheduler.add_request(prompt_len=32, max_tokens=10)   # Small
scheduler.add_request(prompt_len=64, max_tokens=8)    # Medium
scheduler.add_request(prompt_len=48, max_tokens=12)   # Medium
scheduler.add_request(prompt_len=128, max_tokens=5)   # Large prompt
scheduler.add_request(prompt_len=16, max_tokens=15)   # Small, long output

print(f"Initial state: {len(scheduler.waiting)} waiting, {scheduler.total_gpu_blocks} GPU blocks")
print("=" * 70)

step = 0
while scheduler.waiting or scheduler.running:
    step += 1
    result = scheduler.schedule()
    
    if not result['scheduled']:
        break
    
    finished = scheduler.process_outputs(result['scheduled'])
    
    phases = [f"{p}(seq{s.seq_id})" for p, s in result['scheduled']]
    print(f"Step {step:2d}: {', '.join(phases)}")
    print(f"         blocks: {scheduler.used_blocks}/{scheduler.total_gpu_blocks}, "
          f"running: {len(scheduler.running)}, waiting: {len(scheduler.waiting)}")
    if result['preempted']:
        print(f"         PREEMPTED: {[s.seq_id for s in result['preempted']]}")
    if finished:
        print(f"         FINISHED: {[s.seq_id for s in finished]}")

print(f"\nTotal steps: {step}")
print(f"Total preemptions: {scheduler.preemption_count}")
print(f"Finished: {len(scheduler.finished)} requests")
````

## 12. Interaction with Block Manager

````python
block_manager_interaction = """
Scheduler ↔ Block Manager Interaction
══════════════════════════════════════

The scheduler calls these BlockManager methods:

1. can_allocate(seq_group) → bool
   "Can I allocate blocks for this new sequence?"
   Checks: enough free GPU blocks for the prompt length?

2. allocate(seq_group)
   "Allocate blocks for this sequence."
   Maps logical blocks → physical blocks on GPU

3. can_append_slots(seq_group) → bool
   "Can I allocate one more slot for the next token?"
   Checks: current block has space, or a free block exists?

4. append_slots(seq, blocks_to_copy)
   "Allocate slot for the next token."
   May allocate a new block if current one is full.
   May trigger copy-on-write for shared blocks.

5. free(seq)
   "Free all blocks for this sequence."
   Called when sequence finishes or is preempted (recompute).

6. swap_out(seq_group) → Dict[int, int]
   "Move KV cache from GPU to CPU."
   Returns: {gpu_block → cpu_block} mapping.

7. swap_in(seq_group) → Dict[int, int]
   "Move KV cache from CPU back to GPU."
   Returns: {cpu_block → gpu_block} mapping.

8. get_block_table(seq) → List[int]
   "Get the list of physical block IDs for this sequence."
   Used to construct attention metadata for the worker.
"""
print(block_manager_interaction)
````

````python
# Exercise 1: Shortest Job First Policy

class ShortestJobFirst:
    """Priority based on expected output length.
    
    Sequences with fewer remaining tokens to generate
    get higher priority (lower priority value).
    """
    
    def get_priority(self, seq):
        """Priority = remaining tokens to generate."""
        remaining = seq.max_tokens - seq.generated
        return remaining  # Lower = higher priority
    
    def sort_by_priority(self, seq_list):
        return sorted(seq_list, key=lambda s: self.get_priority(s))

# Demo: Compare FCFS vs SJF
from dataclasses import dataclass

@dataclass
class DemoSeq:
    name: str
    max_tokens: int
    generated: int
    arrival_time: float

sequences = [
    DemoSeq("Long story", max_tokens=500, generated=100, arrival_time=1.0),
    DemoSeq("Short answer", max_tokens=20, generated=15, arrival_time=2.0),
    DemoSeq("Medium task", max_tokens=100, generated=50, arrival_time=3.0),
    DemoSeq("Almost done", max_tokens=30, generated=28, arrival_time=4.0),
]

# FCFS order
print("FCFS Order (by arrival time):")
fcfs_order = sorted(sequences, key=lambda s: s.arrival_time)
for i, seq in enumerate(fcfs_order, 1):
    print(f"  {i}. {seq.name} (remaining: {seq.max_tokens - seq.generated})")

print()

# SJF order
sjf = ShortestJobFirst()
print("SJF Order (by remaining tokens):")
sjf_order = sjf.sort_by_priority(sequences)
for i, seq in enumerate(sjf_order, 1):
    print(f"  {i}. {seq.name} (remaining: {seq.max_tokens - seq.generated})")

print()
print("SJF Advantage: Short requests finish faster, improving average latency.")
print("SJF Disadvantage: Long requests may be starved if short requests keep arriving.")
````

## Summary

Key takeaways from this chapter:

1. **Three queues**: WAITING (new) → RUNNING (active) → FINISHED, with SWAPPED as a side queue for preempted sequences
2. **Three scheduling phases**: Handle running first, then swapped, then waiting. This priority order ensures continuity.
3. **Budget management**: Each step has a token budget and sequence budget. Scheduling stops when either is exhausted.
4. **Preemption**: When GPU memory is insufficient, the scheduler preempts low-priority sequences via RECOMPUTE (discard KV cache) or SWAP (move to CPU).
5. **FCFS policy**: Default is first-come-first-served, but custom policies can be implemented.
6. **Chunked prefill**: Splits long prefills across steps to prevent blocking decode sequences.
